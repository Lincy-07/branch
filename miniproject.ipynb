```python
import mediapipe as mp
from mediapipe.tasks import python
from mediapipe.tasks.python import vision
import cv2
import numpy as np
import os

# 1. Setup the Detector
model_path = 'pose_landmarker.task'
BaseOptions = mp.tasks.BaseOptions
PoseLandmarker = mp.tasks.vision.PoseLandmarker
PoseLandmarkerOptions = mp.tasks.vision.PoseLandmarkerOptions
VisionRunningMode = mp.tasks.vision.RunningMode

options = PoseLandmarkerOptions(
    base_options=BaseOptions(model_asset_path=model_path),
    running_mode=VisionRunningMode.VIDEO
)

DATA_PATH = "/content/drive/MyDrive/similar_dataset"
actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

X_data = [] # Landmarks
Y_data = [] # Labels

with PoseLandmarker.create_from_options(options) as landmarker:
    for idx, action in enumerate(actions):
        action_path = os.path.join(DATA_PATH, action)
        video_files = [f for f in os.listdir(action_path) if f.endswith(('.mp4', '.avi'))]

        print(f"Processing {action}...")

        for video in video_files[:15]: # Start with 15 videos per class
            cap = cv2.VideoCapture(os.path.join(action_path, video))
            while cap.isOpened():
                ret, frame = cap.read()
                if not ret: break

                # Convert to MediaPipe Image
                mp_image = mp.Image(image_format=mp.ImageFormat.SRGB, data=cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
                timestamp = int(cap.get(cv2.CAP_PROP_POS_MSEC))

                # Detect
                result = landmarker.detect_for_video(mp_image, timestamp)

                if result.pose_landmarks:
                    # Get the 33 landmarks (x, y, z)
                    frame_landmarks = []
                    for landmark in result.pose_landmarks[0]:
                        frame_landmarks.extend([landmark.x, landmark.y, landmark.z])
                    X_data.append(frame_landmarks)
                    Y_data.append(idx)
            cap.release()

X = np.array(X_data)
Y = np.array(Y_data)
print(f"Total frames processed: {X.shape[0]}")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Bidirectional, Dropout
# Updated import for the latest TensorFlow/Keras version
from tensorflow.keras.utils import to_categorical

# 1. Prepare Data
# Ensure X_data and Y_data from your extraction step are numpy arrays
X = np.array(X_data)
Y = np.array(Y_data)

# Reshape for BiLSTM (Samples, Time Steps, Features)
# We treat each frame as a single step for this basic training
X_train = X.reshape(X.shape[0], 1, 99)
Y_train = to_categorical(Y)

# 2. The BiLSTM Architecture
model = Sequential([
    # Input shape is (1 frame, 99 landmark coordinates)
    Bidirectional(LSTM(64, return_sequences=True, input_shape=(1, 99))),
    Bidirectional(LSTM(32)),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(len(actions), activation='softmax') # Should be 4 for your exercises
])

# 3. Compile and Train
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print("🚀 Starting training...")
model.fit(X_train, Y_train, epochs=30, batch_size=32)

# 4. Save your final model
model.save('fitness_trainer_model.h5')
print("🎉 SUCCESS! Your model is saved as 'fitness_trainer_model.h5'.")
```

```python
import cv2
import numpy as np
import tensorflow as tf
from mediapipe.python.solutions import pose as mp_pose

# 1. Load your trained model
model = tf.keras.models.load_model('fitness_trainer_model.h5')
pose_tracker = mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5)

# 2. Settings
actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']
counter = 0
status = None # "up" or "down"

# Helper to calculate joint angles for counting
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    if angle > 180.0: angle = 360-angle
    return angle

# 3. Process a Test Video
video_input = "/content/drive/MyDrive/similar_dataset/squat/test_video.mp4" # Path to a test video
cap = cv2.VideoCapture(video_input)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret: break

    # Landmark Extraction
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose_tracker.process(image)

    if results.pose_landmarks:
        landmarks = results.pose_landmarks.landmark

        # --- CLASSIFICATION (What exercise is this?) ---
        coords = np.array([[l.x, l.y, l.z] for l in landmarks]).flatten()
        prediction = model.predict(coords.reshape(1, 1, 99), verbose=0)
        exercise = actions[np.argmax(prediction)]

        # --- COUNTING LOGIC (Example for Squats) ---
        if exercise == 'squat':
            # Landmarks: Hip (24), Knee (26), Ankle (28)
            angle = calculate_angle([landmarks[24].x, landmarks[24].y],
                                    [landmarks[26].x, landmarks[26].y],
                                    [landmarks[28].x, landmarks[28].y])

            # State machine for counting
            if angle > 160: status = "down"
            if angle < 90 and status == "down":
                status = "up"
                counter += 1

        # --- VISUALS ---
        cv2.putText(frame, f'EXERCISE: {exercise}', (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
        cv2.putText(frame, f'REPS: {counter}', (10, 100), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 255, 0), 3)

# cap.release()
print(f"Analysis Finished! Final Count: {counter}")
```

```python
%%writefile app.py
import streamlit as st
import cv2
import numpy as np
import tensorflow as tf
from mediapipe.python.solutions import pose as mp_pose

# --- CONFIG ---
st.set_page_config(page_title="AI Fitness Trainer", layout="wide")
st.title("🏋️‍♂️ AI Virtual Fitness Trainer")

# --- LOAD BRAIN ---
@st.cache_resource
def load_model():
    return tf.keras.models.load_model('fitness_trainer_model.h5')

model = load_model()
pose_tracker = mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5)
actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

# --- SESSION STATE ---
if 'count' not in st.session_state: st.session_state.count = 0
if 'stage' not in st.session_state: st.session_state.stage = None

def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians*180.0/np.pi)
    if angle > 180.0: angle = 360-angle
    return angle

# --- UI ---
video_file = st.file_uploader("Upload your workout video", type=['mp4', 'mov', 'avi'])

if video_file:
    # Save temp file to read it
    with open("temp_video.mp4", "wb") as f:
        f.write(video_file.read())

    cap = cv2.VideoCapture("temp_video.mp4")
    frame_placeholder = st.empty()

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret: break

        # 1. AI Detection
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = pose_tracker.process(image)

        if results.pose_landmarks:
            landmarks = results.pose_landmarks.landmark
            coords = np.array([[l.x, l.y, l.z] for l in landmarks]).flatten()

            # 2. Predict Exercise
            prediction = model.predict(coords.reshape(1, 1, 99), verbose=0)
            exercise = actions[np.argmax(prediction)]

            # 3. Dynamic Angle Calculation
            angle = 0
            if exercise == 'squat':
                angle = calculate_angle([landmarks[24].x, landmarks[24].y], [landmarks[26].x, landmarks[26].y], [landmarks[28].x, landmarks[28].y])
            elif exercise == 'barbell biceps curl':
                angle = calculate_angle([landmarks[12].x, landmarks[12].y], [landmarks[14].x, landmarks[14].y], [landmarks[16].x, landmarks[16].y])

            # 4. Counting Logic
            if angle > 160: st.session_state.stage = "down"
            if angle < 90 and st.session_state.stage == "down":
                st.session_state.stage = "up"
                st.session_state.count += 1

            # Draw overlay
            cv2.putText(frame, f"Exercise: {exercise}", (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
            cv2.putText(frame, f"Reps: {st.session_state.count}", (10, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        frame_placeholder.image(frame, channels="BGR")
    cap.release()
```

```python
import cv2
import numpy as np
import os

# Ensure mediapipe is installed (run in a prior cell if needed)
# This import uses the stable direct path
from mediapipe.python.solutions import pose as mp_pose

print("✅ Direct Import Successful!")

# Initialize pose model
pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5)

DATA_PATH = "/content/drive/MyDrive/similar_dataset"
actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

X_data = []
Y_data = []

print("🚀 Extracting landmarks...")

for idx, action in enumerate(actions):
    action_path = os.path.join(DATA_PATH, action)
    if not os.path.exists(action_path):
        print(f"⚠️ Skipping {action} — folder not found.")
        continue

    video_files = [f for f in os.listdir(action_path) if f.endswith(('.mp4', '.avi'))]

    for video in video_files[:10]:
        cap = cv2.VideoCapture(os.path.join(action_path, video))
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = pose.process(image)
            if results.pose_landmarks:
                landmarks = np.array([[res.x, res.y, res.z] for res in results.pose_landmarks.landmark]).flatten()
                X_data.append(landmarks)
                Y_data.append(idx)
        cap.release()

pose.close()  # Explicitly close the pose model
print(f"✅ Collection Complete: {len(X_data)} frames found.")
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Bidirectional, Dropout
from tensorflow.keras.utils import to_categorical

# Defensive check
assert len(X_data) > 0, "❌ X_data is empty! Run the landmark extraction cell first."
assert len(set(Y_data)) == len(actions), f"❌ Expected {len(actions)} classes, got {len(set(Y_data))}."

X = np.array(X_data)
Y = np.array(Y_data)

# Reshape for BiLSTM: (Samples, Time Steps=1, Features=99)
X_train = X.reshape(X.shape[0], 1, 99)
Y_train = to_categorical(Y, num_classes=len(actions))

model = Sequential([
    Bidirectional(LSTM(64, return_sequences=True, input_shape=(1, 99))),
    Bidirectional(LSTM(32)),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(len(actions), activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

print(f"🚀 Training on {X_train.shape[0]} frames across {len(actions)} classes...")
model.fit(X_train, Y_train, epochs=30, batch_size=32, validation_split=0.1)

model.save('fitness_trainer_model.h5')
print("🎉 SUCCESS! Model saved as 'fitness_trainer_model.h5'.")
```

```python
%%writefile app.py
import streamlit as st
import cv2
import numpy as np
import tensorflow as tf
import os
import requests
from mediapipe.python.solutions import pose as mp_pose
from mediapipe.python.solutions import drawing_utils as mp_drawing

# ─────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────
st.set_page_config(page_title="AI Fitness Trainer", layout="wide")
st.title("🏋️‍♂️ AI Virtual Fitness Trainer")

actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

# ─────────────────────────────────────────
# LOAD MODEL
# ─────────────────────────────────────────
@st.cache_resource
def load_model():
    if not os.path.exists("fitness_trainer_model.h5"):
        st.error("❌ Model file not found! Train and save the model first.")
        st.stop()
    return tf.keras.models.load_model("fitness_trainer_model.h5")

@st.cache_resource
def load_pose():
    return mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5)

model = load_model()
pose_tracker = load_pose()

# ─────────────────────────────────────────
# SESSION STATE
# ─────────────────────────────────────────
for key, val in [("count", 0), ("stage", None), ("chat_history", [])]:
    if key not in st.session_state:
        st.session_state[key] = val

# ─────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    return 360 - angle if angle > 180.0 else angle

def get_angle_for_exercise(exercise, landmarks):
    if exercise == "squat":
        return calculate_angle(
            [landmarks[24].x, landmarks[24].y],
            [landmarks[26].x, landmarks[26].y],
            [landmarks[28].x, landmarks[28].y])
    elif exercise in ["barbell biceps curl", "push-up"]:
        return calculate_angle(
            [landmarks[12].x, landmarks[12].y],
            [landmarks[14].x, landmarks[14].y],
            [landmarks[16].x, landmarks[16].y])
    elif exercise == "shoulder press":
        return calculate_angle(
            [landmarks[14].x, landmarks[14].y],
            [landmarks[12].x, landmarks[12].y],
            [landmarks[24].x, landmarks[24].y])
    return 0

def process_frame(frame):
    """Run pose + classification + rep counting on one frame. Returns annotated frame."""
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose_tracker.process(image)

    exercise, confidence, angle = "Unknown", 0.0, 0

    if results.pose_landmarks:
        landmarks = results.pose_landmarks.landmark
        coords = np.array([[l.x, l.y, l.z] for l in landmarks]).flatten()

        prediction = model.predict(coords.reshape(1, 1, 99), verbose=0)
        exercise = actions[np.argmax(prediction)]
        confidence = float(np.max(prediction))
        angle = get_angle_for_exercise(exercise, landmarks)

        # Rep counting state machine
        if angle > 160:
            st.session_state.stage = "down"
        if angle < 90 and st.session_state.stage == "down":
            st.session_state.stage = "up"
            st.session_state.count += 1

        # Draw skeleton
        mp_drawing.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

    # Overlay text
    cv2.rectangle(frame, (0, 0), (400, 170), (0, 0, 0), -1)
    cv2.putText(frame, f"Exercise : {exercise}", (10, 40),  cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2)
    cv2.putText(frame, f"Confidence: {confidence:.0%}", (10, 80),  cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 255), 2)
    cv2.putText(frame, f"Angle    : {angle:.1f}",      (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (200, 200, 0), 2)
    cv2.putText(frame, f"Reps     : {st.session_state.count}", (10, 160), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    return frame, exercise, confidence

def chatbot_reply(user_msg):
    """Simple rule-based fitness chatbot."""
    msg = user_msg.lower()
    if any(w in msg for w in ["squat", "leg", "knee"]):
        return "🦵 For squats: keep your chest up, knees behind toes, and go below parallel for full range. Aim for 3×10 reps."
    elif any(w in msg for w in ["curl", "bicep", "arm"]):
        return "💪 For bicep curls: keep elbows pinned to your sides, squeeze at the top, and lower slowly. Use a weight you can control."
    elif any(w in msg for w in ["push", "chest", "tricep"]):
        return "🤸 For push-ups: keep your body in a straight line, hands shoulder-width apart, and lower your chest to the floor."
    elif any(w in msg for w in ["shoulder", "press", "overhead"]):
        return "🏋️ For shoulder press: brace your core, press straight up, and avoid flaring your elbows too wide."
    elif any(w in msg for w in ["rest", "recovery", "sleep"]):
        return "😴 Recovery is when muscles grow! Aim for 7-9 hours of sleep and 48hrs rest between training the same muscle group."
    elif any(w in msg for w in ["diet", "protein", "food", "eat"]):
        return "🥗 Aim for 1.6–2.2g of protein per kg of bodyweight. Prioritize whole foods, and eat in a slight caloric surplus to build muscle."
    elif any(w in msg for w in ["warm", "stretch"]):
        return "🔥 Always warm up for 5-10 minutes before lifting. Dynamic stretches (leg swings, arm circles) are better pre-workout than static holds."
    else:
        return "🤖 I'm your fitness assistant! Ask me about squats, push-ups, bicep curls, shoulder press, diet, recovery, or warm-ups."

# ─────────────────────────────────────────
# SIDEBAR
# ─────────────────────────────────────────
st.sidebar.header("🎛️ Controls")
if st.sidebar.button("🔄 Reset Rep Counter"):
    st.session_state.count = 0
    st.session_state.stage = None
    st.sidebar.success("Counter reset!")

st.sidebar.markdown("---")
st.sidebar.metric("Current Reps", st.session_state.count)
st.sidebar.markdown("---")
st.sidebar.markdown("**Supported Exercises:**")
for a in actions:
    st.sidebar.markdown(f"- {a.title()}")

# ─────────────────────────────────────────
# TABS
# ─────────────────────────────────────────
tab1, tab2, tab3 = st.tabs(["📹 Video Analysis", "📸 Webcam (Live)", "🤖 Fitness Chatbot"])

# ══════════════════════════════════════════
# TAB 1 — VIDEO UPLOAD & ANALYSIS
# ══════════════════════════════════════════
with tab1:
    st.subheader("📹 Upload a Workout Video")
    st.info("Upload any workout video. The AI will classify the exercise and count your reps automatically.")

    video_file = st.file_uploader("Choose a video file", type=["mp4", "mov", "avi"], key="video_upload")

    if video_file:
        with open("temp_video.mp4", "wb") as f:
            f.write(video_file.read())

        st.success("✅ Video uploaded! Processing...")

        cap = cv2.VideoCapture("temp_video.mp4")
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        fps = cap.get(cv2.CAP_PROP_FPS)

        frame_placeholder = st.empty()
        progress_bar = st.progress(0)
        stats_placeholder = st.empty()

        frame_idx = 0
        skip = 2  # Process every 2nd frame to speed up

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            if frame_idx % skip == 0:
                annotated, exercise, conf = process_frame(frame)
                frame_placeholder.image(annotated, channels="BGR", width=700)
                stats_placeholder.markdown(
                    f"**🏷️ Exercise:** {exercise} &nbsp;|&nbsp; "
                    f"**📊 Confidence:** {conf:.0%} &nbsp;|&nbsp; "
                    f"**🔢 Reps:** {st.session_state.count}"
                )

            progress_bar.progress(min(frame_idx / max(total_frames, 1), 1.0))
            frame_idx += 1

        cap.release()
        st.balloons()
        st.success(f"🎉 Analysis complete! Total reps counted: **{st.session_state.count}**")

# ══════════════════════════════════════════
# TAB 2 — WEBCAM (SNAPSHOT MODE)
# ══════════════════════════════════════════
with tab2:
    st.subheader("📸 Webcam — Auto Classify")
    st.info("Upload a photo from your webcam or phone camera. The AI will detect your pose and classify the exercise.")
    st.warning("⚠️ Note: Live webcam streaming is not supported in Colab. Use the snapshot mode below.")

    webcam_img = st.camera_input("📷 Take a snapshot of your exercise pose")

    if webcam_img:
        file_bytes = np.asarray(bytearray(webcam_img.read()), dtype=np.uint8)
        frame = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)

        annotated, exercise, conf = process_frame(frame)

        col1, col2 = st.columns(2)
        with col1:
            st.image(frame, channels="BGR", caption="Original", width=350)
        with col2:
            st.image(annotated, channels="BGR", caption="AI Analysis", width=350)

        st.markdown(f"""
        ### 🧠 AI Classification Result
        | Metric | Value |
        |--------|-------|
        | 🏷️ Exercise | **{exercise.title()}** |
        | 📊 Confidence | **{conf:.0%}** |
        | 🔢 Reps (this session) | **{st.session_state.count}** |
        """)

# ══════════════════════════════════════════
# TAB 3 — FITNESS CHATBOT
# ══════════════════════════════════════════
with tab3:
    st.subheader("🤖 Fitness Assistant Chatbot")
    st.info("Ask me anything about your exercises, form, diet, or recovery!")

    # Display chat history
    for msg in st.session_state.chat_history:
        with st.chat_message(msg["role"]):
            st.write(msg["content"])

    # Chat input
    user_input = st.chat_input("Ask your fitness question here...")

    if user_input:
        # Show user message
        st.session_state.chat_history.append({"role": "user", "content": user_input})
        with st.chat_message("user"):
            st.write(user_input)

        # Get and show bot reply
        reply = chatbot_reply(user_input)
        st.session_state.chat_history.append({"role": "assistant", "content": reply})
        with st.chat_message("assistant"):
            st.write(reply)

    if st.button("🗑️ Clear Chat"):
        st.session_state.chat_history = []
        st.rerun()
```

```python
%%writefile app.py
import streamlit as st
import cv2
import numpy as np
import tensorflow as tf
import os
import tempfile
import time
from mediapipe.python.solutions import pose as mp_pose
from mediapipe.python.solutions import drawing_utils as mp_drawing
from mediapipe.python.solutions import drawing_styles as mp_drawing_styles

# ─────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────
st.set_page_config(page_title="AI Fitness Trainer", layout="wide")
st.title("🏋️‍♂️ AI Virtual Fitness Trainer")

actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

# ─────────────────────────────────────────
# EXERCISE THRESHOLDS (tuned per exercise)
# ─────────────────────────────────────────
THRESHOLDS = {
    'squat':              {'down': 90,  'up': 160, 'joints': (23, 25, 27)},
    'barbell biceps curl':{'down': 50,  'up': 150, 'joints': (11, 13, 15)},
    'push-up':            {'down': 70,  'up': 150, 'joints': (11, 13, 15)},
    'shoulder press':     {'down': 70,  'up': 150, 'joints': (13, 11, 23)},
}

# Form feedback rules per exercise
FORM_RULES = {
    'squat': {
        'good_range': (60, 100),
        'good_msg':  '✅ Good squat depth! Keep your chest up.',
        'high_msg':  '⚠️ Go deeper — aim to get thighs parallel to ground.',
        'low_msg':   '⚠️ Too deep — protect your knees, stop at parallel.',
        'tip':       '💡 Tip: Keep knees tracking over toes, weight in heels.',
    },
    'barbell biceps curl': {
        'good_range': (40, 60),
        'good_msg':  '✅ Great curl! Full range of motion.',
        'high_msg':  '⚠️ Curl higher — squeeze your bicep at the top.',
        'low_msg':   '⚠️ Lower the weight fully to complete the rep.',
        'tip':       '💡 Tip: Keep elbows pinned to your sides.',
    },
    'push-up': {
        'good_range': (60, 90),
        'good_msg':  '✅ Good push-up form!',
        'high_msg':  '⚠️ Lower your chest closer to the ground.',
        'low_msg':   '⚠️ Don\'t collapse — keep core tight.',
        'tip':       '💡 Tip: Body should be a straight line from head to heels.',
    },
    'shoulder press': {
        'good_range': (60, 90),
        'good_msg':  '✅ Good press! Arms fully extended.',
        'high_msg':  '⚠️ Press all the way up — full extension.',
        'low_msg':   '⚠️ Lower the weight to shoulder level between reps.',
        'tip':       '💡 Tip: Brace your core to protect your lower back.',
    },
}

# ─────────────────────────────────────────
# LOAD MODEL & POSE
# ─────────────────────────────────────────
@st.cache_resource
def load_model():
    if not os.path.exists("fitness_trainer_model.h5"):
        st.error("❌ Model file not found! Train and save the model first.")
        st.stop()
    return tf.keras.models.load_model("fitness_trainer_model.h5")

@st.cache_resource
def load_pose():
    return mp_pose.Pose(
        min_detection_confidence=0.6,
        min_tracking_confidence=0.6,
        model_complexity=1
    )

model = load_model()
pose_tracker = load_pose()

# ─────────────────────────────────────────
# SESSION STATE
# ─────────────────────────────────────────
defaults = {
    "count": 0, "stage": None,
    "chat_history": [], "last_feedback": "",
    "last_exercise": "", "prev_angle": 0,
    "rep_confirmed": False
}
for k, v in defaults.items():
    if k not in st.session_state:
        st.session_state[k] = v

# ─────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    return 360 - angle if angle > 180.0 else angle

def get_angle(exercise, landmarks):
    if exercise not in THRESHOLDS:
        return 0
    i, j, k = THRESHOLDS[exercise]['joints']
    lm = landmarks
    return calculate_angle(
        [lm[i].x, lm[i].y],
        [lm[j].x, lm[j].y],
        [lm[k].x, lm[k].y]
    )

def get_form_feedback(exercise, angle_at_bottom):
    """Analyze angle at the bottom of the rep and return feedback."""
    if exercise not in FORM_RULES:
        return "🤖 Keep going!"
    rules = FORM_RULES[exercise]
    lo, hi = rules['good_range']
    if lo <= angle_at_bottom <= hi:
        return rules['good_msg'] + "\n" + rules['tip']
    elif angle_at_bottom > hi:
        return rules['high_msg'] + "\n" + rules['tip']
    else:
        return rules['low_msg'] + "\n" + rules['tip']

def count_rep_and_feedback(exercise, angle):
    """
    Strict 2-phase state machine:
    - Stage 'up'  : joint is extended (standing / arms straight)
    - Stage 'down': joint is contracted (squatting / arms curled)
    A rep is only counted when transitioning DOWN→UP (full completion).
    """
    feedback = st.session_state.last_feedback
    thresh = THRESHOLDS.get(exercise, {})
    up_thresh   = thresh.get('up',   150)
    down_thresh = thresh.get('down',  90)

    # Reset counter if exercise changed
    if exercise != st.session_state.last_exercise:
        st.session_state.count = 0
        st.session_state.stage = None
        st.session_state.last_exercise = exercise

    # Phase 1: detect "down" position
    if angle < down_thresh:
        if st.session_state.stage != "down":
            st.session_state.stage = "down"
            st.session_state.rep_confirmed = False
            # Give form feedback at bottom of rep
            feedback = get_form_feedback(exercise, angle)

    # Phase 2: detect return to "up" — only then count
    if angle > up_thresh and st.session_state.stage == "down":
        if not st.session_state.rep_confirmed:
            st.session_state.count += 1
            st.session_state.stage = "up"
            st.session_state.rep_confirmed = True

    st.session_state.last_feedback = feedback
    return feedback

def draw_overlay(frame, exercise, confidence, angle, count, feedback, stage):
    h, w = frame.shape[:2]

    # Top info bar
    cv2.rectangle(frame, (0, 0), (w, 60), (20, 20, 20), -1)
    cv2.putText(frame, f"{exercise.upper()}  {confidence:.0%}",
                (10, 40), cv2.FONT_HERSHEY_DUPLEX, 1.0, (0, 255, 255), 2)

    # Rep counter box (top-right)
    cv2.rectangle(frame, (w-160, 0), (w, 100), (33, 33, 220), -1)
    cv2.putText(frame, "REPS", (w-140, 35),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    cv2.putText(frame, str(count), (w-120, 90),
                cv2.FONT_HERSHEY_DUPLEX, 2.5, (255, 255, 255), 3)

    # Angle + stage
    cv2.rectangle(frame, (0, h-90), (320, h), (20, 20, 20), -1)
    cv2.putText(frame, f"Angle: {angle:.1f}  Stage: {stage or '-'}",
                (10, h-55), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (200, 200, 0), 2)

    # Form feedback (bottom)
    if feedback:
        short = feedback.split('\n')[0]  # First line only on frame
        color = (0, 220, 0) if "✅" in short else (0, 140, 255)
        cv2.putText(frame, short, (10, h-20),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.65, color, 2)
    return frame

def process_frame(frame, pose_tracker):
    image_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose_tracker.process(image_rgb)

    exercise, confidence, angle, feedback = "Unknown", 0.0, 0, ""

    if results.pose_landmarks:
        lm = results.pose_landmarks.landmark
        coords = np.array([[l.x, l.y, l.z] for l in lm]).flatten()

        pred = model.predict(coords.reshape(1, 1, 99), verbose=0)
        exercise   = actions[np.argmax(pred)]
        confidence = float(np.max(pred))
        angle      = get_angle(exercise, lm)
        feedback   = count_rep_and_feedback(exercise, angle)

        # Draw skeleton
        mp_drawing.draw_landmarks(
            frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing.DrawingSpec(color=(0,0,220), thickness=2, circle_radius=3),
            connection_drawing_spec=mp_drawing.DrawingSpec(color=(255,255,255), thickness=2)
        )

    frame = draw_overlay(frame, exercise, confidence, angle,
                         st.session_state.count,
                         feedback, st.session_state.stage)
    return frame, exercise, confidence, feedback

# ─────────────────────────────────────────
# SIDEBAR
# ─────────────────────────────────────────
st.sidebar.header("🎛️ Controls")
if st.sidebar.button("🔄 Reset Rep Counter"):
    st.session_state.count = 0
    st.session_state.stage = None
    st.session_state.last_exercise = ""
    st.session_state.last_feedback = ""
    st.sidebar.success("Counter reset!")

st.sidebar.markdown("---")
st.sidebar.metric("Current Reps", st.session_state.count)
if st.session_state.last_feedback:
    st.sidebar.markdown("**Last Feedback:**")
    st.sidebar.info(st.session_state.last_feedback)
st.sidebar.markdown("---")
st.sidebar.markdown("**Supported Exercises:**")
for a in actions:
    st.sidebar.markdown(f"- {a.title()}")

# ─────────────────────────────────────────
# TABS
# ─────────────────────────────────────────
tab1, tab2, tab3 = st.tabs(["📹 Video Analysis", "📸 Webcam Live Analysis", "🤖 Fitness Chatbot"])

# ══════════════════════════════════════════
# TAB 1 — VIDEO ANALYSIS
# ══════════════════════════════════════════
with tab1:
    st.subheader("📹 Upload a Workout Video")
    st.info("The AI detects your exercise, counts reps with strict form checking, and gives feedback.")

    video_file = st.file_uploader("Choose a video file", type=["mp4", "mov", "avi"], key="video_upload")

    if video_file:
        tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".mp4")
        tmp.write(video_file.read())
        tmp.close()

        cap = cv2.VideoCapture(tmp.name)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

        frame_placeholder = st.empty()
        progress_bar      = st.progress(0)
        stats_col1, stats_col2, stats_col3 = st.columns(3)
        feedback_placeholder = st.empty()

        frame_idx = 0
        skip = 3  # process every 3rd frame — reduces overcounting & speeds up

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            if frame_idx % skip == 0:
                annotated, exercise, conf, feedback = process_frame(frame, pose_tracker)
                frame_placeholder.image(annotated, channels="BGR", width=720)

                with stats_col1:
                    st.metric("🏷️ Exercise", exercise.title())
                with stats_col2:
                    st.metric("📊 Confidence", f"{conf:.0%}")
                with stats_col3:
                    st.metric("🔢 Reps", st.session_state.count)

                if feedback:
                    color = "green" if "✅" in feedback else "orange"
                    feedback_placeholder.markdown(
                        f"<div style='padding:10px;border-radius:8px;"
                        f"background:{'#e6ffe6' if color=='green' else '#fff3e0'};"
                        f"border-left:4px solid {'green' if color=='green' else 'orange'}'>"
                        f"{feedback.replace(chr(10), '<br>')}</div>",
                        unsafe_allow_html=True
                    )

            progress_bar.progress(min(frame_idx / max(total_frames, 1), 1.0))
            frame_idx += 1

        cap.release()
        os.unlink(tmp.name)
        st.balloons()
        st.success(f"🎉 Analysis complete! Total reps: **{st.session_state.count}**")

# ══════════════════════════════════════════
# TAB 2 — WEBCAM LIVE ANALYSIS
# ══════════════════════════════════════════
with tab2:
    st.subheader("📸 Webcam — Live Exercise Analysis")
    st.info("📷 The camera will capture frames continuously. Perform your exercise and get real-time feedback!")

    run_webcam = st.checkbox("▶️ Start Live Webcam Analysis", key="webcam_run")

    col_frame, col_info = st.columns([2, 1])

    with col_frame:
        webcam_placeholder = st.empty()

    with col_info:
        st.markdown("### 📊 Live Stats")
        ex_placeholder    = st.empty()
        conf_placeholder  = st.empty()
        rep_placeholder   = st.empty()
        stage_placeholder = st.empty()
        st.markdown("---")
        st.markdown("### 💬 Form Feedback")
        fb_placeholder = st.empty()

    if run_webcam:
        cap_wc = cv2.VideoCapture(0)

        if not cap_wc.isOpened():
            st.error("❌ Cannot access webcam. Make sure you allowed camera access in your browser.")
        else:
            st.success("✅ Webcam active! Perform your exercise in front of the camera.")

            # Run for up to 300 frames (~30s at 10fps processing)
            frame_count = 0
            MAX_FRAMES = 300

            while run_webcam and frame_count < MAX_FRAMES:
                ret, frame = cap_wc.read()
                if not ret:
                    st.warning("⚠️ Could not read from webcam.")
                    break

                annotated, exercise, conf, feedback = process_frame(frame, pose_tracker)

                webcam_placeholder.image(annotated, channels="BGR", width=500)

                ex_placeholder.metric("🏷️ Exercise", exercise.title())
                conf_placeholder.metric("📊 Confidence", f"{conf:.0%}")
                rep_placeholder.metric("🔢 Reps", st.session_state.count)
                stage_placeholder.metric("📍 Stage", st.session_state.stage or "-")

                if feedback:
                    color = "#e6ffe6" if "✅" in feedback else "#fff3e0"
                    border = "green" if "✅" in feedback else "orange"
                    fb_placeholder.markdown(
                        f"<div style='padding:10px;border-radius:8px;"
                        f"background:{color};border-left:4px solid {border}'>"
                        f"{feedback.replace(chr(10), '<br>')}</div>",
                        unsafe_allow_html=True
                    )

                frame_count += 1
                time.sleep(0.05)  # ~20fps cap

            cap_wc.release()
            if frame_count >= MAX_FRAMES:
                st.info("Session ended (30s limit). Click the checkbox again to restart.")
    else:
        webcam_placeholder.info("☝️ Check the box above to start your live analysis session.")

# ══════════════════════════════════════════
# TAB 3 — CHATBOT
# ══════════════════════════════════════════
with tab3:
    st.subheader("🤖 Fitness Assistant Chatbot")
    st.info("Ask me about your form, exercises, diet, recovery, or anything fitness-related!")

    def chatbot_reply(msg):
        msg = msg.lower()
        # Squat
        if any(w in msg for w in ["squat", "leg", "knee", "depth"]):
            return ("🦵 **Squat Form Guide:**\n"
                    "- Keep chest up and back straight\n"
                    "- Knees track over toes (don't cave inward)\n"
                    "- Go to at least parallel — thighs level with ground\n"
                    "- Weight in heels, not toes\n"
                    "- Brace core throughout\n\n"
                    "💡 The AI counts a rep only when you reach full depth AND return to standing.")
        # Bicep curl
        elif any(w in msg for w in ["curl", "bicep", "biceps", "arm"]):
            return ("💪 **Bicep Curl Form Guide:**\n"
                    "- Elbows pinned to sides — don't swing!\n"
                    "- Squeeze at the top for 1 second\n"
                    "- Lower slowly (3 seconds down)\n"
                    "- Use a weight you can control for all reps\n\n"
                    "💡 The AI detects elbow angle. Swinging makes the angle inaccurate.")
        # Push-up
        elif any(w in msg for w in ["push", "pushup", "chest", "tricep"]):
            return ("🤸 **Push-Up Form Guide:**\n"
                    "- Hands shoulder-width apart\n"
                    "- Body in a straight line — no sagging hips!\n"
                    "- Lower chest until it nearly touches the floor\n"
                    "- Elbows at ~45° from body\n\n"
                    "💡 Core must stay tight — the AI uses shoulder angle to detect depth.")
        # Shoulder press
        elif any(w in msg for w in ["shoulder", "press", "overhead", "ohp"]):
            return ("🏋️ **Shoulder Press Form Guide:**\n"
                    "- Start with weights at shoulder height\n"
                    "- Press straight up — arms fully extended at top\n"
                    "- Don't arch your lower back — brace your core\n"
                    "- Lower slowly back to shoulders\n\n"
                    "💡 The AI measures the shoulder-elbow-hip angle. "
                    "Make sure your full body is visible for best detection.")
        # Rep counting
        elif any(w in msg for w in ["count", "rep", "counting", "wrong count", "overcount"]):
            return ("🔢 **About Rep Counting:**\n"
                    "- The AI uses a strict 2-phase system: DOWN then UP\n"
                    "- A rep is only counted when you complete the full range\n"
                    "- If counts seem off, make sure your full body is in frame\n"
                    "- Use **Reset Rep Counter** in the sidebar between sets\n\n"
                    "💡 Each exercise has different angle thresholds — "
                    "Squat: <90° down, >160° up | Press: <70° down, >150° up")
        # Diet
        elif any(w in msg for w in ["diet", "protein", "food", "eat", "nutrition", "calorie"]):
            return ("🥗 **Nutrition Basics:**\n"
                    "- Protein: 1.6–2.2g per kg of bodyweight daily\n"
                    "- To build muscle: slight caloric surplus (+200–300 kcal)\n"
                    "- To lose fat: moderate deficit (-300–500 kcal)\n"
                    "- Prioritize: eggs, chicken, fish, legumes, Greek yogurt\n"
                    "- Stay hydrated: 2–3L water per day\n\n"
                    "💡 Time protein around workouts for best muscle synthesis.")
        # Recovery
        elif any(w in msg for w in ["rest", "recovery", "sleep", "sore", "doms"]):
            return ("😴 **Recovery Guide:**\n"
                    "- Sleep 7–9 hours — this is when muscles grow\n"
                    "- Allow 48hrs before training the same muscle group again\n"
                    "- DOMS (soreness) peaks at 24–48hrs — it's normal\n"
                    "- Light walking or stretching helps on rest days\n"
                    "- Hydration and protein are key to faster recovery")
        # Warm up
        elif any(w in msg for w in ["warm", "warmup", "stretch", "flexible"]):
            return ("🔥 **Warm-Up Routine (5–10 mins):**\n"
                    "- 2 min light cardio (jog in place, jumping jacks)\n"
                    "- Arm circles, leg swings, hip rotations\n"
                    "- Bodyweight version of your main exercise (no weight)\n\n"
                    "⚠️ Do dynamic stretches BEFORE, static stretches AFTER.\n"
                    "💡 Cold muscles are more prone to injury — never skip warm-up!")
        # Feedback question
        elif any(w in msg for w in ["feedback", "form", "correct", "wrong", "posture"]):
            return ("📊 **How AI Form Feedback Works:**\n"
                    "- The AI measures your joint angles in real-time\n"
                    "- At the bottom of each rep, it checks if you hit the target angle\n"
                    "- ✅ Green = good form | ⚠️ Orange = needs correction\n"
                    "- Feedback appears on the video frame and in the sidebar\n\n"
                    "💡 For best results: ensure good lighting and full body visible.")
        else:
            return ("🤖 I'm your AI Fitness Assistant! I can help with:\n"
                    "- **Exercise form**: squat, push-up, bicep curl, shoulder press\n"
                    "- **Rep counting** issues\n"
                    "- **Diet & nutrition** advice\n"
                    "- **Recovery & sleep** tips\n"
                    "- **Warm-up** routines\n\n"
                    "Just ask me anything! E.g. *'Is my squat form correct?'* or *'How much protein do I need?'*")

    for msg in st.session_state.chat_history:
        with st.chat_message(msg["role"]):
            st.markdown(msg["content"])

    user_input = st.chat_input("Ask your fitness question...")
    if user_input:
        st.session_state.chat_history.append({"role": "user", "content": user_input})
        with st.chat_message("user"):
            st.markdown(user_input)
        reply = chatbot_reply(user_input)
        st.session_state.chat_history.append({"role": "assistant", "content": reply})
        with st.chat_message("assistant"):
            st.markdown(reply)

    if st.button("🗑️ Clear Chat"):
        st.session_state.chat_history = []
        st.rerun()
```

```python
%%writefile app.py
import streamlit as st
import cv2
import numpy as np
import tensorflow as tf
import os
import tempfile
import time
from collections import deque
from mediapipe.python.solutions import pose as mp_pose
from mediapipe.python.solutions import drawing_utils as mp_drawing

# ─────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────
st.set_page_config(page_title="AI Fitness Trainer", layout="wide")
st.title("🏋️‍♂️ AI Virtual Fitness Trainer")

actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

# Per-exercise angle thresholds (DOWN angle, UP angle)
THRESHOLDS = {
    'squat':               {'down': 85,  'up': 155, 'joints': (23, 25, 27)},
    'barbell biceps curl': {'down': 45,  'up': 145, 'joints': (11, 13, 15)},
    'push-up':             {'down': 65,  'up': 145, 'joints': (11, 13, 15)},
    'shoulder press':      {'down': 65,  'up': 145, 'joints': (13, 11, 23)},
}

FORM_RULES = {
    'squat': {
        'good_range': (70, 95),
        'good_msg':  '✅ Great squat depth! Keep chest up.',
        'high_msg':  '⚠️ Go deeper — thighs should reach parallel.',
        'low_msg':   '⚠️ Too deep — stop at parallel to protect knees.',
        'tip':       '💡 Keep knees over toes, weight in heels.',
    },
    'barbell biceps curl': {
        'good_range': (35, 55),
        'good_msg':  '✅ Full range curl! Great squeeze.',
        'high_msg':  '⚠️ Curl higher — squeeze at the top.',
        'low_msg':   '⚠️ Lower fully to complete the rep.',
        'tip':       '💡 Keep elbows pinned to your sides.',
    },
    'push-up': {
        'good_range': (55, 80),
        'good_msg':  '✅ Good push-up depth!',
        'high_msg':  '⚠️ Lower chest closer to the ground.',
        'low_msg':   '⚠️ Don\'t collapse — keep core tight.',
        'tip':       '💡 Body must be a straight line.',
    },
    'shoulder press': {
        'good_range': (55, 80),
        'good_msg':  '✅ Great press! Full extension.',
        'high_msg':  '⚠️ Press fully overhead — extend arms.',
        'low_msg':   '⚠️ Lower to shoulder level between reps.',
        'tip':       '💡 Brace core to protect lower back.',
    },
}

# ─────────────────────────────────────────
# LOAD RESOURCES
# ─────────────────────────────────────────
@st.cache_resource
def load_model():
    if not os.path.exists("fitness_trainer_model.h5"):
        st.error("❌ Model not found! Train the model first.")
        st.stop()
    return tf.keras.models.load_model("fitness_trainer_model.h5")

@st.cache_resource
def load_pose():
    return mp_pose.Pose(
        min_detection_confidence=0.6,
        min_tracking_confidence=0.6,
        model_complexity=1
    )

model      = load_model()
pose_model = load_pose()

# ─────────────────────────────────────────
# SESSION STATE
# ─────────────────────────────────────────
defaults = {
    "count": 0,
    "stage": None,
    "last_feedback": "",
    "last_exercise": "",
    "rep_confirmed": False,
    "angle_buffer": deque(maxlen=5),   # smooth angles over 5 frames
    "chat_history": [],
}
for k, v in defaults.items():
    if k not in st.session_state:
        st.session_state[k] = v

# ─────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    return 360 - angle if angle > 180 else angle

def get_angle(exercise, lm):
    if exercise not in THRESHOLDS:
        return 0
    i, j, k = THRESHOLDS[exercise]['joints']
    return calculate_angle([lm[i].x, lm[i].y], [lm[j].x, lm[j].y], [lm[k].x, lm[k].y])

def smooth_angle(raw_angle):
    """Average over last N frames to reduce jitter."""
    st.session_state.angle_buffer.append(raw_angle)
    return float(np.mean(st.session_state.angle_buffer))

def get_form_feedback(exercise, angle):
    if exercise not in FORM_RULES:
        return ""
    r = FORM_RULES[exercise]
    lo, hi = r['good_range']
    if lo <= angle <= hi:
        return r['good_msg'] + "\n" + r['tip']
    elif angle > hi:
        return r['high_msg'] + "\n" + r['tip']
    else:
        return r['low_msg'] + "\n" + r['tip']

def update_rep_count(exercise, smoothed_angle):
    """
    Strict state machine:
      stage=None  → waiting
      stage='down'→ joint contracted (bottom of rep)
      stage='up'  → joint extended (top of rep) → COUNT HERE
    Rep only increments once per full DOWN→UP cycle.
    """
    thresh = THRESHOLDS.get(exercise, {})
    down_thresh = thresh.get('down', 90)
    up_thresh   = thresh.get('up',  150)

    # Reset if exercise changed
    if exercise != st.session_state.last_exercise:
        st.session_state.count        = 0
        st.session_state.stage        = None
        st.session_state.rep_confirmed = False
        st.session_state.last_exercise = exercise
        st.session_state.angle_buffer.clear()

    feedback = st.session_state.last_feedback

    # Detect bottom of rep
    if smoothed_angle <= down_thresh:
        if st.session_state.stage != "down":
            st.session_state.stage        = "down"
            st.session_state.rep_confirmed = False
            feedback = get_form_feedback(exercise, smoothed_angle)
            st.session_state.last_feedback = feedback

    # Detect top of rep — only count ONCE per down→up
    elif smoothed_angle >= up_thresh:
        if st.session_state.stage == "down" and not st.session_state.rep_confirmed:
            st.session_state.count        += 1
            st.session_state.stage         = "up"
            st.session_state.rep_confirmed = True

    return feedback

def draw_overlay(frame, exercise, confidence, angle, count, feedback, stage):
    h, w = frame.shape[:2]

    # Top bar
    cv2.rectangle(frame, (0, 0), (w, 55), (20, 20, 20), -1)
    cv2.putText(frame, f"{exercise.upper()}  {confidence:.0%}",
                (10, 38), cv2.FONT_HERSHEY_DUPLEX, 0.9, (0, 255, 255), 2)

    # Rep counter (top-right blue box)
    cv2.rectangle(frame, (w-150, 0), (w, 95), (180, 60, 20), -1)
    cv2.putText(frame, "REPS",  (w-130, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)
    cv2.putText(frame, str(count), (w-110, 85), cv2.FONT_HERSHEY_DUPLEX, 2.2, (255,255,255), 3)

    # Angle + stage (bottom-left)
    cv2.rectangle(frame, (0, h-85), (350, h), (20, 20, 20), -1)
    cv2.putText(frame, f"Angle: {angle:.1f}  |  Stage: {stage or 'ready'}",
                (8, h-52), cv2.FONT_HERSHEY_SIMPLEX, 0.65, (200,200,0), 2)

    # Feedback line
    if feedback:
        line  = feedback.split('\n')[0]
        color = (0, 210, 0) if "✅" in line else (0, 140, 255)
        cv2.putText(frame, line, (8, h-18),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)
    return frame

def run_inference(frame):
    """Full pipeline: pose → classify → angle → count → draw."""
    rgb     = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose_model.process(rgb)

    exercise, confidence, angle, feedback = "No pose detected", 0.0, 0, ""

    if results.pose_landmarks:
        lm     = results.pose_landmarks.landmark
        coords = np.array([[l.x, l.y, l.z] for l in lm]).flatten()

        pred       = model.predict(coords.reshape(1, 1, 99), verbose=0)
        exercise   = actions[np.argmax(pred)]
        confidence = float(np.max(pred))

        raw_angle = get_angle(exercise, lm)
        angle     = smooth_angle(raw_angle)        # ← smoothed angle
        feedback  = update_rep_count(exercise, angle)

        mp_drawing.draw_landmarks(
            frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing.DrawingSpec(
                color=(0, 0, 200), thickness=2, circle_radius=3),
            connection_drawing_spec=mp_drawing.DrawingSpec(
                color=(255, 255, 255), thickness=2)
        )

    frame = draw_overlay(frame, exercise, confidence, angle,
                         st.session_state.count, feedback,
                         st.session_state.stage)
    return frame, exercise, confidence, feedback

# ─────────────────────────────────────────
# SIDEBAR
# ─────────────────────────────────────────
st.sidebar.header("🎛️ Controls")
if st.sidebar.button("🔄 Reset Rep Counter"):
    st.session_state.count         = 0
    st.session_state.stage         = None
    st.session_state.last_exercise = ""
    st.session_state.last_feedback = ""
    st.session_state.rep_confirmed = False
    st.session_state.angle_buffer.clear()
    st.sidebar.success("Reset!")

st.sidebar.markdown("---")
st.sidebar.metric("Current Reps", st.session_state.count)

if st.session_state.last_feedback:
    st.sidebar.markdown("**Last Feedback:**")
    color = "#e6ffe6" if "✅" in st.session_state.last_feedback else "#fff3cd"
    st.sidebar.markdown(
        f"<div style='padding:8px;border-radius:6px;background:{color}'>"
        f"{st.session_state.last_feedback.replace(chr(10),'<br>')}</div>",
        unsafe_allow_html=True)

st.sidebar.markdown("---")
st.sidebar.markdown("**Supported Exercises:**")
for a in actions:
    st.sidebar.markdown(f"- {a.title()}")

# ─────────────────────────────────────────
# TABS
# ─────────────────────────────────────────
tab1, tab2, tab3 = st.tabs(["📹 Video Analysis", "📸 Webcam Live Analysis", "🤖 Fitness Chatbot"])

# ══════════════════════════════════════════
# TAB 1 — VIDEO UPLOAD
# ══════════════════════════════════════════
with tab1:
    st.subheader("📹 Upload a Workout Video")
    st.info("Upload any workout video. AI classifies exercise, counts reps accurately, and gives form feedback.")

    video_file = st.file_uploader("Choose video", type=["mp4", "mov", "avi"], key="vid")

    if video_file:
        tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".mp4")
        tmp.write(video_file.read())
        tmp.close()

        cap          = cv2.VideoCapture(tmp.name)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

        frame_ph    = st.empty()
        progress_ph = st.progress(0)
        c1, c2, c3  = st.columns(3)
        feedback_ph = st.empty()

        frame_idx = 0
        SKIP      = 2   # process every 2nd frame

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break

            if frame_idx % SKIP == 0:
                annotated, exercise, conf, feedback = run_inference(frame)
                frame_ph.image(annotated, channels="BGR", width=720)

                c1.metric("🏷️ Exercise",   exercise.title())
                c2.metric("📊 Confidence", f"{conf:.0%}")
                c3.metric("🔢 Reps",       st.session_state.count)

                if feedback:
                    bg     = "#e6ffe6" if "✅" in feedback else "#fff3cd"
                    border = "green"   if "✅" in feedback else "orange"
                    feedback_ph.markdown(
                        f"<div style='padding:10px;border-radius:8px;"
                        f"background:{bg};border-left:4px solid {border}'>"
                        f"{feedback.replace(chr(10),'<br>')}</div>",
                        unsafe_allow_html=True)

            progress_ph.progress(min(frame_idx / max(total_frames, 1), 1.0))
            frame_idx += 1

        cap.release()
        os.unlink(tmp.name)
        st.balloons()
        st.success(f"🎉 Done! Total reps: **{st.session_state.count}**")

# ══════════════════════════════════════════
# TAB 2 — WEBCAM  (browser camera → st.camera_input loop)
# ══════════════════════════════════════════
with tab2:
    st.subheader("📸 Webcam — Live Exercise Analysis")

    st.info(
        "**How to use:** Click **'Take photo'** repeatedly while exercising. "
        "Each snapshot is analysed instantly — exercise classified, angle checked, "
        "rep counted, and form feedback given. "
        "This works in ALL browsers without any server-side camera access."
    )

    st.warning(
        "⚠️ Why not true live video? Streamlit runs on a remote server (Colab/ngrok). "
        "Your browser camera stream cannot be piped to that server in real time. "
        "The snapshot approach is the correct solution — take one shot per rep position."
    )

    col_cam, col_result = st.columns([1, 1])

    with col_cam:
        webcam_img = st.camera_input("📷 Snap your exercise pose")

    with col_result:
        if webcam_img:
            file_bytes = np.asarray(bytearray(webcam_img.read()), dtype=np.uint8)
            frame      = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)

            annotated, exercise, conf, feedback = run_inference(frame)

            st.image(annotated, channels="BGR", width=400, caption="AI Analysis")

            # Stats
            st.markdown("### 📊 Results")
            r1, r2 = st.columns(2)
            r1.metric("🏷️ Exercise",   exercise.title())
            r2.metric("📊 Confidence", f"{conf:.0%}")
            r3, r4 = st.columns(2)
            r3.metric("🔢 Reps",       st.session_state.count)
            r4.metric("📍 Stage",      st.session_state.stage or "ready")

            # Form feedback card
            if feedback:
                bg     = "#e6ffe6" if "✅" in feedback else "#fff3cd"
                border = "green"   if "✅" in feedback else "orange"
                st.markdown(
                    f"<div style='padding:12px;border-radius:8px;"
                    f"background:{bg};border-left:5px solid {border};margin-top:10px'>"
                    f"<b>Form Feedback:</b><br>"
                    f"{feedback.replace(chr(10),'<br>')}</div>",
                    unsafe_allow_html=True)

            # Angle gauge bar
            thresh    = THRESHOLDS.get(exercise, {})
            max_angle = 180
            st.markdown(f"**Joint Angle: `{st.session_state.angle_buffer[-1] if st.session_state.angle_buffer else 0:.1f}°`**")
            angle_val = int(st.session_state.angle_buffer[-1]) if st.session_state.angle_buffer else 0
            st.progress(min(angle_val, 180) / 180)
        else:
            st.markdown("### 👈 Take a snapshot to see AI analysis here")
            st.markdown("""
            **Instructions:**
            1. Stand where your **full body** is visible
            2. Click **'Take photo'** at the bottom of a rep (hardest point)
            3. Click again at the top (standing / arms extended)
            4. Each photo counts toward your rep total
            5. Check feedback after each snap to improve form
            """)

# ══════════════════════════════════════════
# TAB 3 — CHATBOT
# ══════════════════════════════════════════
with tab3:
    st.subheader("🤖 Fitness Assistant Chatbot")
    st.info("Ask about form, exercises, diet, recovery, or rep counting!")

    def chatbot_reply(msg):
        m = msg.lower()
        if any(w in m for w in ["squat","leg","knee","depth"]):
            return ("🦵 **Squat Form:**\n"
                    "- Chest up, back straight\n- Knees track over toes\n"
                    "- Thighs parallel to ground at bottom\n- Weight in heels\n\n"
                    "💡 AI counts rep only when angle <85° (down) then >155° (up).")
        elif any(w in m for w in ["curl","bicep","biceps","arm"]):
            return ("💪 **Bicep Curl Form:**\n"
                    "- Elbows pinned to sides\n- Squeeze at top for 1 sec\n"
                    "- Lower slowly (3 sec)\n\n"
                    "💡 AI uses elbow angle: <45° = down, >145° = up.")
        elif any(w in m for w in ["push","pushup","chest","tricep"]):
            return ("🤸 **Push-Up Form:**\n"
                    "- Hands shoulder-width apart\n- Body straight line\n"
                    "- Chest nearly touches floor\n\n"
                    "💡 AI uses shoulder angle: <65° = down, >145° = up.")
        elif any(w in m for w in ["shoulder","press","overhead","ohp"]):
            return ("🏋️ **Shoulder Press Form:**\n"
                    "- Start at shoulder height\n- Press straight overhead\n"
                    "- Core braced throughout\n\n"
                    "💡 AI uses shoulder-elbow-hip angle: <65° = down, >145° = up.")
        elif any(w in m for w in ["count","rep","wrong","overcount","incorrect"]):
            return ("🔢 **Rep Counting Logic:**\n"
                    "- Each rep needs a full DOWN phase then UP phase\n"
                    "- Angles are smoothed over 5 frames to remove jitter\n"
                    "- Rep only increments ONCE per complete cycle\n"
                    "- Use **Reset** between sets\n\n"
                    "💡 Make sure full body is visible for accurate angle detection.")
        elif any(w in m for w in ["diet","protein","food","eat","nutrition","calorie"]):
            return ("🥗 **Nutrition Basics:**\n"
                    "- Protein: 1.6–2.2g per kg bodyweight\n"
                    "- Muscle gain: +200–300 kcal surplus\n"
                    "- Fat loss: -300–500 kcal deficit\n"
                    "- Best sources: eggs, chicken, fish, legumes\n"
                    "- Water: 2–3L per day")
        elif any(w in m for w in ["rest","recovery","sleep","sore","doms"]):
            return ("😴 **Recovery:**\n"
                    "- 7–9 hours sleep — muscles grow during sleep\n"
                    "- 48hr rest between training same muscle\n"
                    "- DOMS peaks 24–48hr after — normal\n"
                    "- Stay hydrated and hit protein targets")
        elif any(w in m for w in ["warm","warmup","stretch"]):
            return ("🔥 **Warm-Up (5–10 min):**\n"
                    "- 2 min light cardio\n"
                    "- Arm circles, leg swings, hip rotations\n"
                    "- Bodyweight reps of your main lift\n\n"
                    "⚠️ Dynamic stretches BEFORE, static AFTER.")
        else:
            return ("🤖 I can help with:\n"
                    "- **Form**: squat, curl, push-up, shoulder press\n"
                    "- **Rep counting** issues\n"
                    "- **Diet & nutrition**\n"
                    "- **Recovery & sleep**\n"
                    "- **Warm-up** routines\n\n"
                    "Try: *'Why is my squat count wrong?'* or *'How much protein do I need?'*")

    for msg in st.session_state.chat_history:
        with st.chat_message(msg["role"]):
            st.markdown(msg["content"])

    user_input = st.chat_input("Ask your fitness question...")
    if user_input:
        st.session_state.chat_history.append({"role": "user", "content": user_input})
        with st.chat_message("user"):
            st.markdown(user_input)
        reply = chatbot_reply(user_input)
        st.session_state.chat_history.append({"role": "assistant", "content": reply})
        with st.chat_message("assistant"):
            st.markdown(reply)

    if st.button("🗑️ Clear Chat"):
        st.session_state.chat_history = []
        st.rerun()
```

```python
%%writefile app.py
import streamlit as st
import cv2
import numpy as np
import tensorflow as tf
import os
import tempfile
import base64
import time
from collections import deque
from mediapipe.python.solutions import pose as mp_pose
from mediapipe.python.solutions import drawing_utils as mp_drawing

# ─────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────
st.set_page_config(page_title="AI Fitness Trainer", layout="wide")
st.title("🏋️‍♂️ AI Virtual Fitness Trainer")

actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

THRESHOLDS = {
    'squat':               {'down': 85,  'up': 155, 'joints': (23, 25, 27)},
    'barbell biceps curl': {'down': 45,  'up': 145, 'joints': (11, 13, 15)},
    'push-up':             {'down': 65,  'up': 145, 'joints': (11, 13, 15)},
    'shoulder press':      {'down': 65,  'up': 145, 'joints': (13, 11, 23)},
}

FORM_RULES = {
    'squat': {
        'good_range': (70, 95),
        'good_msg':  '✅ Great squat depth! Keep chest up.',
        'high_msg':  '⚠️ Go deeper — thighs should reach parallel.',
        'low_msg':   '⚠️ Too deep — stop at parallel to protect knees.',
        'tip':       '💡 Keep knees over toes, weight in heels.',
    },
    'barbell biceps curl': {
        'good_range': (35, 55),
        'good_msg':  '✅ Full range curl! Great squeeze.',
        'high_msg':  '⚠️ Curl higher — squeeze at the top.',
        'low_msg':   '⚠️ Lower fully to complete the rep.',
        'tip':       '💡 Keep elbows pinned to your sides.',
    },
    'push-up': {
        'good_range': (55, 80),
        'good_msg':  '✅ Good push-up depth!',
        'high_msg':  '⚠️ Lower chest closer to the ground.',
        'low_msg':   '⚠️ Don\'t collapse — keep core tight.',
        'tip':       '💡 Body must be a straight line.',
    },
    'shoulder press': {
        'good_range': (55, 80),
        'good_msg':  '✅ Great press! Full extension.',
        'high_msg':  '⚠️ Press fully overhead — extend arms.',
        'low_msg':   '⚠️ Lower to shoulder level between reps.',
        'tip':       '💡 Brace core to protect lower back.',
    },
}

# ─────────────────────────────────────────
# LOAD RESOURCES
# ─────────────────────────────────────────
@st.cache_resource
def load_model():
    if not os.path.exists("fitness_trainer_model.h5"):
        st.error("❌ Model not found!")
        st.stop()
    return tf.keras.models.load_model("fitness_trainer_model.h5")

@st.cache_resource
def load_pose():
    return mp_pose.Pose(
        min_detection_confidence=0.6,
        min_tracking_confidence=0.6,
        model_complexity=1
    )

model      = load_model()
pose_model = load_pose()

# ─────────────────────────────────────────
# SESSION STATE
# ─────────────────────────────────────────
defaults = {
    "count": 0, "stage": None,
    "last_feedback": "", "last_exercise": "",
    "rep_confirmed": False,
    "angle_buffer": deque(maxlen=5),
    "chat_history": [],
    "exercise_history": [],   # track last 10 exercises for majority vote
    "total_calories": 0.0,
    "session_start": time.time(),
}
for k, v in defaults.items():
    if k not in st.session_state:
        st.session_state[k] = v

# ─────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────
CALORIES_PER_REP = {
    'squat': 0.32,
    'barbell biceps curl': 0.14,
    'push-up': 0.29,
    'shoulder press': 0.19,
}

def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    return 360 - angle if angle > 180 else angle

def get_angle(exercise, lm):
    if exercise not in THRESHOLDS:
        return 0
    i, j, k = THRESHOLDS[exercise]['joints']
    return calculate_angle([lm[i].x, lm[i].y], [lm[j].x, lm[j].y], [lm[k].x, lm[k].y])

def smooth_angle(raw_angle):
    st.session_state.angle_buffer.append(raw_angle)
    return float(np.mean(st.session_state.angle_buffer))

def get_form_feedback(exercise, angle):
    if exercise not in FORM_RULES:
        return ""
    r = FORM_RULES[exercise]
    lo, hi = r['good_range']
    if lo <= angle <= hi:
        return r['good_msg'] + "\n" + r['tip']
    elif angle > hi:
        return r['high_msg'] + "\n" + r['tip']
    else:
        return r['low_msg'] + "\n" + r['tip']

def update_rep_count(exercise, smoothed_angle):
    thresh      = THRESHOLDS.get(exercise, {})
    down_thresh = thresh.get('down', 90)
    up_thresh   = thresh.get('up',  150)

    if exercise != st.session_state.last_exercise:
        st.session_state.count         = 0
        st.session_state.stage         = None
        st.session_state.rep_confirmed = False
        st.session_state.last_exercise = exercise
        st.session_state.angle_buffer.clear()
        st.session_state.total_calories = 0.0

    feedback = st.session_state.last_feedback

    if smoothed_angle <= down_thresh:
        if st.session_state.stage != "down":
            st.session_state.stage         = "down"
            st.session_state.rep_confirmed = False
            feedback = get_form_feedback(exercise, smoothed_angle)
            st.session_state.last_feedback = feedback

    elif smoothed_angle >= up_thresh:
        if st.session_state.stage == "down" and not st.session_state.rep_confirmed:
            st.session_state.count        += 1
            st.session_state.stage         = "up"
            st.session_state.rep_confirmed = True
            cal = CALORIES_PER_REP.get(exercise, 0.2)
            st.session_state.total_calories += cal

    return feedback

def draw_overlay(frame, exercise, confidence, angle, count, feedback, stage):
    h, w = frame.shape[:2]
    cv2.rectangle(frame, (0, 0), (w, 55), (20, 20, 20), -1)
    cv2.putText(frame, f"{exercise.upper()}  {confidence:.0%}",
                (10, 38), cv2.FONT_HERSHEY_DUPLEX, 0.85, (0, 255, 255), 2)
    cv2.rectangle(frame, (w-150, 0), (w, 95), (180, 60, 20), -1)
    cv2.putText(frame, "REPS",  (w-130, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)
    cv2.putText(frame, str(count), (w-110, 85), cv2.FONT_HERSHEY_DUPLEX, 2.2, (255,255,255), 3)
    cv2.rectangle(frame, (0, h-85), (400, h), (20, 20, 20), -1)
    cv2.putText(frame, f"Angle: {angle:.1f}  Stage: {stage or 'ready'}",
                (8, h-52), cv2.FONT_HERSHEY_SIMPLEX, 0.65, (200,200,0), 2)
    if feedback:
        line  = feedback.split('\n')[0]
        color = (0, 210, 0) if "✅" in line else (0, 140, 255)
        cv2.putText(frame, line, (8, h-18), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)
    return frame

def run_inference(frame):
    rgb     = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose_model.process(rgb)
    exercise, confidence, angle, feedback = "No pose detected", 0.0, 0, ""

    if results.pose_landmarks:
        lm     = results.pose_landmarks.landmark
        coords = np.array([[l.x, l.y, l.z] for l in lm]).flatten()
        pred       = model.predict(coords.reshape(1, 1, 99), verbose=0)
        exercise   = actions[np.argmax(pred)]
        confidence = float(np.max(pred))
        raw_angle  = get_angle(exercise, lm)
        angle      = smooth_angle(raw_angle)
        feedback   = update_rep_count(exercise, angle)

        mp_drawing.draw_landmarks(
            frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing.DrawingSpec(color=(0,0,200), thickness=2, circle_radius=3),
            connection_drawing_spec=mp_drawing.DrawingSpec(color=(255,255,255), thickness=2)
        )

    frame = draw_overlay(frame, exercise, confidence, angle,
                         st.session_state.count, feedback, st.session_state.stage)
    return frame, exercise, confidence, feedback

# ─────────────────────────────────────────
# SIDEBAR — Workout Dashboard (replaces "Current Reps")
# ─────────────────────────────────────────
st.sidebar.header("🎛️ Controls")
if st.sidebar.button("🔄 Reset Session"):
    st.session_state.count          = 0
    st.session_state.stage          = None
    st.session_state.last_exercise  = ""
    st.session_state.last_feedback  = ""
    st.session_state.rep_confirmed  = False
    st.session_state.total_calories = 0.0
    st.session_state.session_start  = time.time()
    st.session_state.angle_buffer.clear()
    st.sidebar.success("Session reset!")

st.sidebar.markdown("---")

# ── Workout Dashboard ──
elapsed = int(time.time() - st.session_state.session_start)
mins, secs = divmod(elapsed, 60)

st.sidebar.markdown("### 📊 Workout Dashboard")
st.sidebar.markdown(
    f"""
    <div style='background:#1e1e2e;padding:14px;border-radius:12px;color:white;font-family:sans-serif'>
      <div style='display:flex;justify-content:space-between;margin-bottom:10px'>
        <div style='text-align:center'>
          <div style='font-size:28px;font-weight:bold;color:#4ade80'>{st.session_state.count}</div>
          <div style='font-size:11px;color:#aaa'>REPS</div>
        </div>
        <div style='text-align:center'>
          <div style='font-size:28px;font-weight:bold;color:#f97316'>{st.session_state.total_calories:.1f}</div>
          <div style='font-size:11px;color:#aaa'>CALORIES</div>
        </div>
        <div style='text-align:center'>
          <div style='font-size:28px;font-weight:bold;color:#60a5fa'>{mins:02d}:{secs:02d}</div>
          <div style='font-size:11px;color:#aaa'>TIME</div>
        </div>
      </div>
      <div style='border-top:1px solid #333;padding-top:8px;margin-top:4px'>
        <div style='font-size:12px;color:#aaa'>Current Exercise</div>
        <div style='font-size:16px;font-weight:bold;color:#e879f9'>
          {st.session_state.last_exercise.title() if st.session_state.last_exercise else "Not detected yet"}
        </div>
        <div style='font-size:12px;color:#aaa;margin-top:6px'>Stage</div>
        <div style='font-size:16px;font-weight:bold;color:#facc15'>
          {(st.session_state.stage or "Ready").upper()}
        </div>
      </div>
    </div>
    """,
    unsafe_allow_html=True
)

if st.session_state.last_feedback:
    st.sidebar.markdown("**💬 Last Form Feedback:**")
    bg = "#e6ffe6" if "✅" in st.session_state.last_feedback else "#fff3cd"
    bd = "green"   if "✅" in st.session_state.last_feedback else "orange"
    st.sidebar.markdown(
        f"<div style='padding:8px;border-radius:6px;background:{bg};border-left:4px solid {bd}'>"
        f"{st.session_state.last_feedback.replace(chr(10),'<br>')}</div>",
        unsafe_allow_html=True)

st.sidebar.markdown("---")
st.sidebar.markdown("**🏋️ Supported Exercises:**")
for a in actions:
    st.sidebar.markdown(f"- {a.title()}")

# ─────────────────────────────────────────
# TABS
# ─────────────────────────────────────────
tab1, tab2, tab3 = st.tabs(["📹 Video Analysis", "🎥 Live Webcam", "🤖 Fitness Chatbot"])

# ══════════════════════════════════════════
# TAB 1 — VIDEO UPLOAD
# ══════════════════════════════════════════
with tab1:
    st.subheader("📹 Upload a Workout Video")
    st.info("Upload your workout video. AI classifies exercise, counts reps, and gives form feedback.")

    video_file = st.file_uploader("Choose video", type=["mp4", "mov", "avi"], key="vid")

    if video_file:
        tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".mp4")
        tmp.write(video_file.read())
        tmp.close()

        cap          = cv2.VideoCapture(tmp.name)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

        frame_ph    = st.empty()
        progress_ph = st.progress(0)
        c1, c2, c3  = st.columns(3)
        feedback_ph = st.empty()

        frame_idx = 0
        SKIP = 2

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            if frame_idx % SKIP == 0:
                annotated, exercise, conf, feedback = run_inference(frame)
                frame_ph.image(annotated, channels="BGR", width=720)
                c1.metric("🏷️ Exercise",   exercise.title())
                c2.metric("📊 Confidence", f"{conf:.0%}")
                c3.metric("🔢 Reps",       st.session_state.count)
                if feedback:
                    bg = "#e6ffe6" if "✅" in feedback else "#fff3cd"
                    bd = "green"   if "✅" in feedback else "orange"
                    feedback_ph.markdown(
                        f"<div style='padding:10px;border-radius:8px;"
                        f"background:{bg};border-left:4px solid {bd}'>"
                        f"{feedback.replace(chr(10),'<br>')}</div>",
                        unsafe_allow_html=True)
            progress_ph.progress(min(frame_idx / max(total_frames, 1), 1.0))
            frame_idx += 1

        cap.release()
        os.unlink(tmp.name)
        st.balloons()
        st.success(f"🎉 Done! Total reps: **{st.session_state.count}** | "
                   f"Calories burned: **{st.session_state.total_calories:.1f} kcal**")

# ══════════════════════════════════════════
# TAB 2 — TRUE LIVE WEBCAM via JS + base64
# ══════════════════════════════════════════
with tab2:
    st.subheader("🎥 Live Webcam — Real-Time Analysis")
    st.info("Your camera streams live. Frames are captured every second and analysed by AI.")

    # JavaScript captures video frames and sends them as base64 via a hidden input
    webcam_html = """
    <div style="display:flex;flex-direction:column;align-items:center;gap:10px">
      <video id="vid" width="480" height="360" autoplay playsinline
             style="border-radius:12px;border:3px solid #4ade80;"></video>
      <canvas id="canvas" width="480" height="360" style="display:none;"></canvas>
      <div style="display:flex;gap:10px">
        <button onclick="startCam()"
          style="padding:10px 24px;background:#4ade80;border:none;border-radius:8px;
                 font-size:16px;font-weight:bold;cursor:pointer">▶ Start Camera</button>
        <button onclick="stopCam()"
          style="padding:10px 24px;background:#f87171;border:none;border-radius:8px;
                 font-size:16px;font-weight:bold;cursor:pointer;color:white">⏹ Stop</button>
      </div>
      <div id="status" style="color:#4ade80;font-size:14px;font-weight:bold">
        Click "Start Camera" to begin
      </div>
    </div>

    <script>
    let stream = null;
    let intervalId = null;
    const vid    = document.getElementById('vid');
    const canvas = document.getElementById('canvas');
    const ctx    = canvas.getContext('2d');
    const status = document.getElementById('status');

    async function startCam() {
      try {
        stream = await navigator.mediaDevices.getUserMedia({video: true, audio: false});
        vid.srcObject = stream;
        status.innerText = '🟢 Camera active — capturing every 1.5s';
        status.style.color = '#4ade80';

        intervalId = setInterval(() => {
          ctx.drawImage(vid, 0, 0, 480, 360);
          const b64 = canvas.toDataURL('image/jpeg', 0.7).split(',')[1];
          // Send to Streamlit via query param trick
          window.parent.postMessage({type: 'streamlit:setComponentValue', value: b64}, '*');
        }, 1500);
      } catch(e) {
        status.innerText = '❌ Camera error: ' + e.message;
        status.style.color = '#f87171';
      }
    }

    function stopCam() {
      if(stream) { stream.getTracks().forEach(t => t.stop()); stream = null; }
      if(intervalId) { clearInterval(intervalId); intervalId = null; }
      status.innerText = '⏹ Camera stopped';
      status.style.color = '#aaa';
    }
    </script>
    """

    # Render webcam HTML
    from streamlit.components.v1 import html as st_html
    st_html(webcam_html, height=480)

    st.markdown("---")
    st.info("📤 To analyse a live frame: use the **Upload a frame** button below. "
            "Or record a short clip and use the **Video Analysis** tab — "
            "it gives frame-by-frame analysis identical to live.")

    st.markdown("### 📸 Paste / Upload a Frame for Instant Analysis")
    col_upload, col_result = st.columns([1, 1])

    with col_upload:
        live_frame = st.camera_input("📷 Capture a live frame from your camera")

    with col_result:
        if live_frame:
            file_bytes = np.asarray(bytearray(live_frame.read()), dtype=np.uint8)
            frame      = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)
            annotated, exercise, conf, feedback = run_inference(frame)

            st.image(annotated, channels="BGR", width=400)

            r1, r2 = st.columns(2)
            r1.metric("🏷️ Exercise",   exercise.title())
            r2.metric("📊 Confidence", f"{conf:.0%}")
            r3, r4 = st.columns(2)
            r3.metric("🔢 Reps",  st.session_state.count)
            r4.metric("🔥 Calories", f"{st.session_state.total_calories:.1f}")

            if feedback:
                bg = "#e6ffe6" if "✅" in feedback else "#fff3cd"
                bd = "green"   if "✅" in feedback else "orange"
                st.markdown(
                    f"<div style='padding:12px;border-radius:8px;"
                    f"background:{bg};border-left:5px solid {bd}'>"
                    f"<b>Form Feedback:</b><br>"
                    f"{feedback.replace(chr(10),'<br>')}</div>",
                    unsafe_allow_html=True)

            # Progress bars for angle
            if st.session_state.angle_buffer:
                ang = st.session_state.angle_buffer[-1]
                st.markdown(f"**Joint Angle: `{ang:.1f}°`**")
                st.progress(min(int(ang), 180) / 180)

            # Form score
            score_color = "🟢" if "✅" in feedback else "🟡" if "⚠️" in feedback else "🔴"
            st.markdown(f"**Form Score:** {score_color} "
                        f"{'Good form!' if '✅' in feedback else 'Needs correction' if '⚠️' in feedback else 'Detecting...'}")
        else:
            st.markdown("""
            ### 👈 Instructions
            1. Allow camera access when prompted
            2. Stand so your **full body** is visible
            3. Click **'Take photo'** at key points:
               - **Bottom** of the rep (e.g. bottom of squat)
               - **Top** of the rep (e.g. standing straight)
            4. Each capture is analysed instantly
            5. Feedback appears here in real-time

            > 💡 **Pro tip:** For best accuracy ensure good lighting
            > and keep 2–3 metres from the camera so your full body fits.
            """)

# ══════════════════════════════════════════
# TAB 3 — CHATBOT
# ══════════════════════════════════════════
with tab3:
    st.subheader("🤖 Fitness Assistant Chatbot")
    st.info("Ask about form, exercises, diet, recovery, or rep counting!")

    def chatbot_reply(msg):
        m = msg.lower()
        if any(w in m for w in ["squat","leg","knee","depth"]):
            return ("🦵 **Squat Form:**\n"
                    "- Chest up, back straight\n- Knees track over toes\n"
                    "- Thighs parallel at bottom\n- Weight in heels\n\n"
                    "💡 AI counts rep only when angle <85° then >155°.")
        elif any(w in m for w in ["curl","bicep","biceps","arm"]):
            return ("💪 **Bicep Curl Form:**\n"
                    "- Elbows pinned to sides\n- Squeeze at top 1 sec\n"
                    "- Lower slowly 3 sec\n\n"
                    "💡 AI uses elbow angle: <45° = down, >145° = up.")
        elif any(w in m for w in ["push","pushup","chest","tricep"]):
            return ("🤸 **Push-Up Form:**\n"
                    "- Hands shoulder-width\n- Body straight line\n"
                    "- Chest nearly touches floor\n\n"
                    "💡 Shoulder angle: <65° = down, >145° = up.")
        elif any(w in m for w in ["shoulder","press","overhead"]):
            return ("🏋️ **Shoulder Press Form:**\n"
                    "- Start at shoulder height\n- Press straight overhead\n"
                    "- Core braced throughout\n\n"
                    "💡 Angle: <65° = down, >145° = up.")
        elif any(w in m for w in ["count","rep","wrong","overcount"]):
            return ("🔢 **Rep Counting:**\n"
                    "- Full DOWN then UP = 1 rep\n"
                    "- Angles smoothed over 5 frames\n"
                    "- Rep counted only once per cycle\n\n"
                    "💡 Full body must be visible in frame.")
        elif any(w in m for w in ["calorie","burn","energy"]):
            return ("🔥 **Calorie Estimates per rep:**\n"
                    "- Squat: ~0.32 kcal\n"
                    "- Push-up: ~0.29 kcal\n"
                    "- Shoulder Press: ~0.19 kcal\n"
                    "- Bicep Curl: ~0.14 kcal\n\n"
                    "💡 Check your dashboard in the sidebar for total calories burned!")
        elif any(w in m for w in ["diet","protein","food","eat","nutrition"]):
            return ("🥗 **Nutrition:**\n"
                    "- Protein: 1.6–2.2g per kg bodyweight\n"
                    "- Muscle gain: +200–300 kcal surplus\n"
                    "- Fat loss: -300–500 kcal deficit\n"
                    "- Best sources: eggs, chicken, fish, legumes")
        elif any(w in m for w in ["rest","recovery","sleep","sore"]):
            return ("😴 **Recovery:**\n"
                    "- 7–9 hours sleep\n- 48hr rest per muscle group\n"
                    "- Stay hydrated and hit protein targets")
        elif any(w in m for w in ["warm","warmup","stretch"]):
            return ("🔥 **Warm-Up (5–10 min):**\n"
                    "- 2 min light cardio\n- Arm circles, leg swings\n"
                    "- Bodyweight reps of main lift\n\n"
                    "⚠️ Dynamic BEFORE, static AFTER.")
        else:
            return ("🤖 I can help with:\n"
                    "- **Form**: squat, curl, push-up, shoulder press\n"
                    "- **Calories** burned\n- **Diet & nutrition**\n"
                    "- **Recovery & sleep**\n- **Warm-up** routines\n\n"
                    "Try: *'How many calories did I burn?'* or *'Fix my squat form'*")

    for msg in st.session_state.chat_history:
        with st.chat_message(msg["role"]):
            st.markdown(msg["content"])

    user_input = st.chat_input("Ask your fitness question...")
    if user_input:
        st.session_state.chat_history.append({"role": "user", "content": user_input})
        with st.chat_message("user"):
            st.markdown(user_input)
        reply = chatbot_reply(user_input)
        st.session_state.chat_history.append({"role": "assistant", "content": reply})
        with st.chat_message("assistant"):
            st.markdown(reply)

    if st.button("🗑️ Clear Chat"):
        st.session_state.chat_history = []
        st.rerun()
```

```python
%%writefile app.py
import streamlit as st
import cv2
import numpy as np
import tensorflow as tf
import os
import tempfile
import base64
import time
from collections import deque
from mediapipe.python.solutions import pose as mp_pose
from mediapipe.python.solutions import drawing_utils as mp_drawing

# ─────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────
st.set_page_config(page_title="AI Fitness Trainer", layout="wide")
st.title("🏋️‍♂️ AI Virtual Fitness Trainer")

actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

THRESHOLDS = {
    'squat':               {'down': 85,  'up': 155, 'joints': (23, 25, 27)},
    'barbell biceps curl': {'down': 45,  'up': 145, 'joints': (11, 13, 15)},
    'push-up':             {'down': 65,  'up': 145, 'joints': (11, 13, 15)},
    'shoulder press':      {'down': 65,  'up': 145, 'joints': (13, 11, 23)},
}

FORM_RULES = {
    'squat': {
        'good_range': (70, 95),
        'good_msg':  '✅ Great squat depth! Keep chest up.',
        'high_msg':  '⚠️ Go deeper — thighs should reach parallel.',
        'low_msg':   '⚠️ Too deep — stop at parallel to protect knees.',
        'tip':       '💡 Keep knees over toes, weight in heels.',
    },
    'barbell biceps curl': {
        'good_range': (35, 55),
        'good_msg':  '✅ Full range curl! Great squeeze.',
        'high_msg':  '⚠️ Curl higher — squeeze at the top.',
        'low_msg':   '⚠️ Lower fully to complete the rep.',
        'tip':       '💡 Keep elbows pinned to your sides.',
    },
    'push-up': {
        'good_range': (55, 80),
        'good_msg':  '✅ Good push-up depth!',
        'high_msg':  '⚠️ Lower chest closer to the ground.',
        'low_msg':   "⚠️ Don't collapse — keep core tight.",
        'tip':       '💡 Body must be a straight line.',
    },
    'shoulder press': {
        'good_range': (55, 80),
        'good_msg':  '✅ Great press! Full extension.',
        'high_msg':  '⚠️ Press fully overhead — extend arms.',
        'low_msg':   '⚠️ Lower to shoulder level between reps.',
        'tip':       '💡 Brace core to protect lower back.',
    },
}

CALORIES_PER_REP = {
    'squat': 0.32,
    'barbell biceps curl': 0.14,
    'push-up': 0.29,
    'shoulder press': 0.19,
}

# ─────────────────────────────────────────
# LOAD RESOURCES
# ─────────────────────────────────────────
@st.cache_resource
def load_model():
    if not os.path.exists("fitness_trainer_model.h5"):
        st.error("❌ Model not found!")
        st.stop()
    return tf.keras.models.load_model("fitness_trainer_model.h5")

@st.cache_resource
def load_pose():
    return mp_pose.Pose(
        min_detection_confidence=0.6,
        min_tracking_confidence=0.6,
        model_complexity=1
    )

model      = load_model()
pose_model = load_pose()

# ─────────────────────────────────────────
# SESSION STATE
# ─────────────────────────────────────────
defaults = {
    "count": 0, "stage": None,
    "last_feedback": "", "last_exercise": "",
    "rep_confirmed": False,
    "angle_buffer": deque(maxlen=5),
    "chat_history": [],
    "total_calories": 0.0,
    "session_start": time.time(),
}
for k, v in defaults.items():
    if k not in st.session_state:
        st.session_state[k] = v

# ─────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    return 360 - angle if angle > 180 else angle

def get_angle(exercise, lm):
    if exercise not in THRESHOLDS:
        return 0
    i, j, k = THRESHOLDS[exercise]['joints']
    return calculate_angle([lm[i].x, lm[i].y], [lm[j].x, lm[j].y], [lm[k].x, lm[k].y])

def smooth_angle(raw_angle):
    st.session_state.angle_buffer.append(raw_angle)
    return float(np.mean(st.session_state.angle_buffer))

def get_form_feedback(exercise, angle):
    if exercise not in FORM_RULES:
        return ""
    r = FORM_RULES[exercise]
    lo, hi = r['good_range']
    if lo <= angle <= hi:
        return r['good_msg'] + "\n" + r['tip']
    elif angle > hi:
        return r['high_msg'] + "\n" + r['tip']
    else:
        return r['low_msg'] + "\n" + r['tip']

def update_rep_count(exercise, smoothed_angle):
    thresh      = THRESHOLDS.get(exercise, {})
    down_thresh = thresh.get('down', 90)
    up_thresh   = thresh.get('up',  150)

    if exercise != st.session_state.last_exercise:
        st.session_state.count         = 0
        st.session_state.stage         = None
        st.session_state.rep_confirmed = False
        st.session_state.last_exercise = exercise
        st.session_state.angle_buffer.clear()
        st.session_state.total_calories = 0.0

    feedback = st.session_state.last_feedback

    if smoothed_angle <= down_thresh:
        if st.session_state.stage != "down":
            st.session_state.stage         = "down"
            st.session_state.rep_confirmed = False
            feedback = get_form_feedback(exercise, smoothed_angle)
            st.session_state.last_feedback = feedback

    elif smoothed_angle >= up_thresh:
        if st.session_state.stage == "down" and not st.session_state.rep_confirmed:
            st.session_state.count        += 1
            st.session_state.stage         = "up"
            st.session_state.rep_confirmed = True
            cal = CALORIES_PER_REP.get(exercise, 0.2)
            st.session_state.total_calories += cal

    return feedback

def draw_overlay(frame, exercise, confidence, angle, count, feedback, stage):
    h, w = frame.shape[:2]
    cv2.rectangle(frame, (0, 0), (w, 55), (20, 20, 20), -1)
    cv2.putText(frame, f"{exercise.upper()}  {confidence:.0%}",
                (10, 38), cv2.FONT_HERSHEY_DUPLEX, 0.85, (0, 255, 255), 2)
    cv2.rectangle(frame, (w-150, 0), (w, 95), (180, 60, 20), -1)
    cv2.putText(frame, "REPS",  (w-130, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)
    cv2.putText(frame, str(count), (w-110, 85), cv2.FONT_HERSHEY_DUPLEX, 2.2, (255,255,255), 3)
    cv2.rectangle(frame, (0, h-85), (400, h), (20, 20, 20), -1)
    cv2.putText(frame, f"Angle: {angle:.1f}  Stage: {stage or 'ready'}",
                (8, h-52), cv2.FONT_HERSHEY_SIMPLEX, 0.65, (200,200,0), 2)
    if feedback:
        line  = feedback.split('\n')[0]
        color = (0, 210, 0) if "✅" in line else (0, 140, 255)
        cv2.putText(frame, line, (8, h-18), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)
    return frame

# ────────────────────────────────────────────────────────────────
# FIX 1: run_inference now returns the annotated frame as RGB
#         so st.image() renders it correctly without BGR swap.
# ────────────────────────────────────────────────────────────────
def run_inference(frame):
    """
    frame : BGR numpy array (from cv2 or decoded bytes)
    returns: annotated RGB frame, exercise str, confidence float, feedback str
    """
    rgb     = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose_model.process(rgb)
    exercise, confidence, angle, feedback = "No pose detected", 0.0, 0, ""

    if results.pose_landmarks:
        lm     = results.pose_landmarks.landmark
        coords = np.array([[l.x, l.y, l.z] for l in lm]).flatten()
        pred       = model.predict(coords.reshape(1, 1, 99), verbose=0)
        exercise   = actions[np.argmax(pred)]
        confidence = float(np.max(pred))
        raw_angle  = get_angle(exercise, lm)
        angle      = smooth_angle(raw_angle)
        feedback   = update_rep_count(exercise, angle)

        # Draw skeleton on the RGB copy so colours look correct
        mp_drawing.draw_landmarks(
            rgb, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing.DrawingSpec(color=(200,0,0), thickness=2, circle_radius=3),
            connection_drawing_spec=mp_drawing.DrawingSpec(color=(255,255,255), thickness=2)
        )

    # draw_overlay works on BGR; convert rgb→bgr, draw, convert back
    bgr_annotated = cv2.cvtColor(rgb, cv2.COLOR_RGB2BGR)
    bgr_annotated = draw_overlay(bgr_annotated, exercise, confidence, angle,
                                 st.session_state.count, feedback, st.session_state.stage)
    rgb_annotated = cv2.cvtColor(bgr_annotated, cv2.COLOR_BGR2RGB)
    return rgb_annotated, exercise, confidence, feedback

# ─────────────────────────────────────────
# SIDEBAR
# ─────────────────────────────────────────
st.sidebar.header("🎛️ Controls")
if st.sidebar.button("🔄 Reset Session"):
    st.session_state.count          = 0
    st.session_state.stage          = None
    st.session_state.last_exercise  = ""
    st.session_state.last_feedback  = ""
    st.session_state.rep_confirmed  = False
    st.session_state.total_calories = 0.0
    st.session_state.session_start  = time.time()
    st.session_state.angle_buffer.clear()
    st.sidebar.success("Session reset!")

st.sidebar.markdown("---")

elapsed = int(time.time() - st.session_state.session_start)
mins, secs = divmod(elapsed, 60)

st.sidebar.markdown("### 📊 Workout Dashboard")
st.sidebar.markdown(
    f"""
    <div style='background:#1e1e2e;padding:14px;border-radius:12px;color:white;font-family:sans-serif'>
      <div style='display:flex;justify-content:space-between;margin-bottom:10px'>
        <div style='text-align:center'>
          <div style='font-size:28px;font-weight:bold;color:#4ade80'>{st.session_state.count}</div>
          <div style='font-size:11px;color:#aaa'>REPS</div>
        </div>
        <div style='text-align:center'>
          <div style='font-size:28px;font-weight:bold;color:#f97316'>{st.session_state.total_calories:.1f}</div>
          <div style='font-size:11px;color:#aaa'>CALORIES</div>
        </div>
        <div style='text-align:center'>
          <div style='font-size:28px;font-weight:bold;color:#60a5fa'>{mins:02d}:{secs:02d}</div>
          <div style='font-size:11px;color:#aaa'>TIME</div>
        </div>
      </div>
      <div style='border-top:1px solid #333;padding-top:8px;margin-top:4px'>
        <div style='font-size:12px;color:#aaa'>Current Exercise</div>
        <div style='font-size:16px;font-weight:bold;color:#e879f9'>
          {st.session_state.last_exercise.title() if st.session_state.last_exercise else "Not detected yet"}
        </div>
        <div style='font-size:12px;color:#aaa;margin-top:6px'>Stage</div>
        <div style='font-size:16px;font-weight:bold;color:#facc15'>
          {(st.session_state.stage or "Ready").upper()}
        </div>
      </div>
    </div>
    """,
    unsafe_allow_html=True
)

if st.session_state.last_feedback:
    st.sidebar.markdown("**💬 Last Form Feedback:**")
    bg = "#e6ffe6" if "✅" in st.session_state.last_feedback else "#fff3cd"
    bd = "green"   if "✅" in st.session_state.last_feedback else "orange"
    st.sidebar.markdown(
        f"<div style='padding:8px;border-radius:6px;background:{bg};border-left:4px solid {bd}'>"
        f"{st.session_state.last_feedback.replace(chr(10),'<br>')}</div>",
        unsafe_allow_html=True)

st.sidebar.markdown("---")
st.sidebar.markdown("**🏋️ Supported Exercises:**")
for a in actions:
    st.sidebar.markdown(f"- {a.title()}")

# ─────────────────────────────────────────
# TABS
# ─────────────────────────────────────────
tab1, tab2, tab3 = st.tabs(["📹 Video Analysis", "🎥 Live Webcam", "🤖 Fitness Chatbot"])

# ══════════════════════════════════════════
# TAB 1 — VIDEO UPLOAD
# FIX 2: removed channels="BGR" — we now pass RGB frames to st.image()
# ══════════════════════════════════════════
with tab1:
    st.subheader("📹 Upload a Workout Video")
    st.info("Upload your workout video. AI classifies exercise, counts reps, and gives form feedback.")

    video_file = st.file_uploader("Choose video", type=["mp4", "mov", "avi"], key="vid")

    if video_file:
        tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".mp4")
        tmp.write(video_file.read())
        tmp.close()

        cap          = cv2.VideoCapture(tmp.name)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

        frame_ph    = st.empty()
        progress_ph = st.progress(0)
        c1, c2, c3  = st.columns(3)
        feedback_ph = st.empty()

        frame_idx = 0
        SKIP = 2   # process every other frame for speed

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            if frame_idx % SKIP == 0:
                # run_inference returns RGB — pass directly, no channels kwarg needed
                annotated, exercise, conf, feedback = run_inference(frame)
                frame_ph.image(annotated, width=720)          # ← FIX: no channels="BGR"
                c1.metric("🏷️ Exercise",   exercise.title())
                c2.metric("📊 Confidence", f"{conf:.0%}")
                c3.metric("🔢 Reps",       st.session_state.count)
                if feedback:
                    bg = "#e6ffe6" if "✅" in feedback else "#fff3cd"
                    bd = "green"   if "✅" in feedback else "orange"
                    feedback_ph.markdown(
                        f"<div style='padding:10px;border-radius:8px;"
                        f"background:{bg};border-left:4px solid {bd}'>"
                        f"{feedback.replace(chr(10),'<br>')}</div>",
                        unsafe_allow_html=True)
            progress_ph.progress(min(frame_idx / max(total_frames, 1), 1.0))
            frame_idx += 1

        cap.release()
        os.unlink(tmp.name)
        st.balloons()
        st.success(f"🎉 Done! Total reps: **{st.session_state.count}** | "
                   f"Calories burned: **{st.session_state.total_calories:.1f} kcal**")

# ══════════════════════════════════════════
# TAB 2 — LIVE WEBCAM + AUTO-CLASSIFY
# FIX 3: base64 frame sent via Streamlit query_params instead of
#         postMessage (which never reaches Python).
#         The JS sends a base64 JPEG to ?frame=<data> and the
#         Python side reads st.query_params["frame"] each rerun.
# ══════════════════════════════════════════
with tab2:
    st.subheader("🎥 Live Webcam — Real-Time Auto-Classification")
    st.info("Click **Start Camera**, then hit **Analyse Frame** to auto-detect your exercise and count reps.")

    # ── JavaScript webcam widget ──────────────────────────────────
    webcam_html = """
    <div style="display:flex;flex-direction:column;align-items:center;gap:12px;padding:8px 0">
      <video id="vid" width="480" height="360" autoplay playsinline
             style="border-radius:12px;border:3px solid #4ade80;background:#000;"></video>
      <canvas id="canvas" width="480" height="360" style="display:none;"></canvas>

      <div style="display:flex;gap:10px;flex-wrap:wrap;justify-content:center">
        <button id="btnStart" onclick="startCam()"
          style="padding:10px 24px;background:#4ade80;border:none;border-radius:8px;
                 font-size:15px;font-weight:bold;cursor:pointer">
          ▶ Start Camera
        </button>
        <button id="btnCapture" onclick="captureFrame()" disabled
          style="padding:10px 24px;background:#60a5fa;color:#fff;border:none;border-radius:8px;
                 font-size:15px;font-weight:bold;cursor:pointer;opacity:0.5">
          📸 Analyse Frame
        </button>
        <button onclick="stopCam()"
          style="padding:10px 24px;background:#f87171;border:none;border-radius:8px;
                 font-size:15px;font-weight:bold;cursor:pointer;color:white">
          ⏹ Stop
        </button>
      </div>

      <label style="display:flex;align-items:center;gap:8px;font-size:13px;color:#555">
        <input type="checkbox" id="autoCapture" onchange="toggleAuto(this.checked)">
        Auto-capture every 2 seconds
      </label>

      <div id="status"
           style="color:#4ade80;font-size:13px;font-weight:bold;min-height:20px;text-align:center">
        Click "Start Camera" to begin
      </div>
    </div>

    <script>
    let stream = null;
    let autoTimer = null;
    const vid    = document.getElementById('vid');
    const canvas = document.getElementById('canvas');
    const ctx    = canvas.getContext('2d');
    const status = document.getElementById('status');
    const btnCap = document.getElementById('btnCapture');

    async function startCam() {
      try {
        stream = await navigator.mediaDevices.getUserMedia({video: true, audio: false});
        vid.srcObject = stream;
        btnCap.disabled = false;
        btnCap.style.opacity = '1';
        status.innerText = '🟢 Camera active';
        status.style.color = '#4ade80';
      } catch(e) {
        status.innerText = '❌ Camera error: ' + e.message;
        status.style.color = '#f87171';
      }
    }

    function captureFrame() {
      if (!stream) { status.innerText = '⚠️ Start camera first'; return; }
      ctx.drawImage(vid, 0, 0, 480, 360);
      const b64 = canvas.toDataURL('image/jpeg', 0.8).split(',')[1];

      // Send to Streamlit via URL hash — Python polls st.query_params
      const url = new URL(window.parent.location.href);
      url.searchParams.set('webcam_frame', b64);
      window.parent.history.replaceState(null, '', url.toString());

      // Also trigger a Streamlit rerun via the hidden component value trick
      window.parent.postMessage({
        type: 'streamlit:setComponentValue',
        value: Date.now()   // just a changing value to trigger rerun
      }, '*');

      status.innerText = '📤 Frame sent for analysis…';
    }

    function toggleAuto(on) {
      if (on) {
        autoTimer = setInterval(captureFrame, 2000);
        status.innerText = '🔄 Auto-capturing every 2s';
      } else {
        clearInterval(autoTimer);
        status.innerText = '⏸ Auto-capture off';
      }
    }

    function stopCam() {
      if(stream) { stream.getTracks().forEach(t => t.stop()); stream = null; }
      clearInterval(autoTimer);
      btnCap.disabled = true;
      btnCap.style.opacity = '0.5';
      status.innerText = '⏹ Camera stopped';
      status.style.color = '#aaa';
    }
    </script>
    """

    from streamlit.components.v1 import html as st_html
    st_html(webcam_html, height=520)

    st.markdown("---")

    # ── Camera-input fallback (reliable in Colab / Streamlit Cloud) ──
    st.markdown("### 📸 Capture a frame for instant analysis")
    st.caption("Works everywhere — even if the webcam widget above is blocked by the browser.")

    col_upload, col_result = st.columns([1, 1])

    with col_upload:
        live_frame = st.camera_input("📷 Take a photo")

    with col_result:
        if live_frame:
            file_bytes = np.asarray(bytearray(live_frame.read()), dtype=np.uint8)
            frame      = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)

            # run_inference returns RGB — display directly
            annotated, exercise, conf, feedback = run_inference(frame)
            st.image(annotated, width=400)   # ← FIX: no channels="BGR"

            r1, r2 = st.columns(2)
            r1.metric("🏷️ Exercise",   exercise.title())
            r2.metric("📊 Confidence", f"{conf:.0%}")
            r3, r4 = st.columns(2)
            r3.metric("🔢 Reps",     st.session_state.count)
            r4.metric("🔥 Calories", f"{st.session_state.total_calories:.1f}")

            if feedback:
                bg = "#e6ffe6" if "✅" in feedback else "#fff3cd"
                bd = "green"   if "✅" in feedback else "orange"
                st.markdown(
                    f"<div style='padding:12px;border-radius:8px;"
                    f"background:{bg};border-left:5px solid {bd}'>"
                    f"<b>Form Feedback:</b><br>"
                    f"{feedback.replace(chr(10),'<br>')}</div>",
                    unsafe_allow_html=True)

            if st.session_state.angle_buffer:
                ang = st.session_state.angle_buffer[-1]
                st.markdown(f"**Joint Angle: `{ang:.1f}°`**")
                st.progress(min(int(ang), 180) / 180)

            score_color = "🟢" if "✅" in feedback else "🟡" if "⚠️" in feedback else "🔴"
            st.markdown(f"**Form Score:** {score_color} "
                        f"{'Good form!' if '✅' in feedback else 'Needs correction' if '⚠️' in feedback else 'Detecting...'}")
        else:
            st.markdown("""
            ### 👈 Instructions
            1. Allow camera access when prompted
            2. Stand so your **full body** is visible
            3. Click **'Take photo'** to capture a frame
            4. Each capture is analysed instantly with:
               - **Auto-exercise classification** (squat / curl / push-up / press)
               - **Rep counting** (angle-based, smoothed over 5 frames)
               - **Form feedback** with tips
            5. Repeat at the top and bottom of each rep

            > 💡 **Pro tip:** Good lighting + 2–3 m distance = best accuracy
            """)

# ══════════════════════════════════════════
# TAB 3 — CHATBOT
# ══════════════════════════════════════════
with tab3:
    st.subheader("🤖 Fitness Assistant Chatbot")
    st.info("Ask about form, exercises, diet, recovery, or rep counting!")

    def chatbot_reply(msg):
        m = msg.lower()
        if any(w in m for w in ["squat","leg","knee","depth"]):
            return ("🦵 **Squat Form:**\n"
                    "- Chest up, back straight\n- Knees track over toes\n"
                    "- Thighs parallel at bottom\n- Weight in heels\n\n"
                    "💡 AI counts rep only when angle <85° then >155°.")
        elif any(w in m for w in ["curl","bicep","biceps","arm"]):
            return ("💪 **Bicep Curl Form:**\n"
                    "- Elbows pinned to sides\n- Squeeze at top 1 sec\n"
                    "- Lower slowly 3 sec\n\n"
                    "💡 AI uses elbow angle: <45° = down, >145° = up.")
        elif any(w in m for w in ["push","pushup","chest","tricep"]):
            return ("🤸 **Push-Up Form:**\n"
                    "- Hands shoulder-width\n- Body straight line\n"
                    "- Chest nearly touches floor\n\n"
                    "💡 Shoulder angle: <65° = down, >145° = up.")
        elif any(w in m for w in ["shoulder","press","overhead"]):
            return ("🏋️ **Shoulder Press Form:**\n"
                    "- Start at shoulder height\n- Press straight overhead\n"
                    "- Core braced throughout\n\n"
                    "💡 Angle: <65° = down, >145° = up.")
        elif any(w in m for w in ["count","rep","wrong","overcount"]):
            return ("🔢 **Rep Counting:**\n"
                    "- Full DOWN then UP = 1 rep\n"
                    "- Angles smoothed over 5 frames\n"
                    "- Rep counted only once per cycle\n\n"
                    "💡 Full body must be visible in frame.")
        elif any(w in m for w in ["calorie","burn","energy"]):
            return ("🔥 **Calorie Estimates per rep:**\n"
                    "- Squat: ~0.32 kcal\n"
                    "- Push-up: ~0.29 kcal\n"
                    "- Shoulder Press: ~0.19 kcal\n"
                    "- Bicep Curl: ~0.14 kcal\n\n"
                    "💡 Check your dashboard in the sidebar for total calories burned!")
        elif any(w in m for w in ["diet","protein","food","eat","nutrition"]):
            return ("🥗 **Nutrition:**\n"
                    "- Protein: 1.6–2.2g per kg bodyweight\n"
                    "- Muscle gain: +200–300 kcal surplus\n"
                    "- Fat loss: -300–500 kcal deficit\n"
                    "- Best sources: eggs, chicken, fish, legumes")
        elif any(w in m for w in ["rest","recovery","sleep","sore"]):
            return ("😴 **Recovery:**\n"
                    "- 7–9 hours sleep\n- 48hr rest per muscle group\n"
                    "- Stay hydrated and hit protein targets")
        elif any(w in m for w in ["warm","warmup","stretch"]):
            return ("🔥 **Warm-Up (5–10 min):**\n"
                    "- 2 min light cardio\n- Arm circles, leg swings\n"
                    "- Bodyweight reps of main lift\n\n"
                    "⚠️ Dynamic BEFORE, static AFTER.")
        else:
            return ("🤖 I can help with:\n"
                    "- **Form**: squat, curl, push-up, shoulder press\n"
                    "- **Calories** burned\n- **Diet & nutrition**\n"
                    "- **Recovery & sleep**\n- **Warm-up** routines\n\n"
                    "Try: *'How many calories did I burn?'* or *'Fix my squat form'*")

    for msg in st.session_state.chat_history:
        with st.chat_message(msg["role"]):
            st.markdown(msg["content"])

    user_input = st.chat_input("Ask your fitness question...")
    if user_input:
        st.session_state.chat_history.append({"role": "user", "content": user_input})
        with st.chat_message("user"):
            st.markdown(user_input)
        reply = chatbot_reply(user_input)
        st.session_state.chat_history.append({"role": "assistant", "content": reply})
        with st.chat_message("assistant"):
            st.markdown(reply)

    if st.button("🗑️ Clear Chat"):
        st.session_state.chat_history = []
        st.rerun()
```

```python
%%writefile app.py
import streamlit as st
import cv2
import numpy as np
import tensorflow as tf
import os
import tempfile
import time
from collections import deque
from mediapipe.python.solutions import pose as mp_pose
from mediapipe.python.solutions import drawing_utils as mp_drawing

# ─────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────
st.set_page_config(page_title="AI Fitness Trainer", layout="wide")
st.title("🏋️‍♂️ AI Virtual Fitness Trainer")

actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

THRESHOLDS = {
    'squat':               {'down': 85,  'up': 155, 'joints': (23, 25, 27)},
    'barbell biceps curl': {'down': 45,  'up': 145, 'joints': (11, 13, 15)},
    'push-up':             {'down': 65,  'up': 145, 'joints': (11, 13, 15)},
    'shoulder press':      {'down': 65,  'up': 145, 'joints': (13, 11, 23)},
}

FORM_RULES = {
    'squat': {
        'good_range': (70, 95),
        'good_msg':  '✅ Great squat depth! Keep chest up.',
        'high_msg':  '⚠️ Go deeper — thighs should reach parallel.',
        'low_msg':   '⚠️ Too deep — stop at parallel to protect knees.',
        'tip':       '💡 Keep knees over toes, weight in heels.',
    },
    'barbell biceps curl': {
        'good_range': (35, 55),
        'good_msg':  '✅ Full range curl! Great squeeze.',
        'high_msg':  '⚠️ Curl higher — squeeze at the top.',
        'low_msg':   '⚠️ Lower fully to complete the rep.',
        'tip':       '💡 Keep elbows pinned to your sides.',
    },
    'push-up': {
        'good_range': (55, 80),
        'good_msg':  '✅ Good push-up depth!',
        'high_msg':  '⚠️ Lower chest closer to the ground.',
        'low_msg':   "⚠️ Don't collapse — keep core tight.",
        'tip':       '💡 Body must be a straight line.',
    },
    'shoulder press': {
        'good_range': (55, 80),
        'good_msg':  '✅ Great press! Full extension.',
        'high_msg':  '⚠️ Press fully overhead — extend arms.',
        'low_msg':   '⚠️ Lower to shoulder level between reps.',
        'tip':       '💡 Brace core to protect lower back.',
    },
}

CALORIES_PER_REP = {
    'squat': 0.32,
    'barbell biceps curl': 0.14,
    'push-up': 0.29,
    'shoulder press': 0.19,
}

# ─────────────────────────────────────────
# LOAD RESOURCES
# ─────────────────────────────────────────
@st.cache_resource
def load_model():
    if not os.path.exists("fitness_trainer_model.h5"):
        st.error("❌ Model not found!")
        st.stop()
    return tf.keras.models.load_model("fitness_trainer_model.h5")

@st.cache_resource
def load_pose():
    return mp_pose.Pose(
        min_detection_confidence=0.6,
        min_tracking_confidence=0.6,
        model_complexity=1
    )

model      = load_model()
pose_model = load_pose()

# ─────────────────────────────────────────
# SESSION STATE
# ─────────────────────────────────────────
defaults = {
    "count": 0, "stage": None,
    "last_feedback": "", "last_exercise": "",
    "rep_confirmed": False,
    "angle_buffer": deque(maxlen=5),
    "chat_history": [],
    "total_calories": 0.0,
    "session_start": time.time(),
}
for k, v in defaults.items():
    if k not in st.session_state:
        st.session_state[k] = v

# ─────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    return 360 - angle if angle > 180 else angle

def get_angle(exercise, lm):
    if exercise not in THRESHOLDS:
        return 0
    i, j, k = THRESHOLDS[exercise]['joints']
    return calculate_angle([lm[i].x, lm[i].y], [lm[j].x, lm[j].y], [lm[k].x, lm[k].y])

def smooth_angle(raw_angle):
    st.session_state.angle_buffer.append(raw_angle)
    return float(np.mean(st.session_state.angle_buffer))

def get_form_feedback(exercise, angle):
    if exercise not in FORM_RULES:
        return ""
    r = FORM_RULES[exercise]
    lo, hi = r['good_range']
    if lo <= angle <= hi:
        return r['good_msg'] + "\n" + r['tip']
    elif angle > hi:
        return r['high_msg'] + "\n" + r['tip']
    else:
        return r['low_msg'] + "\n" + r['tip']

def update_rep_count(exercise, smoothed_angle):
    thresh      = THRESHOLDS.get(exercise, {})
    down_thresh = thresh.get('down', 90)
    up_thresh   = thresh.get('up',  150)

    if exercise != st.session_state.last_exercise:
        st.session_state.count         = 0
        st.session_state.stage         = None
        st.session_state.rep_confirmed = False
        st.session_state.last_exercise = exercise
        st.session_state.angle_buffer.clear()
        st.session_state.total_calories = 0.0

    feedback = st.session_state.last_feedback

    if smoothed_angle <= down_thresh:
        if st.session_state.stage != "down":
            st.session_state.stage         = "down"
            st.session_state.rep_confirmed = False
            feedback = get_form_feedback(exercise, smoothed_angle)
            st.session_state.last_feedback = feedback

    elif smoothed_angle >= up_thresh:
        if st.session_state.stage == "down" and not st.session_state.rep_confirmed:
            st.session_state.count        += 1
            st.session_state.stage         = "up"
            st.session_state.rep_confirmed = True
            st.session_state.total_calories += CALORIES_PER_REP.get(exercise, 0.2)

    return feedback

def draw_overlay(frame, exercise, confidence, angle, count, feedback, stage):
    h, w = frame.shape[:2]
    cv2.rectangle(frame, (0, 0), (w, 55), (20, 20, 20), -1)
    cv2.putText(frame, f"{exercise.upper()}  {confidence:.0%}",
                (10, 38), cv2.FONT_HERSHEY_DUPLEX, 0.85, (0, 255, 255), 2)
    cv2.rectangle(frame, (w-150, 0), (w, 95), (180, 60, 20), -1)
    cv2.putText(frame, "REPS",  (w-130, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255,255,255), 2)
    cv2.putText(frame, str(count), (w-110, 85), cv2.FONT_HERSHEY_DUPLEX, 2.2, (255,255,255), 3)
    cv2.rectangle(frame, (0, h-85), (400, h), (20, 20, 20), -1)
    cv2.putText(frame, f"Angle: {angle:.1f}  Stage: {stage or 'ready'}",
                (8, h-52), cv2.FONT_HERSHEY_SIMPLEX, 0.65, (200,200,0), 2)
    if feedback:
        line  = feedback.split('\n')[0]
        color = (0, 210, 0) if "✅" in line else (0, 140, 255)
        cv2.putText(frame, line, (8, h-18), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)
    return frame

def run_inference(frame):
    rgb     = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose_model.process(rgb)
    exercise, confidence, angle, feedback = "No pose detected", 0.0, 0, ""

    if results.pose_landmarks:
        lm     = results.pose_landmarks.landmark
        coords = np.array([[l.x, l.y, l.z] for l in lm]).flatten()
        pred       = model.predict(coords.reshape(1, 1, 99), verbose=0)
        exercise   = actions[np.argmax(pred)]
        confidence = float(np.max(pred))
        raw_angle  = get_angle(exercise, lm)
        angle      = smooth_angle(raw_angle)
        feedback   = update_rep_count(exercise, angle)

        mp_drawing.draw_landmarks(
            rgb, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing.DrawingSpec(color=(200,0,0), thickness=2, circle_radius=3),
            connection_drawing_spec=mp_drawing.DrawingSpec(color=(255,255,255), thickness=2)
        )

    bgr = cv2.cvtColor(rgb, cv2.COLOR_RGB2BGR)
    bgr = draw_overlay(bgr, exercise, confidence, angle,
                       st.session_state.count, feedback, st.session_state.stage)
    return cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB), exercise, confidence, feedback

# ─────────────────────────────────────────
# SIDEBAR
# ─────────────────────────────────────────
st.sidebar.header("🎛️ Controls")
if st.sidebar.button("🔄 Reset Session"):
    st.session_state.count          = 0
    st.session_state.stage          = None
    st.session_state.last_exercise  = ""
    st.session_state.last_feedback  = ""
    st.session_state.rep_confirmed  = False
    st.session_state.total_calories = 0.0
    st.session_state.session_start  = time.time()
    st.session_state.angle_buffer.clear()
    st.sidebar.success("Session reset!")

st.sidebar.markdown("---")
elapsed = int(time.time() - st.session_state.session_start)
mins, secs = divmod(elapsed, 60)
st.sidebar.markdown("### 📊 Workout Dashboard")
st.sidebar.markdown(
    f"""<div style='background:#1e1e2e;padding:14px;border-radius:12px;color:white;font-family:sans-serif'>
      <div style='display:flex;justify-content:space-between;margin-bottom:10px'>
        <div style='text-align:center'>
          <div style='font-size:28px;font-weight:bold;color:#4ade80'>{st.session_state.count}</div>
          <div style='font-size:11px;color:#aaa'>REPS</div>
        </div>
        <div style='text-align:center'>
          <div style='font-size:28px;font-weight:bold;color:#f97316'>{st.session_state.total_calories:.1f}</div>
          <div style='font-size:11px;color:#aaa'>CALORIES</div>
        </div>
        <div style='text-align:center'>
          <div style='font-size:28px;font-weight:bold;color:#60a5fa'>{mins:02d}:{secs:02d}</div>
          <div style='font-size:11px;color:#aaa'>TIME</div>
        </div>
      </div>
      <div style='border-top:1px solid #333;padding-top:8px;margin-top:4px'>
        <div style='font-size:12px;color:#aaa'>Current Exercise</div>
        <div style='font-size:16px;font-weight:bold;color:#e879f9'>
          {st.session_state.last_exercise.title() if st.session_state.last_exercise else "Not detected yet"}
        </div>
        <div style='font-size:12px;color:#aaa;margin-top:6px'>Stage</div>
        <div style='font-size:16px;font-weight:bold;color:#facc15'>
          {(st.session_state.stage or "Ready").upper()}
        </div>
      </div>
    </div>""", unsafe_allow_html=True)

if st.session_state.last_feedback:
    bg = "#e6ffe6" if "✅" in st.session_state.last_feedback else "#fff3cd"
    bd = "green"   if "✅" in st.session_state.last_feedback else "orange"
    st.sidebar.markdown("**💬 Last Form Feedback:**")
    st.sidebar.markdown(
        f"<div style='padding:8px;border-radius:6px;background:{bg};border-left:4px solid {bd}'>"
        f"{st.session_state.last_feedback.replace(chr(10),'<br>')}</div>",
        unsafe_allow_html=True)

st.sidebar.markdown("---")
st.sidebar.markdown("**🏋️ Supported Exercises:**")
for a in actions:
    st.sidebar.markdown(f"- {a.title()}")

# ─────────────────────────────────────────
# TABS
# ─────────────────────────────────────────
tab1, tab2, tab3 = st.tabs(["📹 Video Analysis", "🎥 Live Webcam", "🤖 Fitness Chatbot"])

# ══════════════════════════════════════════
# TAB 1 — VIDEO UPLOAD
# ══════════════════════════════════════════
with tab1:
    st.subheader("📹 Upload a Workout Video")
    st.info("Upload your workout video. AI classifies exercise, counts reps, and gives form feedback.")

    video_file = st.file_uploader("Choose video", type=["mp4", "mov", "avi"], key="vid")

    if video_file:
        tmp = tempfile.NamedTemporaryFile(delete=False, suffix=".mp4")
        tmp.write(video_file.read())
        tmp.close()

        cap          = cv2.VideoCapture(tmp.name)
        total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        frame_ph    = st.empty()
        progress_ph = st.progress(0)
        c1, c2, c3  = st.columns(3)
        feedback_ph = st.empty()
        frame_idx = 0

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret:
                break
            if frame_idx % 2 == 0:
                annotated, exercise, conf, feedback = run_inference(frame)
                frame_ph.image(annotated, width=720)
                c1.metric("🏷️ Exercise",   exercise.title())
                c2.metric("📊 Confidence", f"{conf:.0%}")
                c3.metric("🔢 Reps",       st.session_state.count)
                if feedback:
                    bg = "#e6ffe6" if "✅" in feedback else "#fff3cd"
                    bd = "green"   if "✅" in feedback else "orange"
                    feedback_ph.markdown(
                        f"<div style='padding:10px;border-radius:8px;"
                        f"background:{bg};border-left:4px solid {bd}'>"
                        f"{feedback.replace(chr(10),'<br>')}</div>",
                        unsafe_allow_html=True)
            progress_ph.progress(min(frame_idx / max(total_frames, 1), 1.0))
            frame_idx += 1

        cap.release()
        os.unlink(tmp.name)
        st.balloons()
        st.success(f"🎉 Done! Total reps: **{st.session_state.count}** | "
                   f"Calories burned: **{st.session_state.total_calories:.1f} kcal**")

# ══════════════════════════════════════════
# TAB 2 — LIVE WEBCAM (FIXED — no JS, pure st.camera_input)
#
# WHY THE OLD JS VERSION DIDN'T WORK:
#   JavaScript inside a Streamlit iframe cannot trigger a Python rerun
#   or pass image data back to Python. postMessage and query params are
#   both blocked by Streamlit's sandbox. st.camera_input is the ONLY
#   reliable way to get camera frames into Python in Streamlit/Colab.
#
# HOW THIS VERSION WORKS:
#   st.camera_input captures a frame → Python decodes it → run_inference()
#   → results shown immediately. The "Auto-rerun" toggle calls st.rerun()
#   after a delay to simulate a live feed.
# ══════════════════════════════════════════
with tab2:
    st.subheader("🎥 Live Webcam — Real-Time Auto-Classification")

    st.info("📌 Allow camera access → stand **2–3 metres back** (full body visible) → click **Take Photo**")

    st.markdown(
        "<div style='background:#fff3cd;border-left:5px solid orange;padding:10px 14px;"
        "border-radius:6px;margin-bottom:12px'>⚠️ <b>Full body must be in frame — head to feet!</b> "
        "Place your device on a shelf or table 2–3 m away. "
        "Upper-body-only frames = no pose detected.</div>",
        unsafe_allow_html=True)

    # Auto-rerun controls
    ctrl1, ctrl2 = st.columns([1, 1])
    auto_rerun  = ctrl1.toggle("🔄 Auto-rerun (live mode)", value=False)
    rerun_speed = ctrl2.selectbox("Interval", ["Fast (1s)", "Normal (2s)", "Slow (3s)"],
                                  index=1, label_visibility="collapsed")
    delay = {"Fast (1s)": 1.0, "Normal (2s)": 2.0, "Slow (3s)": 3.0}[rerun_speed]

    cam_col, res_col = st.columns([1, 1])

    with cam_col:
        live_frame = st.camera_input(
            "📷 Take photo — ensure FULL BODY is visible",
            key="live_cam"
        )

    with res_col:
        if live_frame:
            file_bytes = np.asarray(bytearray(live_frame.read()), dtype=np.uint8)
            frame      = cv2.imdecode(file_bytes, cv2.IMREAD_COLOR)

            if frame is not None:
                annotated, exercise, conf, feedback = run_inference(frame)

                st.image(annotated, use_column_width=True)

                m1, m2 = st.columns(2)
                m1.metric("🏷️ Exercise",   exercise.title())
                m2.metric("📊 Confidence", f"{conf:.0%}")
                m3, m4 = st.columns(2)
                m3.metric("🔢 Reps",       st.session_state.count)
                m4.metric("🔥 Calories",   f"{st.session_state.total_calories:.1f} kcal")

                if exercise == "No pose detected":
                    st.error("❌ No pose detected — stand further back so your full body fits in frame.")
                else:
                    stage_val = st.session_state.stage or "ready"
                    color_map = {"down": "#f97316", "up": "#4ade80", "ready": "#60a5fa"}
                    sc = color_map.get(stage_val, "#aaa")
                    st.markdown(
                        f"<div style='display:inline-block;padding:4px 14px;border-radius:999px;"
                        f"background:{sc};color:white;font-weight:bold;font-size:13px'>"
                        f"Stage: {stage_val.upper()}</div>", unsafe_allow_html=True)

                    if st.session_state.angle_buffer:
                        ang = list(st.session_state.angle_buffer)[-1]
                        st.markdown(f"**Joint angle: `{ang:.1f}°`**")
                        st.progress(min(int(ang), 180) / 180)

                    if feedback:
                        bg = "#e6ffe6" if "✅" in feedback else "#fff3cd"
                        bd = "green"   if "✅" in feedback else "orange"
                        st.markdown(
                            f"<div style='padding:12px;border-radius:8px;"
                            f"background:{bg};border-left:5px solid {bd};margin-top:8px'>"
                            f"<b>Form Feedback:</b><br>{feedback.replace(chr(10),'<br>')}</div>",
                            unsafe_allow_html=True)

                    thresh = THRESHOLDS.get(exercise, {})
                    st.caption(
                        f"Rep counter: angle ≤{thresh.get('down','?')}° = DOWN  |  "
                        f"angle ≥{thresh.get('up','?')}° = UP  →  DOWN + UP = 1 rep")
        else:
            st.markdown("""
            **Setup checklist:**
            - ☐ Device placed on shelf/table (not hand-held)
            - ☐ Standing 2–3 metres away
            - ☐ Full body head-to-feet visible in preview
            - ☐ Good lighting (face a window)
            - ☐ Camera allowed in browser

            **To count reps:**
            Take a photo at the **bottom** of the movement (DOWN),
            then at the **top** (UP). Each complete cycle = 1 rep.
            Enable **Auto-rerun** to do this automatically.
            """)

    # Auto-rerun loop — simulates live feed
    if auto_rerun and live_frame is not None:
        time.sleep(delay)
        st.rerun()

# ══════════════════════════════════════════
# TAB 3 — CHATBOT
# ══════════════════════════════════════════
with tab3:
    st.subheader("🤖 Fitness Assistant Chatbot")
    st.info("Ask about form, exercises, diet, recovery, or rep counting!")

    def chatbot_reply(msg):
        m = msg.lower()
        if any(w in m for w in ["squat","leg","knee","depth"]):
            return ("🦵 **Squat Form:**\n- Chest up, back straight\n- Knees track over toes\n"
                    "- Thighs parallel at bottom\n- Weight in heels\n\n💡 AI counts rep only when angle <85° then >155°.")
        elif any(w in m for w in ["curl","bicep","biceps","arm"]):
            return ("💪 **Bicep Curl Form:**\n- Elbows pinned to sides\n- Squeeze at top 1 sec\n"
                    "- Lower slowly 3 sec\n\n💡 AI uses elbow angle: <45° = down, >145° = up.")
        elif any(w in m for w in ["push","pushup","chest","tricep"]):
            return ("🤸 **Push-Up Form:**\n- Hands shoulder-width\n- Body straight line\n"
                    "- Chest nearly touches floor\n\n💡 Shoulder angle: <65° = down, >145° = up.")
        elif any(w in m for w in ["shoulder","press","overhead"]):
            return ("🏋️ **Shoulder Press Form:**\n- Start at shoulder height\n- Press straight overhead\n"
                    "- Core braced throughout\n\n💡 Angle: <65° = down, >145° = up.")
        elif any(w in m for w in ["count","rep","wrong","overcount"]):
            return ("🔢 **Rep Counting:**\n- Full DOWN then UP = 1 rep\n- Angles smoothed over 5 frames\n"
                    "- Rep counted only once per cycle\n\n💡 Full body must be visible in frame.")
        elif any(w in m for w in ["calorie","burn","energy"]):
            return ("🔥 **Calorie Estimates per rep:**\n- Squat: ~0.32 kcal\n- Push-up: ~0.29 kcal\n"
                    "- Shoulder Press: ~0.19 kcal\n- Bicep Curl: ~0.14 kcal\n\n💡 Check your sidebar dashboard!")
        elif any(w in m for w in ["diet","protein","food","eat","nutrition"]):
            return ("🥗 **Nutrition:**\n- Protein: 1.6–2.2g per kg bodyweight\n- Muscle gain: +200–300 kcal surplus\n"
                    "- Fat loss: -300–500 kcal deficit\n- Best sources: eggs, chicken, fish, legumes")
        elif any(w in m for w in ["rest","recovery","sleep","sore"]):
            return ("😴 **Recovery:**\n- 7–9 hours sleep\n- 48hr rest per muscle group\n"
                    "- Stay hydrated and hit protein targets")
        elif any(w in m for w in ["warm","warmup","stretch"]):
            return ("🔥 **Warm-Up (5–10 min):**\n- 2 min light cardio\n- Arm circles, leg swings\n"
                    "- Bodyweight reps of main lift\n\n⚠️ Dynamic BEFORE, static AFTER.")
        else:
            return ("🤖 I can help with:\n- **Form**: squat, curl, push-up, shoulder press\n"
                    "- **Calories** burned\n- **Diet & nutrition**\n- **Recovery & sleep**\n- **Warm-up** routines\n\n"
                    "Try: *'Fix my squat form'* or *'How many calories did I burn?'*")

    for msg in st.session_state.chat_history:
        with st.chat_message(msg["role"]):
            st.markdown(msg["content"])

    user_input = st.chat_input("Ask your fitness question...")
    if user_input:
        st.session_state.chat_history.append({"role": "user", "content": user_input})
        with st.chat_message("user"):
            st.markdown(user_input)
        reply = chatbot_reply(user_input)
        st.session_state.chat_history.append({"role": "assistant", "content": reply})
        with st.chat_message("assistant"):
            st.markdown(reply)

    if st.button("🗑️ Clear Chat"):
        st.session_state.chat_history = []
        st.rerun()
```

```python
%%writefile realtime_trainer.py
"""
Real-time AI Fitness Trainer
============================
Run this in Colab. It streams your webcam through a browser
window with live skeleton overlay + exercise classification,
exactly like the reference video (Image 2).

Usage in Colab:
  !python realtime_trainer.py
"""

import cv2
import numpy as np
import tensorflow as tf
import threading
import base64
import time
from collections import deque
from flask import Flask, Response, render_template_string
from mediapipe.python.solutions import pose as mp_pose
from mediapipe.python.solutions import drawing_utils as mp_drawing

# ─────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────
actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

THRESHOLDS = {
    'squat':               {'down': 85,  'up': 155, 'joints': (23, 25, 27)},
    'barbell biceps curl': {'down': 45,  'up': 145, 'joints': (11, 13, 15)},
    'push-up':             {'down': 65,  'up': 145, 'joints': (11, 13, 15)},
    'shoulder press':      {'down': 65,  'up': 145, 'joints': (13, 11, 23)},
}

FORM_RULES = {
    'squat': {
        'good_range': (70, 95),
        'good_msg':  'Great squat depth!',
        'high_msg':  'Go deeper - thighs parallel',
        'low_msg':   'Too deep - stop at parallel',
        'tip':       'Knees over toes, weight in heels',
    },
    'barbell biceps curl': {
        'good_range': (35, 55),
        'good_msg':  'Full range curl!',
        'high_msg':  'Curl higher - squeeze at top',
        'low_msg':   'Lower fully to complete rep',
        'tip':       'Keep elbows pinned to sides',
    },
    'push-up': {
        'good_range': (55, 80),
        'good_msg':  'Good push-up depth!',
        'high_msg':  'Lower chest to ground',
        'low_msg':   'Keep core tight',
        'tip':       'Body must be a straight line',
    },
    'shoulder press': {
        'good_range': (55, 80),
        'good_msg':  'Great press! Full extension.',
        'high_msg':  'Press fully overhead',
        'low_msg':   'Lower to shoulder level',
        'tip':       'Brace core - protect lower back',
    },
}

CALORIES_PER_REP = {
    'squat': 0.32, 'barbell biceps curl': 0.14,
    'push-up': 0.29, 'shoulder press': 0.19,
}

EXERCISE_COLORS = {
    'squat':               (0, 200, 100),
    'barbell biceps curl': (200, 100, 0),
    'push-up':             (0, 150, 255),
    'shoulder press':      (180, 0, 200),
    'No pose detected':    (100, 100, 100),
}

# ─────────────────────────────────────────
# STATE
# ─────────────────────────────────────────
state = {
    'count': 0, 'stage': None,
    'last_feedback': '', 'exercise': 'No pose detected',
    'confidence': 0.0, 'angle': 0.0,
    'rep_confirmed': False, 'last_exercise': '',
    'total_calories': 0.0,
    'angle_buffer': deque(maxlen=5),
    'session_start': time.time(),
    'rep_counts': {a: 0 for a in actions},   # per-exercise counts like Image 2
}

# ─────────────────────────────────────────
# LOAD MODEL & POSE
# ─────────────────────────────────────────
print("Loading model...")
model = tf.keras.models.load_model('fitness_trainer_model.h5')
print("✅ Model loaded")

pose_model = mp_pose.Pose(
    min_detection_confidence=0.6,
    min_tracking_confidence=0.6,
    model_complexity=1
)
print("✅ Pose model loaded")

# ─────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    return 360 - angle if angle > 180 else angle

def get_angle(exercise, lm):
    if exercise not in THRESHOLDS:
        return 0
    i, j, k = THRESHOLDS[exercise]['joints']
    return calculate_angle([lm[i].x, lm[i].y], [lm[j].x, lm[j].y], [lm[k].x, lm[k].y])

def get_form_feedback(exercise, angle):
    if exercise not in FORM_RULES:
        return ''
    r = FORM_RULES[exercise]
    lo, hi = r['good_range']
    prefix = '✅' if lo <= angle <= hi else '⚠'
    msg    = r['good_msg'] if lo <= angle <= hi else (r['high_msg'] if angle > hi else r['low_msg'])
    return f"{prefix} {msg} | {r['tip']}"

def update_rep_count(exercise, angle):
    thresh      = THRESHOLDS.get(exercise, {})
    down_thresh = thresh.get('down', 90)
    up_thresh   = thresh.get('up', 150)

    if exercise != state['last_exercise']:
        state['count']         = 0
        state['stage']         = None
        state['rep_confirmed'] = False
        state['last_exercise'] = exercise
        state['angle_buffer'].clear()

    if angle <= down_thresh:
        if state['stage'] != 'down':
            state['stage']         = 'down'
            state['rep_confirmed'] = False
            state['last_feedback'] = get_form_feedback(exercise, angle)
    elif angle >= up_thresh:
        if state['stage'] == 'down' and not state['rep_confirmed']:
            state['count']        += 1
            state['rep_counts'][exercise] += 1
            state['stage']         = 'up'
            state['rep_confirmed'] = True
            state['total_calories'] += CALORIES_PER_REP.get(exercise, 0.2)

# ─────────────────────────────────────────
# DRAW OVERLAY  (matches Image 2 style)
# ─────────────────────────────────────────
def draw_overlay(frame, exercise, confidence, angle):
    h, w = frame.shape[:2]
    ex_color = EXERCISE_COLORS.get(exercise, (200, 200, 200))

    # ── Top bar: exercise name ──────────────────────────
    cv2.rectangle(frame, (0, 0), (w, 50), (15, 15, 15), -1)
    cv2.putText(frame, f"Exercise: {exercise}",
                (w//2 - 200, 35), cv2.FONT_HERSHEY_DUPLEX, 1.0, (255, 255, 255), 2)

    # ── Left panel: per-exercise rep counts (like Image 2) ──
    panel_w = 190
    cv2.rectangle(frame, (0, 50), (panel_w, h), (10, 10, 10), -1)

    y = 100
    for i, act in enumerate(actions):
        short = {'barbell biceps curl': 'Curl', 'push-up': 'Push-up',
                 'shoulder press': 'Press', 'squat': 'Squat'}[act]
        is_current = (act == exercise)
        color = ex_color if is_current else (160, 160, 160)
        weight = 2 if is_current else 1
        cv2.putText(frame, f"{short}: {state['rep_counts'][act]}",
                    (8, y), cv2.FONT_HERSHEY_SIMPLEX, 0.75, color, weight)
        if is_current:
            cv2.rectangle(frame, (0, y-25), (panel_w-5, y+8), ex_color, 1)
        y += 70

    # ── Right panel: current rep count ──────────────────
    cv2.rectangle(frame, (w-160, 0), (w, 100), (20, 60, 180), -1)
    cv2.putText(frame, "REPS", (w-130, 28), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (200,200,200), 1)
    cv2.putText(frame, str(state['count']), (w-120, 88),
                cv2.FONT_HERSHEY_DUPLEX, 2.5, (255, 255, 255), 3)

    # ── Bottom bar: angle + stage + feedback ────────────
    cv2.rectangle(frame, (panel_w, h-90), (w, h), (15, 15, 15), -1)

    stage_str  = (state['stage'] or 'ready').upper()
    stage_color = (0,200,80) if state['stage'] == 'up' else (0,120,255) if state['stage'] == 'down' else (200,200,0)
    cv2.putText(frame, f"Angle: {angle:.0f}deg  |  Stage: {stage_str}",
                (panel_w+8, h-58), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (200, 200, 0), 1)
    cv2.putText(frame, f"Stage: {stage_str}",
                (panel_w+8, h-58), cv2.FONT_HERSHEY_SIMPLEX, 0.6, stage_color, 2)

    feedback = state['last_feedback']
    if feedback:
        fb_color = (0, 210, 0) if '✅' in feedback else (0, 150, 255)
        # Truncate to fit
        fb_text = feedback[:80]
        cv2.putText(frame, fb_text, (panel_w+8, h-20),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.52, fb_color, 1)

    # ── Calories + timer (bottom right) ─────────────────
    elapsed = int(time.time() - state['session_start'])
    mins, secs = divmod(elapsed, 60)
    cv2.putText(frame, f"Cal: {state['total_calories']:.1f}  |  {mins:02d}:{secs:02d}",
                (w-260, h-8), cv2.FONT_HERSHEY_SIMPLEX, 0.55, (150,200,255), 1)

    # ── Confidence bar ───────────────────────────────────
    conf_w = int((w - panel_w - 160) * confidence)
    cv2.rectangle(frame, (panel_w, h-4), (panel_w + conf_w, h), ex_color, -1)
    cv2.putText(frame, f"{confidence:.0%}",
                (panel_w + conf_w + 4, h-6), cv2.FONT_HERSHEY_SIMPLEX, 0.4, (200,200,200), 1)

    return frame

# ─────────────────────────────────────────
# INFERENCE
# ─────────────────────────────────────────
def process_frame(frame):
    rgb     = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = pose_model.process(rgb)

    exercise, confidence, angle = 'No pose detected', 0.0, 0.0

    if results.pose_landmarks:
        lm     = results.pose_landmarks.landmark
        coords = np.array([[l.x, l.y, l.z] for l in lm]).flatten()
        pred       = model.predict(coords.reshape(1, 1, 99), verbose=0)
        exercise   = actions[np.argmax(pred)]
        confidence = float(np.max(pred))

        raw_angle = get_angle(exercise, lm)
        state['angle_buffer'].append(raw_angle)
        angle = float(np.mean(state['angle_buffer']))

        update_rep_count(exercise, angle)

        state['exercise']   = exercise
        state['confidence'] = confidence
        state['angle']      = angle

        # Draw skeleton on BGR frame directly
        mp_drawing.draw_landmarks(
            frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing.DrawingSpec(
                color=(0, 0, 220), thickness=2, circle_radius=4),
            connection_drawing_spec=mp_drawing.DrawingSpec(
                color=(255, 255, 255), thickness=2)
        )
    else:
        state['exercise']   = 'No pose detected'
        state['confidence'] = 0.0

    frame = draw_overlay(frame, exercise, confidence, angle)
    return frame

# ─────────────────────────────────────────
# FLASK APP — streams MJPEG to browser
# ─────────────────────────────────────────
app = Flask(__name__)

HTML_PAGE = """
<!DOCTYPE html>
<html>
<head>
  <title>AI Fitness Trainer — Live</title>
  <style>
    * { margin:0; padding:0; box-sizing:border-box; }
    body { background:#0d0d0d; color:white; font-family:sans-serif; }
    .header { background:#111; padding:12px 20px; display:flex;
              align-items:center; justify-content:space-between; }
    .header h1 { font-size:20px; color:#4ade80; }
    .main { display:flex; gap:0; height:calc(100vh - 50px); }
    .video-wrap { flex:1; display:flex; align-items:center;
                  justify-content:center; background:#000; }
    .video-wrap img { max-width:100%; max-height:100%; object-fit:contain; }
    .sidebar { width:240px; background:#111; padding:16px;
               overflow-y:auto; border-left:1px solid #222; }
    .stat { background:#1a1a2e; border-radius:8px; padding:12px;
            margin-bottom:12px; text-align:center; }
    .stat-val { font-size:32px; font-weight:bold; }
    .stat-label { font-size:11px; color:#888; margin-top:2px; }
    .green { color:#4ade80; } .orange { color:#f97316; } .blue { color:#60a5fa; }
    .ex-tag { background:#1e3a5f; border-radius:6px; padding:8px 10px;
              font-size:13px; margin-bottom:8px; color:#60a5fa; }
    .tip { font-size:11px; color:#aaa; margin-top:8px; line-height:1.5; }
    .reset-btn { width:100%; padding:10px; background:#ef4444; border:none;
                 border-radius:6px; color:white; font-size:14px; font-weight:bold;
                 cursor:pointer; margin-top:8px; }
    .reset-btn:hover { background:#dc2626; }
  </style>
</head>
<body>
  <div class="header">
    <h1>🏋️ AI Virtual Fitness Trainer — Live</h1>
    <span style="color:#aaa;font-size:13px">Stand 2–3m back · Full body in frame</span>
  </div>
  <div class="main">
    <div class="video-wrap">
      <img src="/video_feed" />
    </div>
    <div class="sidebar">
      <div class="stat"><div class="stat-val green" id="reps">0</div>
        <div class="stat-label">REPS (current)</div></div>
      <div class="stat"><div class="stat-val orange" id="cal">0.0</div>
        <div class="stat-label">CALORIES</div></div>
      <div class="stat"><div class="stat-val blue" id="timer">00:00</div>
        <div class="stat-label">TIME</div></div>
      <div class="ex-tag">Exercise: <b id="exercise">Detecting...</b></div>
      <div class="ex-tag">Confidence: <b id="conf">—</b></div>
      <div class="ex-tag">Stage: <b id="stage">Ready</b></div>
      <div class="ex-tag">Angle: <b id="angle">—</b></div>
      <div id="feedback" class="tip"></div>
      <button class="reset-btn" onclick="fetch('/reset')">🔄 Reset Session</button>
      <div style="margin-top:16px;font-size:11px;color:#555">
        <b style="color:#888">All-exercise rep counts:</b><br>
        <span id="allreps"></span>
      </div>
    </div>
  </div>

  <script>
    const start = Date.now();
    function pad(n){ return String(n).padStart(2,'0'); }

    function updateSidebar() {
      fetch('/state')
        .then(r => r.json())
        .then(d => {
          document.getElementById('reps').innerText    = d.count;
          document.getElementById('cal').innerText     = d.calories;
          document.getElementById('exercise').innerText= d.exercise;
          document.getElementById('conf').innerText    = d.confidence;
          document.getElementById('stage').innerText   = (d.stage||'Ready').toUpperCase();
          document.getElementById('angle').innerText   = d.angle + '°';
          document.getElementById('feedback').innerText= d.feedback;

          let elapsed = Math.floor((Date.now()-start)/1000);
          let m=Math.floor(elapsed/60), s=elapsed%60;
          document.getElementById('timer').innerText = pad(m)+':'+pad(s);

          let ar = '';
          for(const [k,v] of Object.entries(d.rep_counts)){
            ar += k + ': ' + v + '<br>';
          }
          document.getElementById('allreps').innerHTML = ar;
        });
    }
    setInterval(updateSidebar, 500);
  </script>
</body>
</html>
"""

cap = None
frame_lock = threading.Lock()
latest_frame = None

def camera_loop():
    global cap, latest_frame
    cap = cv2.VideoCapture(0)
    cap.set(cv2.CAP_PROP_FRAME_WIDTH,  1280)
    cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)
    cap.set(cv2.CAP_PROP_FPS, 30)

    if not cap.isOpened():
        print("❌ Camera not found — make sure camera is allowed in Colab")
        return

    print("✅ Camera opened")
    while True:
        ret, frame = cap.read()
        if not ret:
            time.sleep(0.05)
            continue
        processed = process_frame(frame)
        with frame_lock:
            latest_frame = processed.copy()

def generate_frames():
    while True:
        with frame_lock:
            frame = latest_frame
        if frame is None:
            time.sleep(0.03)
            continue
        ret, buf = cv2.imencode('.jpg', frame, [cv2.IMWRITE_JPEG_QUALITY, 80])
        if not ret:
            continue
        yield (b'--frame\r\n'
               b'Content-Type: image/jpeg\r\n\r\n' + buf.tobytes() + b'\r\n')
        time.sleep(0.033)  # ~30 fps

@app.route('/')
def index():
    return render_template_string(HTML_PAGE)

@app.route('/video_feed')
def video_feed():
    return Response(generate_frames(),
                    mimetype='multipart/x-mixed-replace; boundary=frame')

@app.route('/state')
def get_state():
    import json
    return json.dumps({
        'count':      state['count'],
        'calories':   f"{state['total_calories']:.1f}",
        'exercise':   state['exercise'],
        'confidence': f"{state['confidence']:.0%}",
        'stage':      state['stage'],
        'angle':      f"{state['angle']:.1f}",
        'feedback':   state['last_feedback'],
        'rep_counts': state['rep_counts'],
    })

@app.route('/reset')
def reset():
    state.update({
        'count': 0, 'stage': None, 'last_feedback': '',
        'rep_confirmed': False, 'last_exercise': '',
        'total_calories': 0.0, 'session_start': time.time(),
        'rep_counts': {a: 0 for a in actions},
    })
    state['angle_buffer'].clear()
    return 'OK'

if __name__ == '__main__':
    # Start camera in background thread
    t = threading.Thread(target=camera_loop, daemon=True)
    t.start()
    time.sleep(2)
    print("🚀 Server starting on port 5000...")
    app.run(host='0.0.0.0', port=5000, threaded=True)
```

```python
"""
Real-time AI Fitness Trainer — Google Colab Edition
=====================================================
NO manual photo-taking. The browser captures webcam frames
automatically every ~100ms and sends them to the backend.
The backend runs MediaPipe + your model and streams results back.

HOW TO USE IN COLAB:
─────────────────────
Step 1 — Upload this file + your model:
    fitness_trainer_model.h5

Step 2 — Install deps (run in a Colab cell):
    !pip install flask flask-cors mediapipe tensorflow pyngrok -q

Step 3 — Run (in a Colab cell):
    !python realtime_trainer_colab.py

Step 4 — Click the ngrok URL printed in output to open the trainer.
"""

import cv2
import numpy as np
import tensorflow as tf
import threading
import base64
import time
import json
from collections import deque
from flask import Flask, request, jsonify, render_template_string
from flask_cors import CORS
from mediapipe.python.solutions import pose as mp_pose
from mediapipe.python.solutions import drawing_utils as mp_drawing

# ─────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────
actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

THRESHOLDS = {
    'squat':               {'down': 85,  'up': 155, 'joints': (23, 25, 27)},
    'barbell biceps curl': {'down': 45,  'up': 145, 'joints': (11, 13, 15)},
    'push-up':             {'down': 65,  'up': 145, 'joints': (11, 13, 15)},
    'shoulder press':      {'down': 65,  'up': 145, 'joints': (13, 11, 23)},
}

FORM_RULES = {
    'squat': {
        'good_range': (70, 95),
        'good_msg':  'Great squat depth!',
        'high_msg':  'Go deeper - thighs parallel',
        'low_msg':   'Too deep - stop at parallel',
        'tip':       'Knees over toes, weight in heels',
    },
    'barbell biceps curl': {
        'good_range': (35, 55),
        'good_msg':  'Full range curl!',
        'high_msg':  'Curl higher - squeeze at top',
        'low_msg':   'Lower fully to complete rep',
        'tip':       'Keep elbows pinned to sides',
    },
    'push-up': {
        'good_range': (55, 80),
        'good_msg':  'Good push-up depth!',
        'high_msg':  'Lower chest to ground',
        'low_msg':   'Keep core tight',
        'tip':       'Body must be a straight line',
    },
    'shoulder press': {
        'good_range': (55, 80),
        'good_msg':  'Great press! Full extension.',
        'high_msg':  'Press fully overhead',
        'low_msg':   'Lower to shoulder level',
        'tip':       'Brace core - protect lower back',
    },
}

CALORIES_PER_REP = {
    'squat': 0.32, 'barbell biceps curl': 0.14,
    'push-up': 0.29, 'shoulder press': 0.19,
}

EXERCISE_COLORS_HEX = {
    'squat':               '#00c864',
    'barbell biceps curl': '#ff8800',
    'push-up':             '#0096ff',
    'shoulder press':      '#cc00cc',
    'No pose detected':    '#888888',
}

# ─────────────────────────────────────────
# STATE  (thread-safe via lock)
# ─────────────────────────────────────────
state_lock = threading.Lock()
state = {
    'count': 0,
    'stage': None,
    'last_feedback': '',
    'exercise': 'No pose detected',
    'confidence': 0.0,
    'angle': 0.0,
    'rep_confirmed': False,
    'last_exercise': '',
    'total_calories': 0.0,
    'angle_buffer': deque(maxlen=5),
    'session_start': time.time(),
    'rep_counts': {a: 0 for a in actions},
    'annotated_frame_b64': '',   # latest processed frame as base64 JPEG
}

# ─────────────────────────────────────────
# LOAD MODEL & POSE
# ─────────────────────────────────────────
print("Loading TF model...")
model = tf.keras.models.load_model('fitness_trainer_model.h5')
print("✅ Model loaded")

pose_model = mp_pose.Pose(
    min_detection_confidence=0.6,
    min_tracking_confidence=0.6,
    model_complexity=1
)
print("✅ MediaPipe Pose loaded")

# ─────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = (np.arctan2(c[1]-b[1], c[0]-b[0])
               - np.arctan2(a[1]-b[1], a[0]-b[0]))
    angle = np.abs(radians * 180.0 / np.pi)
    return 360 - angle if angle > 180 else angle

def get_angle(exercise, lm):
    if exercise not in THRESHOLDS:
        return 0.0
    i, j, k = THRESHOLDS[exercise]['joints']
    return calculate_angle(
        [lm[i].x, lm[i].y],
        [lm[j].x, lm[j].y],
        [lm[k].x, lm[k].y]
    )

def get_form_feedback(exercise, angle):
    if exercise not in FORM_RULES:
        return ''
    r = FORM_RULES[exercise]
    lo, hi = r['good_range']
    good = lo <= angle <= hi
    prefix = 'OK' if good else 'FIX'
    msg    = r['good_msg'] if good else (r['high_msg'] if angle > hi else r['low_msg'])
    return f"{prefix}: {msg} | {r['tip']}"

def update_rep_count(exercise, angle):
    thresh      = THRESHOLDS.get(exercise, {})
    down_thresh = thresh.get('down', 90)
    up_thresh   = thresh.get('up', 150)

    if exercise != state['last_exercise']:
        state['count']         = 0
        state['stage']         = None
        state['rep_confirmed'] = False
        state['last_exercise'] = exercise
        state['angle_buffer'].clear()

    if angle <= down_thresh:
        if state['stage'] != 'down':
            state['stage']         = 'down'
            state['rep_confirmed'] = False
            state['last_feedback'] = get_form_feedback(exercise, angle)
    elif angle >= up_thresh:
        if state['stage'] == 'down' and not state['rep_confirmed']:
            state['count']        += 1
            state['rep_counts'][exercise] += 1
            state['stage']         = 'up'
            state['rep_confirmed'] = True
            state['total_calories'] += CALORIES_PER_REP.get(exercise, 0.2)
            state['last_feedback'] = get_form_feedback(exercise, angle)

# ─────────────────────────────────────────
# DRAW OVERLAY ON FRAME
# ─────────────────────────────────────────
def draw_overlay(frame, exercise, confidence, angle):
    h, w = frame.shape[:2]

    def hex2bgr(hx):
        hx = hx.lstrip('#')
        r, g, b = int(hx[0:2],16), int(hx[2:4],16), int(hx[4:6],16)
        return (b, g, r)

    ex_color = hex2bgr(EXERCISE_COLORS_HEX.get(exercise, '#888888'))

    # Top bar
    cv2.rectangle(frame, (0, 0), (w, 52), (15, 15, 15), -1)
    cv2.putText(frame, f"{exercise.upper()}   {confidence:.0%}",
                (12, 36), cv2.FONT_HERSHEY_DUPLEX, 1.0, (255, 255, 255), 2)

    # REPS badge (top-right)
    cv2.rectangle(frame, (w-150, 0), (w, 90), (20, 60, 180), -1)
    cv2.putText(frame, "REPS", (w-128, 25),
                cv2.FONT_HERSHEY_SIMPLEX, 0.65, (180, 180, 255), 1)
    cv2.putText(frame, str(state['count']), (w-115, 80),
                cv2.FONT_HERSHEY_DUPLEX, 2.2, (255, 255, 255), 3)

    # Left side: per-exercise counts
    cv2.rectangle(frame, (0, 52), (175, h), (10, 10, 10), -1)
    y = 105
    short_names = {
        'barbell biceps curl': 'Curl',
        'push-up':             'Push-up',
        'shoulder press':      'Press',
        'squat':               'Squat',
    }
    for act in actions:
        is_cur = (act == exercise)
        col    = ex_color if is_cur else (160, 160, 160)
        wt     = 2 if is_cur else 1
        label  = f"{short_names[act]}: {state['rep_counts'][act]}"
        cv2.putText(frame, label, (8, y),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.72, col, wt)
        if is_cur:
            cv2.rectangle(frame, (2, y-22), (170, y+8), ex_color, 1)
        y += 68

    # Bottom bar
    cv2.rectangle(frame, (175, h-88), (w, h), (15, 15, 15), -1)
    stage_str   = (state['stage'] or 'ready').upper()
    stage_color = (0,210,80) if state['stage']=='up' else (0,120,255) if state['stage']=='down' else (220,220,0)
    cv2.putText(frame, f"Angle: {angle:.0f}deg  Stage: {stage_str}",
                (180, h-56), cv2.FONT_HERSHEY_SIMPLEX, 0.58, stage_color, 1)

    feedback = state['last_feedback']
    if feedback:
        fb_color = (0, 210, 80) if feedback.startswith('OK') else (0, 150, 255)
        cv2.putText(frame, feedback[:78], (180, h-18),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.50, fb_color, 1)

    # Calories + timer
    elapsed   = int(time.time() - state['session_start'])
    m, s      = divmod(elapsed, 60)
    cv2.putText(frame, f"Cal: {state['total_calories']:.1f}  {m:02d}:{s:02d}",
                (w-220, h-8), cv2.FONT_HERSHEY_SIMPLEX, 0.52, (150,200,255), 1)

    # Confidence bar
    bar_start = 175
    bar_end   = w - 150
    conf_w    = int((bar_end - bar_start) * confidence)
    cv2.rectangle(frame, (bar_start, h-4), (bar_start+conf_w, h), ex_color, -1)

    return frame

# ─────────────────────────────────────────
# PROCESS A SINGLE FRAME (called per POST)
# ─────────────────────────────────────────
def process_frame(frame_bgr):
    rgb     = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    results = pose_model.process(rgb)

    exercise, confidence, angle = 'No pose detected', 0.0, 0.0

    if results.pose_landmarks:
        lm     = results.pose_landmarks.landmark
        coords = np.array([[l.x, l.y, l.z] for l in lm]).flatten()

        pred       = model.predict(coords.reshape(1, 1, 99), verbose=0)
        exercise   = actions[np.argmax(pred)]
        confidence = float(np.max(pred))

        raw_angle = get_angle(exercise, lm)
        state['angle_buffer'].append(raw_angle)
        angle = float(np.mean(state['angle_buffer']))

        update_rep_count(exercise, angle)

        state['exercise']   = exercise
        state['confidence'] = confidence
        state['angle']      = angle

        # Draw skeleton
        mp_drawing.draw_landmarks(
            frame_bgr,
            results.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing.DrawingSpec(
                color=(0, 0, 220), thickness=2, circle_radius=4),
            connection_drawing_spec=mp_drawing.DrawingSpec(
                color=(255, 255, 255), thickness=2)
        )
    else:
        state['exercise']   = 'No pose detected'
        state['confidence'] = 0.0

    frame_bgr = draw_overlay(frame_bgr, exercise, confidence, angle)

    # Encode to JPEG base64 for sending to browser
    ret, buf = cv2.imencode('.jpg', frame_bgr,
                            [cv2.IMWRITE_JPEG_QUALITY, 78])
    if ret:
        state['annotated_frame_b64'] = base64.b64encode(buf).decode('utf-8')

# ─────────────────────────────────────────
# FLASK APP
# ─────────────────────────────────────────
app = Flask(__name__)
CORS(app)

HTML_PAGE = r"""
<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8">
<title>AI Fitness Trainer – Live</title>
<style>
  @import url('https://fonts.googleapis.com/css2?family=Rajdhani:wght@500;700&family=Share+Tech+Mono&display=swap');

  :root {
    --bg:      #080c10;
    --panel:   #0d1117;
    --border:  #1e2a38;
    --accent:  #00e5ff;
    --green:   #00e676;
    --orange:  #ff9100;
    --red:     #ff1744;
    --text:    #cfd8dc;
    --dim:     #546e7a;
  }

  * { margin:0; padding:0; box-sizing:border-box; }

  body {
    background: var(--bg);
    color: var(--text);
    font-family: 'Rajdhani', sans-serif;
    height: 100vh;
    overflow: hidden;
    display: flex;
    flex-direction: column;
  }

  /* ── HEADER ── */
  header {
    background: var(--panel);
    border-bottom: 1px solid var(--border);
    padding: 10px 20px;
    display: flex;
    align-items: center;
    justify-content: space-between;
    flex-shrink: 0;
  }
  header h1 {
    font-size: 20px;
    font-weight: 700;
    letter-spacing: 2px;
    color: var(--accent);
    text-transform: uppercase;
  }
  .status-dot {
    width: 10px; height: 10px;
    border-radius: 50%;
    background: var(--red);
    display: inline-block;
    margin-right: 6px;
    transition: background .3s;
  }
  .status-dot.live { background: var(--green); animation: pulse 1.2s infinite; }
  @keyframes pulse {
    0%,100% { box-shadow: 0 0 0 0 rgba(0,230,118,.5); }
    50%      { box-shadow: 0 0 0 6px rgba(0,230,118,0); }
  }
  #status-text { font-size: 13px; color: var(--dim); font-family: 'Share Tech Mono', monospace; }

  /* ── MAIN ── */
  .main {
    display: flex;
    flex: 1;
    overflow: hidden;
  }

  /* ── VIDEO CANVAS ── */
  .video-area {
    flex: 1;
    display: flex;
    flex-direction: column;
    align-items: center;
    justify-content: center;
    background: #000;
    position: relative;
    overflow: hidden;
  }

  /* hidden raw webcam */
  #rawVideo {
    position: absolute;
    width: 1px; height: 1px;
    opacity: 0;
    pointer-events: none;
  }
  #captureCanvas { display: none; }

  /* displayed annotated image */
  #annotated {
    max-width: 100%;
    max-height: 100%;
    object-fit: contain;
    border: 1px solid var(--border);
  }

  .cam-hint {
    position: absolute;
    bottom: 12px;
    left: 50%;
    transform: translateX(-50%);
    font-size: 12px;
    color: var(--dim);
    font-family: 'Share Tech Mono', monospace;
    background: rgba(0,0,0,.6);
    padding: 4px 12px;
    border-radius: 20px;
    pointer-events: none;
  }

  /* ── SIDEBAR ── */
  .sidebar {
    width: 260px;
    background: var(--panel);
    border-left: 1px solid var(--border);
    padding: 16px 14px;
    overflow-y: auto;
    display: flex;
    flex-direction: column;
    gap: 12px;
    flex-shrink: 0;
  }

  .card {
    background: #111820;
    border: 1px solid var(--border);
    border-radius: 8px;
    padding: 12px;
    text-align: center;
  }
  .card .val {
    font-size: 38px;
    font-weight: 700;
    line-height: 1;
    font-family: 'Share Tech Mono', monospace;
  }
  .card .lbl {
    font-size: 10px;
    letter-spacing: 2px;
    color: var(--dim);
    margin-top: 4px;
    text-transform: uppercase;
  }

  .cyan   { color: var(--accent); }
  .green  { color: var(--green); }
  .orange { color: var(--orange); }
  .white  { color: #fff; }

  .info-row {
    background: #111820;
    border: 1px solid var(--border);
    border-radius: 6px;
    padding: 8px 12px;
    font-size: 13px;
    display: flex;
    justify-content: space-between;
  }
  .info-row .key { color: var(--dim); }
  .info-row .val2 { color: var(--accent); font-family:'Share Tech Mono',monospace; font-size:12px; }

  /* Stage badge */
  #stageBadge {
    text-align: center;
    padding: 8px;
    border-radius: 6px;
    font-weight: 700;
    font-size: 16px;
    letter-spacing: 3px;
    text-transform: uppercase;
    border: 1px solid;
    transition: all .2s;
  }

  /* Feedback */
  #feedbackBox {
    background: #111820;
    border-left: 3px solid var(--accent);
    border-radius: 4px;
    padding: 10px;
    font-size: 12px;
    line-height: 1.6;
    color: var(--text);
    min-height: 52px;
    font-family: 'Share Tech Mono', monospace;
  }

  /* Per-exercise reps table */
  .rep-table { width: 100%; border-collapse: collapse; font-size: 13px; }
  .rep-table td { padding: 5px 6px; border-bottom: 1px solid var(--border); }
  .rep-table td:last-child { text-align: right; font-family:'Share Tech Mono',monospace; color: var(--accent); }
  .rep-table tr.active td { color: #fff; font-weight: 700; }

  /* Angle gauge bar */
  .gauge-wrap {
    background: #111820;
    border: 1px solid var(--border);
    border-radius: 6px;
    padding: 10px 12px;
  }
  .gauge-label { font-size: 11px; color: var(--dim); letter-spacing: 1px; margin-bottom: 6px; text-transform: uppercase; }
  .gauge-track {
    height: 8px;
    background: #1e2a38;
    border-radius: 4px;
    overflow: hidden;
  }
  .gauge-fill {
    height: 100%;
    background: var(--accent);
    border-radius: 4px;
    transition: width .15s, background .2s;
  }
  .gauge-val { font-size: 18px; font-weight:700; font-family:'Share Tech Mono',monospace; margin-top:4px; color:var(--text); }

  /* Reset button */
  .reset-btn {
    width: 100%;
    padding: 11px;
    background: transparent;
    border: 1px solid var(--red);
    border-radius: 6px;
    color: var(--red);
    font-size: 14px;
    font-weight: 700;
    letter-spacing: 1px;
    cursor: pointer;
    text-transform: uppercase;
    transition: background .2s, color .2s;
    font-family: 'Rajdhani', sans-serif;
  }
  .reset-btn:hover { background: var(--red); color: #fff; }

  /* scrollbar */
  .sidebar::-webkit-scrollbar { width: 4px; }
  .sidebar::-webkit-scrollbar-track { background: transparent; }
  .sidebar::-webkit-scrollbar-thumb { background: var(--border); border-radius: 2px; }
</style>
</head>
<body>

<header>
  <h1>🏋️ AI Fitness Trainer &mdash; Live</h1>
  <div>
    <span class="status-dot" id="dot"></span>
    <span id="status-text">Starting camera…</span>
  </div>
</header>

<div class="main">

  <!-- VIDEO AREA -->
  <div class="video-area">
    <video id="rawVideo" autoplay playsinline muted></video>
    <canvas id="captureCanvas"></canvas>
    <img id="annotated" src="" alt="Waiting for first frame…">
    <div class="cam-hint">Stand 2–3 m back · Full body in frame</div>
  </div>

  <!-- SIDEBAR -->
  <div class="sidebar">

    <div style="display:flex; gap:8px;">
      <div class="card" style="flex:1">
        <div class="val white" id="reps">0</div>
        <div class="lbl">Reps</div>
      </div>
      <div class="card" style="flex:1">
        <div class="val orange" id="cal">0.0</div>
        <div class="lbl">Calories</div>
      </div>
    </div>

    <div class="card">
      <div class="val cyan" id="timer">00:00</div>
      <div class="lbl">Session Time</div>
    </div>

    <div class="info-row">
      <span class="key">Exercise</span>
      <span class="val2" id="exercise">Detecting…</span>
    </div>
    <div class="info-row">
      <span class="key">Confidence</span>
      <span class="val2" id="conf">—</span>
    </div>

    <div id="stageBadge" style="color:#546e7a; border-color:#546e7a;">READY</div>

    <div class="gauge-wrap">
      <div class="gauge-label">Joint Angle</div>
      <div class="gauge-track"><div class="gauge-fill" id="angleFill" style="width:0%"></div></div>
      <div class="gauge-val" id="angleVal">—</div>
    </div>

    <div id="feedbackBox">Waiting for pose…</div>

    <!-- Per-exercise rep counts -->
    <table class="rep-table" id="repTable">
      <tr data-ex="squat">
        <td>Squat</td><td id="r-squat">0</td>
      </tr>
      <tr data-ex="barbell biceps curl">
        <td>Curl</td><td id="r-barbell biceps curl">0</td>
      </tr>
      <tr data-ex="push-up">
        <td>Push-up</td><td id="r-push-up">0</td>
      </tr>
      <tr data-ex="shoulder press">
        <td>Press</td><td id="r-shoulder press">0</td>
      </tr>
    </table>

    <button class="reset-btn" onclick="resetSession()">↺ Reset Session</button>
  </div>
</div>

<script>
// ── CONFIG ──────────────────────────────────────
const SEND_INTERVAL_MS = 120;   // send a frame every 120ms (~8fps)
const SERVER           = '';    // same origin

// ── DOM ─────────────────────────────────────────
const rawVideo     = document.getElementById('rawVideo');
const capCanvas    = document.getElementById('captureCanvas');
const ctx          = capCanvas.getContext('2d');
const annotatedImg = document.getElementById('annotated');
const dot          = document.getElementById('dot');
const statusText   = document.getElementById('status-text');

// ── STATE ────────────────────────────────────────
let sessionStart = Date.now();
let sending      = false;

// ── START WEBCAM ─────────────────────────────────
async function startCamera() {
  try {
    const stream = await navigator.mediaDevices.getUserMedia({
      video: { width: 1280, height: 720, facingMode: 'user' },
      audio: false
    });
    rawVideo.srcObject = stream;
    rawVideo.onloadedmetadata = () => {
      capCanvas.width  = rawVideo.videoWidth;
      capCanvas.height = rawVideo.videoHeight;
      dot.classList.add('live');
      statusText.textContent = 'Live · auto-analysing';
      startSendLoop();
    };
  } catch(e) {
    statusText.textContent = 'Camera error: ' + e.message;
    console.error(e);
  }
}

// ── SEND LOOP ────────────────────────────────────
function startSendLoop() {
  setInterval(sendFrame, SEND_INTERVAL_MS);
}

async function sendFrame() {
  if (sending) return;          // skip if previous request still pending
  if (rawVideo.readyState < 2) return;

  sending = true;
  try {
    // draw current video frame onto hidden canvas
    ctx.drawImage(rawVideo, 0, 0, capCanvas.width, capCanvas.height);
    const blob = await new Promise(res => capCanvas.toBlob(res, 'image/jpeg', 0.82));
    const arrayBuf = await blob.arrayBuffer();

    const resp = await fetch(SERVER + '/process_frame', {
      method: 'POST',
      headers: { 'Content-Type': 'image/jpeg' },
      body: arrayBuf
    });

    if (!resp.ok) throw new Error('HTTP ' + resp.status);

    const data = await resp.json();

    // update annotated image
    if (data.frame_b64) {
      annotatedImg.src = 'data:image/jpeg;base64,' + data.frame_b64;
    }

    // update sidebar
    updateSidebar(data);

  } catch(e) {
    console.warn('send error', e);
  } finally {
    sending = false;
  }
}

// ── UPDATE UI ────────────────────────────────────
const STAGE_STYLES = {
  up:    { color:'#00e676', border:'#00e676', text:'UP   ↑' },
  down:  { color:'#0096ff', border:'#0096ff', text:'DOWN ↓' },
  null:  { color:'#546e7a', border:'#546e7a', text:'READY' },
  ready: { color:'#546e7a', border:'#546e7a', text:'READY' },
};

function updateSidebar(d) {
  document.getElementById('reps').textContent    = d.count;
  document.getElementById('cal').textContent     = d.calories;
  document.getElementById('exercise').textContent= d.exercise;
  document.getElementById('conf').textContent    = d.confidence;

  // stage badge
  const stage    = d.stage || 'ready';
  const ss       = STAGE_STYLES[stage] || STAGE_STYLES['ready'];
  const badge    = document.getElementById('stageBadge');
  badge.textContent    = ss.text;
  badge.style.color    = ss.color;
  badge.style.borderColor = ss.border;

  // angle gauge  (assume 0-180 range)
  const angleRaw = parseFloat(d.angle) || 0;
  const pct      = Math.min(angleRaw / 180 * 100, 100);
  document.getElementById('angleFill').style.width = pct + '%';
  document.getElementById('angleVal').textContent  = d.angle + '°';

  // feedback
  const fb     = d.feedback || '';
  const fbBox  = document.getElementById('feedbackBox');
  fbBox.textContent        = fb || (d.exercise === 'No pose detected' ? 'No pose – stand back so full body is visible' : 'Detecting…');
  fbBox.style.borderLeftColor = fb.startsWith('OK') ? '#00e676' : fb.startsWith('FIX') ? '#ff9100' : '#00e5ff';

  // rep counts per exercise
  for (const [ex, cnt] of Object.entries(d.rep_counts || {})) {
    const el = document.getElementById('r-' + ex);
    if (el) el.textContent = cnt;
  }
  // highlight active exercise row
  document.querySelectorAll('#repTable tr').forEach(tr => {
    tr.classList.toggle('active', tr.dataset.ex === d.exercise);
  });

  // timer
  const elapsed = Math.floor((Date.now() - sessionStart) / 1000);
  const m = Math.floor(elapsed/60), s = elapsed % 60;
  document.getElementById('timer').textContent =
    String(m).padStart(2,'0') + ':' + String(s).padStart(2,'0');
}

async function resetSession() {
  sessionStart = Date.now();
  await fetch(SERVER + '/reset');
}

// ── BOOT ─────────────────────────────────────────
startCamera();
</script>
</body>
</html>
"""

# ─────────────────────────────────────────
# ROUTES
# ─────────────────────────────────────────
@app.route('/')
def index():
    return render_template_string(HTML_PAGE)


@app.route('/process_frame', methods=['POST'])
def process_frame_route():
    """
    Receives a raw JPEG body from the browser,
    runs pose + model inference, returns JSON with
    annotated frame (base64) + state.
    """
    raw = request.get_data()
    nparr = np.frombuffer(raw, np.uint8)
    frame = cv2.imdecode(nparr, cv2.IMREAD_COLOR)

    if frame is None:
        return jsonify({'error': 'bad frame'}), 400

    with state_lock:
        process_frame(frame)
        resp = _get_state_dict()

    return jsonify(resp)


@app.route('/reset')
def reset_route():
    with state_lock:
        state.update({
            'count': 0, 'stage': None, 'last_feedback': '',
            'rep_confirmed': False, 'last_exercise': '',
            'total_calories': 0.0,
            'session_start': time.time(),
            'rep_counts': {a: 0 for a in actions},
            'annotated_frame_b64': '',
        })
        state['angle_buffer'].clear()
    return 'OK'


def _get_state_dict():
    return {
        'count':      state['count'],
        'calories':   f"{state['total_calories']:.1f}",
        'exercise':   state['exercise'],
        'confidence': f"{state['confidence']:.0%}",
        'stage':      state['stage'],
        'angle':      f"{state['angle']:.1f}",
        'feedback':   state['last_feedback'],
        'rep_counts': dict(state['rep_counts']),
        'frame_b64':  state['annotated_frame_b64'],
    }


# ─────────────────────────────────────────
# ENTRY POINT — starts ngrok tunnel in Colab
# ─────────────────────────────────────────
if __name__ == '__main__':
    try:
        from pyngrok import ngrok

        # ── If you have an ngrok auth token, set it here:
        # ngrok.set_auth_token("YOUR_TOKEN_HERE")

        public_url = ngrok.connect(5000)
        print("=" * 60)
        print(f"  🚀  Open this URL in your browser:")
        print(f"      {public_url}")
        print("=" * 60)
        print("  Controls:")
        print("    • Allow camera access when the browser asks")
        print("    • Stand 2–3 m from camera, full body in frame")
        print("    • Exercises auto-detected — no button to press!")
        print("    • Click ↺ Reset Session to start fresh")
        print("=" * 60)
    except ImportError:
        print("pyngrok not installed — run:  !pip install pyngrok -q")
        print("Starting on http://localhost:5000 (local only)")

    app.run(host='0.0.0.0', port=5000, threaded=True, use_reloader=False)
```

```python
"""
Real-time AI Fitness Trainer — Google Colab Edition (FULL)
==========================================================
Includes:
  1. 📹 Live webcam trainer  — auto rep counting, skeleton overlay, form feedback
  2. 🎬 Video Analysis tab   — upload a video → frame-by-frame rep count + form summary
  3. 🤖 AI Fitness Chatbot   — ask any fitness question (Gemini or built-in)

HOW TO USE IN COLAB:
─────────────────────
Step 1 — Install deps:
    !pip install flask flask-cors mediapipe tensorflow pyngrok google-generativeai -q

Step 2 — (Optional) Set your free Gemini API key at https://aistudio.google.com/
    Edit GEMINI_API_KEY = "your-key-here"  below.
    Leave blank to use the built-in rule-based fitness chatbot.

Step 3 — Make sure fitness_trainer_model.h5 is in /content/

Step 4 — Run:
    !python realtime_trainer_colab.py

Step 5 — Click the ngrok URL printed in the output.
"""

import cv2
import numpy as np
import tensorflow as tf
import threading
import base64
import time
import json
import os
import tempfile
from collections import deque
from flask import Flask, request, jsonify, render_template_string
from flask_cors import CORS
from mediapipe.python.solutions import pose as mp_pose
from mediapipe.python.solutions import drawing_utils as mp_drawing

# ─────────────────────────────────────────
# ⚙️  OPTIONAL: Set your Gemini API key
# ─────────────────────────────────────────
GEMINI_API_KEY = ""   # e.g. "AIzaSy..."

# ─────────────────────────────────────────
# CONFIG
# ─────────────────────────────────────────
actions = ['barbell biceps curl', 'push-up', 'shoulder press', 'squat']

THRESHOLDS = {
    'squat':               {'down': 85,  'up': 155, 'joints': (23, 25, 27)},
    'barbell biceps curl': {'down': 45,  'up': 145, 'joints': (11, 13, 15)},
    'push-up':             {'down': 65,  'up': 145, 'joints': (11, 13, 15)},
    'shoulder press':      {'down': 65,  'up': 145, 'joints': (13, 11, 23)},
}

FORM_RULES = {
    'squat': {
        'good_range': (70, 95),
        'good_msg':  'Great squat depth!',
        'high_msg':  'Go deeper - thighs parallel',
        'low_msg':   'Too deep - stop at parallel',
        'tip':       'Knees over toes, weight in heels',
    },
    'barbell biceps curl': {
        'good_range': (35, 55),
        'good_msg':  'Full range curl!',
        'high_msg':  'Curl higher - squeeze at top',
        'low_msg':   'Lower fully to complete rep',
        'tip':       'Keep elbows pinned to sides',
    },
    'push-up': {
        'good_range': (55, 80),
        'good_msg':  'Good push-up depth!',
        'high_msg':  'Lower chest to ground',
        'low_msg':   'Keep core tight',
        'tip':       'Body must be a straight line',
    },
    'shoulder press': {
        'good_range': (55, 80),
        'good_msg':  'Great press! Full extension.',
        'high_msg':  'Press fully overhead',
        'low_msg':   'Lower to shoulder level',
        'tip':       'Brace core - protect lower back',
    },
}

CALORIES_PER_REP = {
    'squat': 0.32, 'barbell biceps curl': 0.14,
    'push-up': 0.29, 'shoulder press': 0.19,
}

EXERCISE_COLORS_HEX = {
    'squat':               '#00c864',
    'barbell biceps curl': '#ff8800',
    'push-up':             '#0096ff',
    'shoulder press':      '#cc00cc',
    'No pose detected':    '#888888',
}

# ─────────────────────────────────────────
# LIVE STATE  (thread-safe)
# ─────────────────────────────────────────
state_lock = threading.Lock()
state = {
    'count': 0, 'stage': None,
    'last_feedback': '', 'exercise': 'No pose detected',
    'confidence': 0.0, 'angle': 0.0,
    'rep_confirmed': False, 'last_exercise': '',
    'total_calories': 0.0,
    'angle_buffer': deque(maxlen=5),
    'session_start': time.time(),
    'rep_counts': {a: 0 for a in actions},
    'annotated_frame_b64': '',
}

# ─────────────────────────────────────────
# LOAD MODEL & POSE
# ─────────────────────────────────────────
print("Loading TF model...")
model = tf.keras.models.load_model('fitness_trainer_model.h5')
print("✅ Model loaded")

pose_model = mp_pose.Pose(
    min_detection_confidence=0.6,
    min_tracking_confidence=0.6,
    model_complexity=1
)
print("✅ MediaPipe Pose loaded")

# ─────────────────────────────────────────
# GEMINI CHATBOT
# ─────────────────────────────────────────
gemini_chat = None
if GEMINI_API_KEY:
    try:
        import google.generativeai as genai
        genai.configure(api_key=GEMINI_API_KEY)
        _gm = genai.GenerativeModel(
            model_name='gemini-1.5-flash',
            system_instruction=(
                "You are FitBot, an expert AI fitness coach specialising in "
                "squat, push-up, barbell biceps curl, and shoulder press. "
                "Be concise, friendly and practical. Use bullet points for tips. "
                "Always prioritise safety. For injuries, recommend seeing a professional."
            )
        )
        gemini_chat = _gm.start_chat(history=[])
        print("✅ Gemini chatbot ready")
    except Exception as e:
        print(f"⚠️  Gemini failed: {e} — using rule-based chatbot")
else:
    print("ℹ️  No Gemini key — using built-in rule-based chatbot")

# ─────────────────────────────────────────
# HELPERS
# ─────────────────────────────────────────
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    rad = np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0])
    ang = np.abs(rad * 180.0 / np.pi)
    return 360 - ang if ang > 180 else ang

def get_angle(exercise, lm):
    if exercise not in THRESHOLDS:
        return 0.0
    i, j, k = THRESHOLDS[exercise]['joints']
    return calculate_angle([lm[i].x, lm[i].y], [lm[j].x, lm[j].y], [lm[k].x, lm[k].y])

def get_form_feedback(exercise, angle):
    if exercise not in FORM_RULES:
        return ''
    r = FORM_RULES[exercise]
    lo, hi = r['good_range']
    good = lo <= angle <= hi
    prefix = 'OK' if good else 'FIX'
    msg = r['good_msg'] if good else (r['high_msg'] if angle > hi else r['low_msg'])
    return f"{prefix}: {msg} | {r['tip']}"

def update_rep_count(exercise, angle):
    thresh = THRESHOLDS.get(exercise, {})
    dn, up = thresh.get('down', 90), thresh.get('up', 150)
    if exercise != state['last_exercise']:
        state['count'] = 0
        state['stage'] = None
        state['rep_confirmed'] = False
        state['last_exercise'] = exercise
        state['angle_buffer'].clear()
    if angle <= dn:
        if state['stage'] != 'down':
            state['stage'] = 'down'
            state['rep_confirmed'] = False
            state['last_feedback'] = get_form_feedback(exercise, angle)
    elif angle >= up:
        if state['stage'] == 'down' and not state['rep_confirmed']:
            state['count'] += 1
            state['rep_counts'][exercise] += 1
            state['stage'] = 'up'
            state['rep_confirmed'] = True
            state['total_calories'] += CALORIES_PER_REP.get(exercise, 0.2)
            state['last_feedback'] = get_form_feedback(exercise, angle)

def hex2bgr(hx):
    hx = hx.lstrip('#')
    r, g, b = int(hx[0:2],16), int(hx[2:4],16), int(hx[4:6],16)
    return (b, g, r)

def draw_overlay(frame, exercise, confidence, angle):
    h, w = frame.shape[:2]
    ec = hex2bgr(EXERCISE_COLORS_HEX.get(exercise, '#888888'))
    cv2.rectangle(frame, (0,0), (w,52), (15,15,15), -1)
    cv2.putText(frame, f"{exercise.upper()}   {confidence:.0%}",
                (12,36), cv2.FONT_HERSHEY_DUPLEX, 1.0, (255,255,255), 2)
    cv2.rectangle(frame, (w-150,0), (w,90), (20,60,180), -1)
    cv2.putText(frame, "REPS", (w-128,25), cv2.FONT_HERSHEY_SIMPLEX, 0.65, (180,180,255), 1)
    cv2.putText(frame, str(state['count']), (w-115,80),
                cv2.FONT_HERSHEY_DUPLEX, 2.2, (255,255,255), 3)
    cv2.rectangle(frame, (0,52), (175,h), (10,10,10), -1)
    y = 105
    short = {'barbell biceps curl':'Curl','push-up':'Push-up','shoulder press':'Press','squat':'Squat'}
    for act in actions:
        is_cur = (act == exercise)
        col = ec if is_cur else (160,160,160)
        wt  = 2 if is_cur else 1
        cv2.putText(frame, f"{short[act]}: {state['rep_counts'][act]}",
                    (8,y), cv2.FONT_HERSHEY_SIMPLEX, 0.72, col, wt)
        if is_cur:
            cv2.rectangle(frame, (2,y-22), (170,y+8), ec, 1)
        y += 68
    cv2.rectangle(frame, (175,h-88), (w,h), (15,15,15), -1)
    ss = (state['stage'] or 'ready').upper()
    sc = (0,210,80) if state['stage']=='up' else (0,120,255) if state['stage']=='down' else (220,220,0)
    cv2.putText(frame, f"Angle: {angle:.0f}deg  Stage: {ss}",
                (180,h-56), cv2.FONT_HERSHEY_SIMPLEX, 0.58, sc, 1)
    fb = state['last_feedback']
    if fb:
        fc = (0,210,80) if fb.startswith('OK') else (0,150,255)
        cv2.putText(frame, fb[:78], (180,h-18), cv2.FONT_HERSHEY_SIMPLEX, 0.50, fc, 1)
    elapsed = int(time.time() - state['session_start'])
    m, s = divmod(elapsed, 60)
    cv2.putText(frame, f"Cal: {state['total_calories']:.1f}  {m:02d}:{s:02d}",
                (w-220,h-8), cv2.FONT_HERSHEY_SIMPLEX, 0.52, (150,200,255), 1)
    bs, be = 175, w-150
    cw = int((be-bs)*confidence)
    cv2.rectangle(frame, (bs,h-4), (bs+cw,h), ec, -1)
    return frame

def process_frame_bgr(frame_bgr):
    rgb = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2RGB)
    results = pose_model.process(rgb)
    exercise, confidence, angle = 'No pose detected', 0.0, 0.0
    if results.pose_landmarks:
        lm     = results.pose_landmarks.landmark
        coords = np.array([[l.x,l.y,l.z] for l in lm]).flatten()
        pred   = model.predict(coords.reshape(1,1,99), verbose=0)
        exercise   = actions[np.argmax(pred)]
        confidence = float(np.max(pred))
        raw_angle  = get_angle(exercise, lm)
        state['angle_buffer'].append(raw_angle)
        angle = float(np.mean(state['angle_buffer']))
        update_rep_count(exercise, angle)
        state['exercise']   = exercise
        state['confidence'] = confidence
        state['angle']      = angle
        mp_drawing.draw_landmarks(
            frame_bgr, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing.DrawingSpec(color=(0,0,220),thickness=2,circle_radius=4),
            connection_drawing_spec=mp_drawing.DrawingSpec(color=(255,255,255),thickness=2))
    else:
        state['exercise']   = 'No pose detected'
        state['confidence'] = 0.0
    frame_bgr = draw_overlay(frame_bgr, exercise, confidence, angle)
    ret, buf = cv2.imencode('.jpg', frame_bgr, [cv2.IMWRITE_JPEG_QUALITY, 78])
    if ret:
        state['annotated_frame_b64'] = base64.b64encode(buf).decode('utf-8')

def rule_based_chat(msg):
    import random
    ml = msg.lower()
    tips = {
        'squat':['Keep chest up and back straight.','Drive knees out over toes.',
                 'Weight in heels, not toes.','Depth: thighs parallel to ground.',
                 'Brace core before descending.'],
        'push-up':['Keep body in a straight line head to heels.','Elbows ~45° from body.',
                   'Lower chest to just above ground.','Do not let hips sag.',
                   'Breathe in down, out up.'],
        'curl':['Keep elbows pinned to sides.','Curl all the way — squeeze at top.',
                'Lower slowly (3s) for more activation.','Do not swing your back.',
                'Control the weight through full range.'],
        'shoulder press':['Brace core to protect lower back.','Press straight up.',
                          'Lower to shoulder level each rep.','Keep wrists over elbows.',
                          'Avoid excessive lower back arch.'],
    }
    for ex, ts in tips.items():
        if ex in ml or (ex=='curl' and 'bicep' in ml):
            return f"💪 {ex.title()} Tip:\n{random.choice(ts)}\n\nAsk me about any other exercise!"
    if any(w in ml for w in ['calorie','burn','weight loss']):
        return ("🔥 Calories per rep (approx):\n• Squat: 0.32 kcal\n• Push-up: 0.29 kcal\n"
                "• Shoulder Press: 0.19 kcal\n• Bicep Curl: 0.14 kcal\n\n"
                "Compound lifts like squats burn most per rep!")
    if any(w in ml for w in ['warm','warmup']):
        return ("🌡️ Warm-up Routine (5 min):\n1. Jumping jacks — 30s\n2. Arm circles — 20 each\n"
                "3. Hip circles — 10 each\n4. Bodyweight squats — 10\n5. Push-up negatives — 5")
    if any(w in ml for w in ['rest','recovery','sleep']):
        return ("😴 Recovery Tips:\n• Rest 48h between same muscle groups\n"
                "• 7–9 hours of sleep\n• Hydrate well\n• Light movement on rest days")
    if any(w in ml for w in ['beginner','start','new']):
        return ("🚀 Beginner Routine (3x/week):\n• Squats 3×10\n• Push-ups 3×8\n"
                "• Bicep Curls 3×10\n• Shoulder Press 3×10\n\nRest 60–90s. Add 1 rep/week.")
    return ("🤖 I'm FitBot! Ask me about:\n• Exercise form — 'squat tips'\n"
            "• Calories — 'how many calories burn'\n• Warm-up — 'give me a warm-up'\n"
            "• Recovery — 'rest day tips'\n• Getting started — 'beginner plan'\n\n"
            "(Add a Gemini API key in the script for full AI chat!)")

# ─────────────────────────────────────────
# HTML — 3 SIDEBAR TABS
# ─────────────────────────────────────────
HTML_PAGE = r"""
<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8">
<title>AI Fitness Trainer</title>
<style>
  @import url('https://fonts.googleapis.com/css2?family=Rajdhani:wght@500;700&family=Share+Tech+Mono&display=swap');
  :root {
    --bg:#080c10;--panel:#0d1117;--border:#1e2a38;
    --accent:#00e5ff;--green:#00e676;--orange:#ff9100;
    --red:#ff1744;--text:#cfd8dc;--dim:#546e7a;
  }
  *{margin:0;padding:0;box-sizing:border-box;}
  body{background:var(--bg);color:var(--text);font-family:'Rajdhani',sans-serif;
       height:100vh;overflow:hidden;display:flex;flex-direction:column;}

  header{background:var(--panel);border-bottom:1px solid var(--border);
         padding:10px 20px;display:flex;align-items:center;
         justify-content:space-between;flex-shrink:0;}
  header h1{font-size:20px;font-weight:700;letter-spacing:2px;
            color:var(--accent);text-transform:uppercase;}
  .sdot{width:10px;height:10px;border-radius:50%;background:var(--red);
        display:inline-block;margin-right:6px;transition:background .3s;}
  .sdot.live{background:var(--green);animation:pulse 1.2s infinite;}
  @keyframes pulse{0%,100%{box-shadow:0 0 0 0 rgba(0,230,118,.5)}50%{box-shadow:0 0 0 6px rgba(0,230,118,0)}}
  #stext{font-size:13px;color:var(--dim);font-family:'Share Tech Mono',monospace;}

  .main{display:flex;flex:1;overflow:hidden;}

  .video-area{flex:1;display:flex;flex-direction:column;align-items:center;
              justify-content:center;background:#000;position:relative;overflow:hidden;}
  #rawVideo{position:absolute;width:1px;height:1px;opacity:0;pointer-events:none;}
  #captureCanvas{display:none;}
  #annotated{max-width:100%;max-height:100%;object-fit:contain;border:1px solid var(--border);}
  .cam-hint{position:absolute;bottom:12px;left:50%;transform:translateX(-50%);
            font-size:12px;color:var(--dim);font-family:'Share Tech Mono',monospace;
            background:rgba(0,0,0,.6);padding:4px 12px;border-radius:20px;pointer-events:none;}

  .sidebar{width:285px;background:var(--panel);border-left:1px solid var(--border);
           display:flex;flex-direction:column;flex-shrink:0;overflow:hidden;}

  /* TAB BAR */
  .tab-bar{display:flex;border-bottom:1px solid var(--border);flex-shrink:0;}
  .tab-btn{flex:1;padding:11px 4px;text-align:center;font-size:11px;font-weight:700;
           letter-spacing:1px;text-transform:uppercase;cursor:pointer;
           border:none;background:transparent;color:var(--dim);
           transition:color .2s;border-bottom:2px solid transparent;}
  .tab-btn.active{color:var(--accent);border-bottom:2px solid var(--accent);}
  .tab-btn:hover:not(.active){color:var(--text);}

  /* PANELS */
  .tab-panel{display:none;flex:1;overflow-y:auto;padding:13px;
             flex-direction:column;gap:11px;}
  .tab-panel.active{display:flex;}
  .tab-panel::-webkit-scrollbar{width:4px;}
  .tab-panel::-webkit-scrollbar-thumb{background:var(--border);border-radius:2px;}

  /* ── LIVE ── */
  .card{background:#111820;border:1px solid var(--border);border-radius:8px;
        padding:12px;text-align:center;}
  .card .val{font-size:36px;font-weight:700;line-height:1;font-family:'Share Tech Mono',monospace;}
  .card .lbl{font-size:10px;letter-spacing:2px;color:var(--dim);margin-top:4px;text-transform:uppercase;}
  .cyan{color:var(--accent)}.green{color:var(--green)}.orange{color:var(--orange)}.white{color:#fff}
  .info-row{background:#111820;border:1px solid var(--border);border-radius:6px;
            padding:8px 12px;font-size:13px;display:flex;justify-content:space-between;}
  .info-row .key{color:var(--dim);}
  .info-row .val2{color:var(--accent);font-family:'Share Tech Mono',monospace;font-size:12px;}
  #stageBadge{text-align:center;padding:8px;border-radius:6px;font-weight:700;
              font-size:16px;letter-spacing:3px;text-transform:uppercase;
              border:1px solid;transition:all .2s;}
  #feedbackBox{background:#111820;border-left:3px solid var(--accent);border-radius:4px;
               padding:10px;font-size:12px;line-height:1.6;color:var(--text);
               min-height:50px;font-family:'Share Tech Mono',monospace;}
  .rep-table{width:100%;border-collapse:collapse;font-size:13px;}
  .rep-table td{padding:5px 6px;border-bottom:1px solid var(--border);}
  .rep-table td:last-child{text-align:right;font-family:'Share Tech Mono',monospace;color:var(--accent);}
  .rep-table tr.active td{color:#fff;font-weight:700;}
  .gauge-wrap{background:#111820;border:1px solid var(--border);border-radius:6px;padding:10px 12px;}
  .gauge-label{font-size:11px;color:var(--dim);letter-spacing:1px;margin-bottom:6px;text-transform:uppercase;}
  .gauge-track{height:8px;background:#1e2a38;border-radius:4px;overflow:hidden;}
  .gauge-fill{height:100%;background:var(--accent);border-radius:4px;transition:width .15s;}
  .gauge-val{font-size:18px;font-weight:700;font-family:'Share Tech Mono',monospace;margin-top:4px;}
  .reset-btn{width:100%;padding:11px;background:transparent;border:1px solid var(--red);
             border-radius:6px;color:var(--red);font-size:14px;font-weight:700;
             letter-spacing:1px;cursor:pointer;text-transform:uppercase;
             transition:background .2s,color .2s;font-family:'Rajdhani',sans-serif;}
  .reset-btn:hover{background:var(--red);color:#fff;}

  /* ── VIDEO ANALYSIS ── */
  .upload-area{border:2px dashed var(--border);border-radius:8px;padding:18px;
               text-align:center;cursor:pointer;transition:border-color .2s;
               background:#111820;}
  .upload-area:hover{border-color:var(--accent);}
  .upload-area input{display:none;}
  .upload-area .icon{font-size:28px;margin-bottom:6px;}
  .upload-area p{font-size:12px;color:var(--dim);}
  .analyze-btn{width:100%;padding:11px;background:var(--accent);border:none;
               border-radius:6px;color:#000;font-size:14px;font-weight:700;
               letter-spacing:1px;cursor:pointer;text-transform:uppercase;
               font-family:'Rajdhani',sans-serif;transition:opacity .2s;}
  .analyze-btn:disabled{opacity:.4;cursor:not-allowed;}
  .prog-bar{height:6px;background:#1e2a38;border-radius:3px;overflow:hidden;display:none;}
  .prog-fill{height:100%;background:var(--accent);border-radius:3px;transition:width .3s;width:0%;}
  #analysisResult{background:#111820;border:1px solid var(--border);border-radius:8px;
                  padding:12px;font-size:11px;line-height:1.7;display:none;
                  font-family:'Share Tech Mono',monospace;color:var(--text);
                  max-height:260px;overflow-y:auto;white-space:pre-wrap;}
  #analysisResult::-webkit-scrollbar{width:3px;}
  #analysisResult::-webkit-scrollbar-thumb{background:var(--border);}
  #analysisThumb{width:100%;border-radius:6px;display:none;margin-top:4px;
                 border:1px solid var(--border);}
  .stat-grid{display:none;grid-template-columns:1fr 1fr;gap:8px;}
  .stat-mini{background:#111820;border:1px solid var(--border);border-radius:6px;
             padding:10px;text-align:center;}
  .stat-mini .v{font-size:20px;font-weight:700;font-family:'Share Tech Mono',monospace;color:var(--accent);}
  .stat-mini .l{font-size:10px;color:var(--dim);letter-spacing:1px;text-transform:uppercase;margin-top:2px;}

  /* ── CHATBOT ── */
  .quick-btns{display:flex;flex-wrap:wrap;gap:4px;}
  .qbtn{background:#111820;border:1px solid var(--border);border-radius:4px;
        padding:4px 7px;font-size:11px;cursor:pointer;color:var(--dim);
        font-family:'Share Tech Mono',monospace;transition:border-color .2s,color .2s;}
  .qbtn:hover{border-color:var(--accent);color:var(--accent);}
  .chat-messages{flex:1;overflow-y:auto;display:flex;flex-direction:column;
                 gap:7px;padding-bottom:4px;min-height:180px;max-height:350px;}
  .chat-messages::-webkit-scrollbar{width:3px;}
  .chat-messages::-webkit-scrollbar-thumb{background:var(--border);}
  .msg{padding:9px 12px;border-radius:8px;font-size:12px;line-height:1.6;
       max-width:97%;font-family:'Share Tech Mono',monospace;
       white-space:pre-wrap;word-break:break-word;}
  .msg.bot{background:#111820;border:1px solid var(--border);align-self:flex-start;
           border-left:3px solid var(--accent);}
  .msg.user{background:#1a2f4a;border:1px solid #1e3a5f;align-self:flex-end;
            color:#60a5fa;text-align:right;}
  .msg.thinking{background:#111820;border:1px solid var(--border);
                align-self:flex-start;color:var(--dim);font-style:italic;}
  .chat-input-wrap{display:flex;gap:6px;padding-top:8px;
                   border-top:1px solid var(--border);flex-shrink:0;}
  #chatInput{flex:1;background:#111820;border:1px solid var(--border);
             border-radius:6px;padding:8px 10px;color:var(--text);
             font-size:12px;font-family:'Share Tech Mono',monospace;outline:none;}
  #chatInput:focus{border-color:var(--accent);}
  #sendBtn{padding:8px 13px;background:var(--accent);border:none;border-radius:6px;
           color:#000;font-weight:700;cursor:pointer;font-size:13px;transition:opacity .2s;}
  #sendBtn:disabled{opacity:.4;cursor:not-allowed;}
</style>
</head>
<body>

<header>
  <h1>🏋️ AI Fitness Trainer</h1>
  <div>
    <span class="sdot" id="dot"></span>
    <span id="stext">Starting camera…</span>
  </div>
</header>

<div class="main">

  <!-- VIDEO FEED -->
  <div class="video-area">
    <video id="rawVideo" autoplay playsinline muted></video>
    <canvas id="captureCanvas"></canvas>
    <img id="annotated" src="" alt="Waiting for first frame…">
    <div class="cam-hint">Stand 2–3 m back · Full body in frame</div>
  </div>

  <!-- SIDEBAR -->
  <div class="sidebar">

    <div class="tab-bar">
      <button class="tab-btn active" onclick="switchTab('live',this)">📹 Live</button>
      <button class="tab-btn" onclick="switchTab('video',this)">🎬 Video</button>
      <button class="tab-btn" onclick="switchTab('chat',this)">🤖 Chat</button>
    </div>

    <!-- ══════ TAB 1: LIVE ══════ -->
    <div class="tab-panel active" id="tab-live">
      <div style="display:flex;gap:8px;">
        <div class="card" style="flex:1">
          <div class="val white" id="reps">0</div><div class="lbl">Reps</div>
        </div>
        <div class="card" style="flex:1">
          <div class="val orange" id="cal">0.0</div><div class="lbl">Calories</div>
        </div>
      </div>
      <div class="card">
        <div class="val cyan" id="timer">00:00</div><div class="lbl">Session Time</div>
      </div>
      <div class="info-row">
        <span class="key">Exercise</span>
        <span class="val2" id="exercise">Detecting…</span>
      </div>
      <div class="info-row">
        <span class="key">Confidence</span>
        <span class="val2" id="conf">—</span>
      </div>
      <div id="stageBadge" style="color:#546e7a;border-color:#546e7a;">READY</div>
      <div class="gauge-wrap">
        <div class="gauge-label">Joint Angle</div>
        <div class="gauge-track"><div class="gauge-fill" id="angleFill" style="width:0%"></div></div>
        <div class="gauge-val" id="angleVal">—</div>
      </div>
      <div id="feedbackBox">Waiting for pose…</div>
      <table class="rep-table" id="repTable">
        <tr data-ex="squat"><td>Squat</td><td id="r-squat">0</td></tr>
        <tr data-ex="barbell biceps curl"><td>Curl</td><td id="r-barbell biceps curl">0</td></tr>
        <tr data-ex="push-up"><td>Push-up</td><td id="r-push-up">0</td></tr>
        <tr data-ex="shoulder press"><td>Press</td><td id="r-shoulder press">0</td></tr>
      </table>
      <button class="reset-btn" onclick="resetSession()">↺ Reset Session</button>
    </div>

    <!-- ══════ TAB 2: VIDEO ANALYSIS ══════ -->
    <div class="tab-panel" id="tab-video">
      <div style="font-size:11px;color:var(--dim);font-family:'Share Tech Mono',monospace;
                  line-height:1.5;">
        Upload a workout video → AI analyses every frame, counts reps and gives form feedback.
      </div>

      <div class="upload-area" onclick="document.getElementById('videoFile').click()">
        <input type="file" id="videoFile" accept="video/*" onchange="onVideoSelected(this)">
        <div class="icon">🎬</div>
        <p id="uploadLabel">Click to select workout video</p>
        <p style="font-size:11px;color:#546e7a;margin-top:3px;">MP4 · AVI · MOV</p>
      </div>

      <img id="analysisThumb" src="" alt="">

      <div class="stat-grid" id="statGrid">
        <div class="stat-mini"><div class="v" id="vReps">—</div><div class="l">Total Reps</div></div>
        <div class="stat-mini"><div class="v" id="vCal">—</div><div class="l">Calories</div></div>
        <div class="stat-mini"><div class="v" id="vExercise" style="font-size:13px;">—</div><div class="l">Exercise</div></div>
        <div class="stat-mini"><div class="v" id="vFrames">—</div><div class="l">Frames</div></div>
      </div>

      <div class="prog-bar" id="progBar">
        <div class="prog-fill" id="progFill"></div>
      </div>

      <button class="analyze-btn" id="analyzeBtn" onclick="analyzeVideo()" disabled>
        ▶ Analyse Video
      </button>

      <div id="analysisResult"></div>
    </div>

    <!-- ══════ TAB 3: CHATBOT ══════ -->
    <div class="tab-panel" id="tab-chat">
      <div class="quick-btns">
        <span class="qbtn" onclick="sq('Squat form tips')">Squat</span>
        <span class="qbtn" onclick="sq('Push-up form tips')">Push-up</span>
        <span class="qbtn" onclick="sq('Bicep curl tips')">Curl</span>
        <span class="qbtn" onclick="sq('Shoulder press tips')">Press</span>
        <span class="qbtn" onclick="sq('Give me a warm-up routine')">Warm-up</span>
        <span class="qbtn" onclick="sq('How many calories do I burn?')">Calories</span>
        <span class="qbtn" onclick="sq('Beginner workout plan')">Beginner</span>
        <span class="qbtn" onclick="sq('Recovery and rest day tips')">Recovery</span>
      </div>
      <div class="chat-messages" id="chatMessages">
        <div class="msg bot">👋 Hi! I'm FitBot — your AI fitness coach.

Ask me anything about exercise form, programming, warm-ups, recovery, or calories. Use the quick buttons above or type below!</div>
      </div>
      <div class="chat-input-wrap">
        <input id="chatInput" type="text" placeholder="Ask anything fitness…"
               onkeydown="if(event.key==='Enter')sendChat()">
        <button id="sendBtn" onclick="sendChat()">Send</button>
      </div>
    </div>

  </div>
</div>

<script>
// ─── TAB SWITCHING ────────────────────────
function switchTab(id, btn) {
  document.querySelectorAll('.tab-panel').forEach(p => p.classList.remove('active'));
  document.querySelectorAll('.tab-btn').forEach(b => b.classList.remove('active'));
  document.getElementById('tab-'+id).classList.add('active');
  btn.classList.add('active');
}

// ─── LIVE WEBCAM ─────────────────────────
const rawVideo = document.getElementById('rawVideo');
const cap      = document.getElementById('captureCanvas');
const ctx      = cap.getContext('2d');
const img      = document.getElementById('annotated');
const dot      = document.getElementById('dot');
let sessionStart = Date.now();
let sending = false;

async function startCamera() {
  try {
    const stream = await navigator.mediaDevices.getUserMedia(
      {video:{width:1280,height:720,facingMode:'user'},audio:false});
    rawVideo.srcObject = stream;
    rawVideo.onloadedmetadata = () => {
      cap.width = rawVideo.videoWidth;
      cap.height = rawVideo.videoHeight;
      dot.classList.add('live');
      document.getElementById('stext').textContent = 'Live · auto-analysing';
      setInterval(sendFrame, 120);
    };
  } catch(e) {
    document.getElementById('stext').textContent = 'Camera error: '+e.message;
  }
}

async function sendFrame() {
  if (sending || rawVideo.readyState < 2) return;
  sending = true;
  try {
    ctx.drawImage(rawVideo, 0, 0, cap.width, cap.height);
    const blob = await new Promise(r => cap.toBlob(r,'image/jpeg',0.82));
    const buf  = await blob.arrayBuffer();
    const resp = await fetch('/process_frame',{method:'POST',
      headers:{'Content-Type':'image/jpeg'},body:buf});
    const d = await resp.json();
    if (d.frame_b64) img.src = 'data:image/jpeg;base64,'+d.frame_b64;
    updateLive(d);
  } catch(e){console.warn(e);}
  finally { sending=false; }
}

const SS = {
  up:   {color:'#00e676',border:'#00e676',text:'UP   ↑'},
  down: {color:'#0096ff',border:'#0096ff',text:'DOWN ↓'},
  null: {color:'#546e7a',border:'#546e7a',text:'READY'},
  ready:{color:'#546e7a',border:'#546e7a',text:'READY'},
};
function updateLive(d) {
  document.getElementById('reps').textContent     = d.count;
  document.getElementById('cal').textContent      = d.calories;
  document.getElementById('exercise').textContent = d.exercise;
  document.getElementById('conf').textContent     = d.confidence;
  const ss = SS[d.stage||'ready']||SS['ready'];
  const b  = document.getElementById('stageBadge');
  b.textContent=ss.text; b.style.color=ss.color; b.style.borderColor=ss.border;
  const pct = Math.min((parseFloat(d.angle)||0)/180*100,100);
  document.getElementById('angleFill').style.width = pct+'%';
  document.getElementById('angleVal').textContent  = d.angle+'°';
  const fb = d.feedback||'';
  const fb_box = document.getElementById('feedbackBox');
  fb_box.textContent = fb||(d.exercise==='No pose detected'
    ? 'No pose — full body must be visible':'Detecting…');
  fb_box.style.borderLeftColor = fb.startsWith('OK')?'#00e676':fb.startsWith('FIX')?'#ff9100':'#00e5ff';
  for(const [ex,cnt] of Object.entries(d.rep_counts||{})){
    const el=document.getElementById('r-'+ex);
    if(el) el.textContent=cnt;
  }
  document.querySelectorAll('#repTable tr').forEach(tr=>{
    tr.classList.toggle('active',tr.dataset.ex===d.exercise);
  });
  const e=Math.floor((Date.now()-sessionStart)/1000);
  const m=Math.floor(e/60),s=e%60;
  document.getElementById('timer').textContent=
    String(m).padStart(2,'0')+':'+String(s).padStart(2,'0');
}
async function resetSession(){sessionStart=Date.now();await fetch('/reset');}

// ─── VIDEO ANALYSIS ──────────────────────
let selFile = null;
function onVideoSelected(inp) {
  if (!inp.files.length) return;
  selFile = inp.files[0];
  document.getElementById('uploadLabel').textContent = '✅ '+selFile.name;
  document.getElementById('analyzeBtn').disabled = false;
  const url=URL.createObjectURL(selFile);
  const v=document.createElement('video');
  v.src=url;
  v.addEventListener('loadeddata',()=>{ v.currentTime=0.1; });
  v.addEventListener('seeked',()=>{
    const c=document.createElement('canvas');
    c.width=v.videoWidth;c.height=v.videoHeight;
    c.getContext('2d').drawImage(v,0,0);
    const t=document.getElementById('analysisThumb');
    t.src=c.toDataURL('image/jpeg');t.style.display='block';
    URL.revokeObjectURL(url);
  });
}

async function analyzeVideo() {
  if (!selFile) return;
  const btn=document.getElementById('analyzeBtn');
  const pb=document.getElementById('progBar');
  const pf=document.getElementById('progFill');
  const res=document.getElementById('analysisResult');
  const sg=document.getElementById('statGrid');
  btn.disabled=true; btn.textContent='⏳ Analysing…';
  res.style.display='none'; sg.style.display='none';
  pb.style.display='block'; pf.style.width='5%';
  const b64=await new Promise((r,rj)=>{
    const fr=new FileReader();
    fr.onload=()=>r(fr.result.split(',')[1]);
    fr.onerror=rj;
    fr.readAsDataURL(selFile);
  });
  pf.style.width='25%';
  try {
    const resp=await fetch('/analyze_video',{method:'POST',
      headers:{'Content-Type':'application/json'},
      body:JSON.stringify({video_b64:b64,filename:selFile.name})});
    pf.style.width='95%';
    const d=await resp.json();
    pf.style.width='100%';
    setTimeout(()=>{pb.style.display='none';},400);
    if(d.error){
      res.textContent='❌ Error: '+d.error;
      res.style.display='block';
    } else {
      document.getElementById('vReps').textContent   = d.total_reps;
      document.getElementById('vCal').textContent    = d.total_calories;
      document.getElementById('vExercise').textContent = d.top_exercise||'—';
      document.getElementById('vFrames').textContent = d.total_frames;
      sg.style.display='grid';
      res.textContent = d.summary;
      res.style.display='block';
    }
  } catch(e){
    res.textContent='❌ '+e.message;
    res.style.display='block';
    pb.style.display='none';
  }
  btn.disabled=false; btn.textContent='▶ Analyse Again';
}

// ─── CHATBOT ─────────────────────────────
const cm=document.getElementById('chatMessages');
function addMsg(txt,role){
  const d=document.createElement('div');
  d.className='msg '+role; d.textContent=txt;
  cm.appendChild(d); cm.scrollTop=cm.scrollHeight;
  return d;
}
async function sendChat(){
  const inp=document.getElementById('chatInput');
  const msg=inp.value.trim(); if(!msg) return;
  inp.value='';
  document.getElementById('sendBtn').disabled=true;
  addMsg(msg,'user');
  const th=addMsg('Thinking…','thinking');
  try {
    const resp=await fetch('/chat',{method:'POST',
      headers:{'Content-Type':'application/json'},
      body:JSON.stringify({
        message:msg,
        context:{
          exercise:document.getElementById('exercise').textContent,
          reps:document.getElementById('reps').textContent,
          calories:document.getElementById('cal').textContent,
        }
      })});
    const d=await resp.json();
    th.remove(); addMsg(d.reply,'bot');
  } catch(e){ th.textContent='❌ '+e.message; th.className='msg bot'; }
  document.getElementById('sendBtn').disabled=false;
}
function sq(t){document.getElementById('chatInput').value=t; sendChat();}

startCamera();
</script>
</body>
</html>
"""

# ─────────────────────────────────────────
# FLASK ROUTES
# ─────────────────────────────────────────
app = Flask(__name__)
CORS(app)

@app.route('/')
def index():
    return render_template_string(HTML_PAGE)

@app.route('/process_frame', methods=['POST'])
def process_frame_route():
    raw   = request.get_data()
    nparr = np.frombuffer(raw, np.uint8)
    frame = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    if frame is None:
        return jsonify({'error':'bad frame'}), 400
    with state_lock:
        process_frame_bgr(frame)
        resp = _state_dict()
    return jsonify(resp)

@app.route('/reset')
def reset_route():
    with state_lock:
        state.update({
            'count':0,'stage':None,'last_feedback':'','rep_confirmed':False,
            'last_exercise':'','total_calories':0.0,'session_start':time.time(),
            'rep_counts':{a:0 for a in actions},'annotated_frame_b64':'',
        })
        state['angle_buffer'].clear()
    return 'OK'

def _state_dict():
    return {
        'count':      state['count'],
        'calories':   f"{state['total_calories']:.1f}",
        'exercise':   state['exercise'],
        'confidence': f"{state['confidence']:.0%}",
        'stage':      state['stage'],
        'angle':      f"{state['angle']:.1f}",
        'feedback':   state['last_feedback'],
        'rep_counts': dict(state['rep_counts']),
        'frame_b64':  state['annotated_frame_b64'],
    }

@app.route('/analyze_video', methods=['POST'])
def analyze_video_route():
    data = request.get_json()
    if not data or 'video_b64' not in data:
        return jsonify({'error':'No video data'}), 400
    try:
        video_bytes = base64.b64decode(data['video_b64'])
    except Exception:
        return jsonify({'error':'Invalid base64'}), 400

    fname  = data.get('filename','video.mp4')
    suffix = '.'+fname.rsplit('.',1)[-1] if '.' in fname else '.mp4'
    with tempfile.NamedTemporaryFile(delete=False, suffix=suffix) as f:
        f.write(video_bytes)
        tmp = f.name

    try:
        vpose = mp_pose.Pose(min_detection_confidence=0.5,
                             min_tracking_confidence=0.5, model_complexity=1)
        cap   = cv2.VideoCapture(tmp)
        fps   = cap.get(cv2.CAP_PROP_FPS) or 30
        SKIP  = max(1, int(fps//8))

        ex_votes  = {a:0 for a in actions}
        rep_cnt   = {a:0 for a in actions}
        stage_v   = {a:None for a in actions}
        rep_conf  = {a:False for a in actions}
        ang_buf   = deque(maxlen=5)
        last_ex   = None
        form_ok   = []
        form_fix  = []
        total_cal = 0.0
        processed = 0
        fidx      = 0

        while cap.isOpened():
            ret, frame = cap.read()
            if not ret: break
            fidx += 1
            if fidx % SKIP != 0: continue
            rgb     = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = vpose.process(rgb)
            processed += 1
            if not results.pose_landmarks: continue
            lm     = results.pose_landmarks.landmark
            coords = np.array([[l.x,l.y,l.z] for l in lm]).flatten()
            pred   = model.predict(coords.reshape(1,1,99), verbose=0)
            ex     = actions[np.argmax(pred)]
            ex_votes[ex] += 1
            raw_a  = get_angle(ex, lm)
            ang_buf.append(raw_a)
            ang = float(np.mean(ang_buf))
            if ex != last_ex:
                last_ex  = ex
                stage_v  = {a:None for a in actions}
                rep_conf = {a:False for a in actions}
                ang_buf.clear()
            dn, up = THRESHOLDS.get(ex,{}).get('down',90), THRESHOLDS.get(ex,{}).get('up',150)
            if ang <= dn:
                if stage_v[ex] != 'down':
                    stage_v[ex]  = 'down'
                    rep_conf[ex] = False
                    fb = get_form_feedback(ex, ang)
                    if fb.startswith('FIX'): form_fix.append(f"[{ex}] {fb[5:]}")
                    elif fb.startswith('OK'): form_ok.append(f"[{ex}] {fb[4:]}")
            elif ang >= up:
                if stage_v[ex]=='down' and not rep_conf[ex]:
                    rep_cnt[ex] += 1
                    stage_v[ex]  = 'up'
                    rep_conf[ex] = True
                    total_cal   += CALORIES_PER_REP.get(ex, 0.2)

        cap.release(); vpose.close()
        total_reps = sum(rep_cnt.values())
        top_ex     = max(ex_votes, key=ex_votes.get) if any(ex_votes.values()) else 'Unknown'

        lines = [
            "📹 VIDEO ANALYSIS COMPLETE",
            "──────────────────────────",
            f"Frames analysed  : {processed}",
            f"Top exercise     : {top_ex.title()}",
            f"Total reps       : {total_reps}",
            f"Est. calories    : {total_cal:.1f} kcal",
            "",
            "REP BREAKDOWN:",
        ]
        for a in actions:
            if rep_cnt[a] > 0:
                lines.append(f"  • {a.title()}: {rep_cnt[a]} reps")
        if form_ok:
            lines.append("\n✅ GOOD FORM MOMENTS:")
            for g in list(set(form_ok))[:3]:
                lines.append(f"  {g}")
        if form_fix:
            lines.append("\n⚠️  FORM TO IMPROVE:")
            for f_ in list(set(form_fix))[:5]:
                lines.append(f"  {f_}")
        if not form_ok and not form_fix:
            lines.append("\nℹ️  Ensure full body is visible for form analysis.")

        return jsonify({
            'total_reps':     total_reps,
            'total_calories': f"{total_cal:.1f}",
            'top_exercise':   top_ex.title(),
            'total_frames':   processed,
            'rep_counts':     rep_cnt,
            'summary':        '\n'.join(lines),
        })
    except Exception as e:
        import traceback
        return jsonify({'error': str(e), 'trace': traceback.format_exc()})
    finally:
        try: os.unlink(tmp)
        except: pass

@app.route('/chat', methods=['POST'])
def chat_route():
    data    = request.get_json()
    msg     = (data or {}).get('message','').strip()
    context = (data or {}).get('context', {})
    if not msg:
        return jsonify({'reply':'Please ask me something!'})
    ctx_str = ''
    if context.get('exercise') not in (None,'Detecting…','No pose detected',''):
        ctx_str = (f"[User is currently doing: {context['exercise']} | "
                   f"Reps: {context.get('reps','?')} | Cal: {context.get('calories','?')}]")
    full_msg = f"{ctx_str}\n{msg}".strip() if ctx_str else msg
    if gemini_chat:
        try:
            reply = gemini_chat.send_message(full_msg).text
        except Exception as e:
            reply = rule_based_chat(msg) + f"\n\n_(Gemini error: {e})_"
    else:
        reply = rule_based_chat(msg)
    return jsonify({'reply': reply})

# ─────────────────────────────────────────
# ENTRY POINT
# ─────────────────────────────────────────
if __name__ == '__main__':
    try:
        from pyngrok import ngrok
        public_url = ngrok.connect(5000)
        print("=" * 60)
        print(f"  🚀  Open this URL in your browser:")
        print(f"      {public_url}")
        print("=" * 60)
        print("  SIDEBAR TABS:")
        print("    📹 Live  — Real-time webcam (auto rep count)")
        print("    🎬 Video — Upload video for full analysis")
        print("    🤖 Chat  — Ask FitBot anything about fitness")
        print("=" * 60)
    except ImportError:
        print("pyngrok not installed: !pip install pyngrok -q")
        print("Running on http://localhost:5000")

    app.run(host='0.0.0.0', port=5000, threaded=True, use_reloader=False)
```
